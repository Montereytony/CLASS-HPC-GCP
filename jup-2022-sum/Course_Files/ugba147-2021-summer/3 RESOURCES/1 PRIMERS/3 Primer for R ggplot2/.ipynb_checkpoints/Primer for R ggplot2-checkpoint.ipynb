## Introduction

### Synopsis

This primer assumes that you are familiar with Jupyter, and data selection and descriptive statistics in R.

R ggplot2 is a library of functions for rendering a variety of plots. 

## Data

To retrieve the dataset stored as file `High-Tech Stocks.csv`, we use the `read.csv` function.  To confirm that the dataset has been retrieved and assigned, we use the `size` and `colnames` functions.

```r
data = read.csv("High-Tech Stocks.csv", header=TRUE)
size(data)
colnames(data)
```

## Layer-by-Layer Construction of a Visualization

A visualization comprises several layers, and is constructed by connecting certain layer-generating functions. 

### Start with a base layer.
The `ggplot` function generates a base layer, to which we will connect other layers.  A table can optionally be provided as a parameter - if it is provided, then that table is set to be the default table used by other layer-generating functions.  Here we specify `data` to be the default table.

```r
ggplot(data)
```

### Add a plot type layer.

To add a plot type layer, we connect the base layer to an appropriate layer-generating function using the `+` connector.  In this context, the `+` connector does not sum numbers in the usual way, but rather it connects layers.  To specify how various aesthetics vary with data, we provide the `aes` function as a parameter to the layer-generating function.

Here we connect the `ggplot` function to the `geom_line` function, and use the `aes` function to specify that the layer's "x" values should vary with `data`'s `Date` values, and "y" values with `data`'s `Apple.Return` (multilpied by 100) values.  Note that `Date` and `Apple.Return` are inferred to be columns of `data` because that was set to be the default table in the `ggplot` layer.

```r
ggplot(data) +
geom_line(aes(x=Date, y=Apple.Return*100))
```

### Add more plot type layers.

To add more plot type layers, we connect additional appropriate layer-generating functions using the `+` connector.

Here we connect the `ggplot` function to two `geom_line` functions.  In the first `geom_line` layer, the aesthetics "x" and "y" vary with `data`'s `Date` and `Apple.Return` (multiplied by 100) values, respectively. In the second `geom_line` layer, the aesthetics "x" and "y" vary with `data`'s `Date` and `SP.500.Return` (multiplied by 100) values.

```r
ggplot(data) +
geom_line(aes(x=Date, y=Apple.Return*100)) +
geom_line(aes(x=Date, y=SP.500.Return*100))
```

### Enhance plot type layers with aesthetics that do not vary with data.

To enhance a plot type layer with aesthetics that do not vary with data, we provide named parameters to a relevant layer-generating function.

Here we connect the `ggplot` function to two `geom_line` functions.

In the first `geom_line` layer, the aesthetics "x" and "y" vary with `data`'s `Date` and `Apple.Return` (multiplied by 100) values, respectively.  We provide the named parameter `color=PALETTE[2]`, which specifies that that layer should be colored red, regardless of how data varies.  Note that `PALETTE[2]` was assigned the code for red previously in the notebook setup.  Note also that this named parameter is _not_ a parameter of the `aes` function, but rather a parameter of the `geom_line` function.

In the second `geom_line` layer, the aesthetics "x" and "y" vary with `data`'s `Date` and `SP.500.Return` (multiplied by 100) values.  We provide named parameters `color="black"` and `linetype="dashed"`, which specifies that that layer should be colored black and drawn as a dashed line.  Again, these named parameters are _not_ parameters of the `aes` function, but rather parameters of the `geom_line` function.

```r
ggplot(data) +
geom_line(aes(x=Date, y=Apple.Return*100), color=PALETTE[2]) +
geom_line(aes(x=Date, y=SP.500.Return*100), color="black", linetype="dashed")
```

### Enhance plot type layers with aesthetics that automatically create a legend.

To enhance a plot type layer with aesthetics that automatically create a legend, we provide an additional parameter to the `aes` function.

Here in the first `geom_line` layer, we provide the named parameter `color="Apple"`, which specifies that that layer should be referred to by color and called Apple in a legend.  Note that this named parameter _is_ a parameter of the `aes` function, and not a parameter of the `geom_line` function.

In the second `geom_line` layer, we provide the named parameter `color="S&P 500"`, which specifies that that layer should be referred to by color and called S&P 500 in a legend.  Again, this named parameter _is_ a parameter of the `aes` function, and not a parameter of the `geom_line` function.  Note that `linetype=dashed` is still a parameter of the `geom_line` function, so it does not affect the legend.

```r
ggplot(data) +
geom_line(aes(x=Date, y=Apple.Return*100, color="Apple")) +
geom_line(aes(x=Date, y=SP.500.Return*100, color="S&P 500"), linetype="dashed")
```

### Add a layer to customize colors.

To add a layer to customize colors, we connect the layer-generating function `scale_color_manual`.

Here we provide the named parameter `values=c(Apple=PALETTE[2], "S&P 500"="black")`, which specifies that the layer referred to as Apple should be colored red (PALETTE[2]), and the layer referred to as S&P 500 should be colored black.

```r
ggplot(data) +
geom_line(aes(x=Date, y=Apple.Return*100, color="Apple")) +
geom_line(aes(x=Date, y=SP.500.Return*100, color="S&P 500")) +
scale_color_manual(values=c(Apple=PALETTE[2], "S&P 500"="black"))
```

### Add a title layer.

To add a title layer, we connect the layer-generating function `ggtitle`.

Here we provide the parameter `"Apple Returns"`, which specifies the text for the title.

```r
ggplot(data) + ggtitle("Apple Returns") +
geom_line(aes(x=Date, y=Apple.Return*100, color="Apple")) +
geom_line(aes(x=Date, y=SP.500.Return*100, color="S&P 500")) +
scale_color_manual(values=c(Apple=PALETTE[2], "S&P 500"="black"))
```

### Add a layer to customize axis labels.

To add a layer to customize axis labels, we connect the layer-generating functions `xlab` and `ylab`.

Here we provide the parameter `"date"` to the `xlab` function, and `"change (%/mo)"` to the `ylab` function, which specify the text of the axis labels.

```r
ggplot(data) + ggtitle("Apple Returns") + xlab("date") + ylab("change (%/mo)") +
geom_line(aes(x=Date, y=Apple.Return*100, color="Apple")) +
geom_line(aes(x=Date, y=SP.500.Return*100, color="S&P 500")) +
scale_color_manual(values=c(Apple=PALETTE[2], "S&P 500"="black"))
```

### Add a layer to customize axis limits.

To add a layer to customize axis limits, we connect the layer-generating functions `xlim` and `ylim`.

Here we provide the parameters `2000` and `2010` to the `xlim` function, which specifies minimum and maximum "x" axis values.  We provide the parameters `-60` and `60` to the `ylim` fuction, which specifies minimum and maximum "y" axis values.  Note that these functions allow you to effectively zoom in and out.

We provide named parameter `na.rm=TRUE` to the `geom_line` functions to prevent warning messages about observations not within the limits.

```r
ggplot(data) + ggtitle("Apple Returns") + xlab("date") + ylab("change (%/mo)") +
xlim(2000,2010) + ylim(-60,60) +
geom_line(aes(x=Date, y=Apple.Return*100, color="Apple"), na.rm=TRUE) +
geom_line(aes(x=Date, y=SP.500.Return*100, color="S&P 500"), na.rm=TRUE) +
scale_color_manual(values=c(Apple=PALETTE[2], "S&P 500"="black"))
```

## Some Examples

### Scatter Plot

Show a scatter plot of Apple returns vs time.

To visualize Apple returns as a scatter plot, we use `ggplot` and related functions.

`ggplot` initializes the visualization.  The parameter is a table that includes the columns of values to be visualized, in this case `data`.

`geom_point` creates a layer of points.  The 1st parameter indicates the points' locations.  We use the `aes` function to specify the points' locations, in this case `data`'s `Date` column of values taken as the points' x coordinates, and `data`'s `Apple.Return` column of values, each multiplied by 100, taken as the points' y coordinates.  The `color` parameter is the points' color.  Note, the column names are described not as strings, and so not enclosed within `"` `"`.  Note also, the R system can recognize many different strings as color names.

`ggtitle` creates a layer with a title.  The parameter is the title text.

`xlab` creates a layer with an x-axis label.  The parameter is the x-axis label text.

`ylab` creates a layer with a y-axis label.  The parameter is the y-axis label text.

```r
ggplot(data) + ggtitle("Apple Returns") + xlab("date") + ylab("change (%/mo)") +
geom_point(aes(x=Date, y=Apple.Return*100), color=PALETTE[2])
```

### Line Plot

Show a line plot of Apple returns vs time.

To visualize Apple returns as a line plot, we use the `geom_line` function instead of the `geom_point` function.

```r
ggplot(data) + ggtitle("Apple Returns") + xlab("date") + ylab("change (%/mo)") +
geom_line(aes(x=Date, y=Apple.Return*100), color=PALETTE[2])
```

### Histogram

Show a histogram of Apple returns.

To visualize Apple returns as a histogram, we use the `geom_histogram` function instead of the `geom_points` or `geom_ line` function.  The 1st parameter is the histogram values.  We use the `aes` function to specify the histogram values, in this case `data`'s `Apple.Returns` column of values, each multiplied by 100.  The `bins` parameter is the histogram's number of bins.  The `color` parameter is the bars' edge color.  The `fill` parameter is the bars' interior color.  Note, no y-axis label need be specified - it defauts to "count".

```r
ggplot(data) + ggtitle("Apple Returns") + xlab("change (%/mo)") +
geom_histogram(aes(x=Apple.Return*100), bins=50, fill=PALETTE[2])
```

### Scatter Plot to Visualize Correlation

Show the correlation between Apple returns and Dell returns as a scatter plot.

To visualize the correlation between Apple returns and Dell returns as a scatter plot, we use the `geom_point` function, but reference the `Apple.Returns` and `Dell.Returns` columns instead of the `Date` and `Apple.Returns` columns.

Note, the `ggtitle` function can take 2 parameters.  If present, the 2nd parameter is the subtitle text.

```r
ggplot(data) +
ggtitle("Correlation of Returns", "Apple vs Dell") + xlab("Apple change (%/mo)") + ylab("Dell change (%/mo)") +
geom_point(aes(x=Apple.Return*100, y=Dell.Return*100))
```

### Line Plot with 2 Lines

Compare Apple returns to S&P 500 returns vs. time as a line plot.

To visualize Apple returns and Dell returns as a line plot, we use the `geom_line` function twice.

The 1st `geom_line` creates a layer with a line of connected points.  The 1st parameter indicates the points' locations and their legend label.  We use the `aes` function to specify the points' locations and their legend label.  Note, the `color` parameter of the `aes` function is the legend label, not to be confused with the `color` parameter of the `geom_line` function that we used earlier.

The 2nd `geom_line` creates another layer with a line of connected points.

`scale_color_manual` creates a layer with a map of which legend lables go with which colors.  The `values` parameter is a vector of color names, each element named by its associated legend label.

Because (in the `geom_line` functions) `aes` functions have been used with `color` parameters, the visualization will automatically include a legend.

```r
ggplot(data) + ggtitle("Apple Returns") + xlab("date") + ylab("change (%/mo)") +
geom_line(aes(x=Date, y=Apple.Return*100, color="Apple")) +
geom_line(aes(x=Date, y=SP.500.Return*100, color="S&P 500")) +
scale_color_manual(values=c(Apple=PALETTE[2], "S&P 500"="black"))
```

### Line Plot with Multiple Lines

Compare the 4 tech returns to S&P 500 returns as a line plot vs time.

To visualize all returns as a line plot, we use the `geom_line` function once for each column of returns.

```r
ggplot(data) + ggtitle("Returns") + xlab("date") + ylab("change (%/mo)") +
geom_line(aes(x=Date, y=Apple.Return*100, color="Apple")) +
geom_line(aes(x=Date, y=Dell.Return*100, color="Dell")) +
geom_line(aes(x=Date, y=IBM.Return*100, color="IBM")) +
geom_line(aes(x=Date, y=Microsoft.Return*100, color="Microsoft")) +
geom_line(aes(x=Date, y=SP.500.Return*100, color="S&P 500")) +
scale_color_manual(values=c(Apple=PALETTE[2], Dell=PALETTE[4], IBM=PALETTE[1], Microsoft=PALETTE[8], "S&P 500"="black"))
```

### Grid-Arranged Plots

Show the correlations between the 4 tech returns as multiple grid-arranged scatterplots.

To size the upcoming visualization, we use the `output_size` function.

To present several visualizations in a grid layout, we first create the individual visualizations and assign each a name.  We use the `xlim` and `ylim` functions to ensure that all the visualization scales match.

Then we use the `grid.arrange` function to specify where the individual visualizations fit in the grid.  The first several parameters are the individual visualizations, in the order in which we want them to be presented.  The `nrow` parameter is the number of rows in the grid.  The `ncol` parameter is the number of columns in the grid.  Note, the `nrow` and `ncol` parameters are not related to any of our table's rows and columns.

To restore the size setting for other visualizations, we use the `output_size` function again, this time providing the parameter `restore`.  Note that `restore` was assigned the default size setting previously in the notebook setup. 

```r
output_size(8,5)

apple.dell = ggplot(data) + ggtitle("Apple vs Dell") + xlab("Apple change (%/mo)") + ylab("Dell change (%/mo)") +
             xlim(-60,60) + ylim(-60,60) +
             geom_point(aes(x=Apple.Return*100, y=Dell.Return*100))

apple.ibm = ggplot(data) + ggtitle("Apple vs IBM") + xlab("Apple change (%/mo)") + ylab("IBM change (%/mo)") +
            xlim(-60,60) + ylim(-60,60) +
            geom_point(aes(x=Apple.Return*100, y=IBM.Return*100))

apple.microsoft = ggplot(data) + ggtitle("Apple vs Microsoft") + xlab("Apple change (%/mo)") + ylab("Microsoft change (%/mo)") +
                  xlim(-60,60) + ylim(-60,60) +
                  geom_point(aes(x=Apple.Return*100, y=Microsoft.Return*100))

dell.ibm = ggplot(data) + ggtitle("Dell vs IBM") + xlab("Dell change (%/mo)") + ylab("IBM change (%/mo)") +
           xlim(-60,60) + ylim(-60,60) +
           geom_point(aes(x=Dell.Return*100, y=IBM.Return*100))

dell.microsoft = ggplot(data) +  ggtitle("Dell vs Microsoft") + xlab("Dell change (%/mo)") + ylab("Microsoft change (%/mo)") +
                 xlim(-60,60) + ylim(-60,60) +
                 geom_point(aes(x=Dell.Return*100, y=Microsoft.Return*100))

ibm.microsoft = ggplot(data) + ggtitle("IBM vs Microsoft") + xlab("IBM change (%/mo)") + ylab("Microsoft change (%/mo)") +
                xlim(-60,60) + ylim(-60,60) +
                geom_point(aes(x=IBM.Return*100, y=Microsoft.Return*100))

grid.arrange(apple.dell, apple.ibm, apple.microsoft, dell.ibm, dell.microsoft, ibm.microsoft, nrow=2, ncol=3)

output_size(restore)
```

### Data Visualization with Special Functions

Show the distributions of the 5 returns and the correlations between them.

To size the upcoming visualization, we use the `output_size` function.

To visualize all returns correlations as scatter plots (and kernel density estimate plots) in a grid layout, we use the `ggscatmat` function.  Note, `ggplot` is not needed because `ggscatmat` initializes the visualization itself.

To restore the size setting for other visualizations, we use the `output_size` function again, this time providing the parameter `restore`.  Note that `restore` was assigned the default size setting previously in the notebook setup. 

```r
output_size(5,5)

ggscatmat(data[,c(2:5,7)]) +
ggtitle("Correlation of Returns")

output_size(restore)
```

## More About R

About naming vector elements:<br>
The `c` function is used to create a vector of values, e.g., `x = c("brown2", "black")` results in a vector of 2 elements, the 1st element value is "brown2", the 2nd element value is "black".  Vector elements can be named with `=` notation, e.g., `x = c("Apple"="brown2", "S&P 500"="black")` results in a vector of 2 elements, the 1st element is named "Apple", the 2nd element is named "S&P 500".  Then, specific elements can be referenced by name, e.g., `x["Apple"]` has the value "brown2", and `x["S&P 500"]` has the value "black".
