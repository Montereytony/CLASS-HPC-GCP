```r
cat("Open code cell here for notebook apparatus.")

options(warn=-1)

# Load some required functions
library(plyr,      verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(ggplot2,   verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(gridExtra, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)

# Set some visualization constants
restore = list(repr.plot.width=8, repr.plot.height=3)
PALETTE = c("#00A9FF", "#F8766D", "#7CAE00", "#C77CFF", "#CD9600", "#00BE67", "#FF61CC", "#00BFC4")
theme.x_axis_only = theme(axis.title.y=element_blank(), axis.text.y=element_blank(), axis.ticks.y=element_blank(), panel.grid.major.y=element_blank(), panel.grid.minor.y=element_blank())
theme.no_axis_titles = theme(axis.title.x=element_blank(), axis.title.y=element_blank())
theme.x_axis_45 = theme(axis.text.x=element_text(angle=45, hjust=1, vjust=1))
theme.no_legend = theme(legend.position="none")

# define some useful functions
size = function(data) data.frame(observations=dim(data)[1], variables=dim(data)[2])
num_columns = function(data) aaply(names(data), 1, function(j) { class(data[, j]) != "factor" })
full_columns = function(data) aaply(names(data), 1, function(j) { !any(is.na(data[, j])) })
vary_columns = function(data) aaply(names(data), 1, function(j) { length(unique(data[, j])) > 1 })
sd_columns = function(data) aaply(1:ncol(data), 1, function(j) sd(data[,j]))
var_columns = function(data) aaply(1:ncol(data), 1, function(j) var(data[,j]))                                   
pc_constituents = function(pc) { pc_names = names(as.data.frame(pc$rotation))
                                 qual = as.data.frame(t(adply(pc_names, 1, function(j) row.names(pc$rotation)[rev(order(abs(pc$rotation[,j])))])[-1]))
                                 names(qual) = pc_names
                                 row.names(qual) = NULL
                                 qual }

# Set some visualization formatting defaults
options(repr.matrix.max.cols=74)
options(restore)
update_geom_defaults("point", list(size=0.01, colour=PALETTE[1]))
update_geom_defaults("vline", list(color="black", size=0.15))
update_geom_defaults("col", list(fill=PALETTE[1]))
update_geom_defaults("density", list(size=3, fill=PALETTE[1], colour=NA))
theme_update(plot.title=element_text(size=10, hjust=0.5), plot.subtitle=element_text(size=8, face="italic", hjust=0.5), axis.title=element_text(size=7), axis.text=element_text(size=7), strip.text=element_text(size=7), strip.text.y=element_text(angle=90), legend.title=element_blank(), legend.text=element_text(size=7))
```

## Situation

LendingTree is an online lending exchange that connects consumers with multiple lenders, banks, and credit partners who compete for business.  Since being founded in 1998 LendingTree has facilitated more than 32 million loan requests.

* **Role:** Banker. 
* **Business Decision:** Lend to whom?
* **Approach:** Use kernel density estimation and principal component transformation to look for features that distinguish known good vs. bad loans, and use that insight to inform decisions about new loans.  
* **Dataset:** Lending Tree Loans 2007-2010 ()

## Select Data 

### Retrieve data.

```r
data.all = read.csv("Lending Tree Loans.csv") # may take about 2 minutes
size(data.all)
head(data.all)
```

### Restrict analysis to inactive loans.  

Note which loans whether or not they were paid back.

```r
as.data.frame(unique(data.all$loan_status))
```

```r
good = which(data.all$loan_status %in% c("Fully Paid", "Does not meet the credit policy. Status:Fully Paid"))
data.frame(good_observations=length(good))
```

```r
bad = which(data.all$loan_status %in% c("Default", "Charged Off", "Does not meet the credit policy. Status:Charged Off"))
data.frame(bad_observations=length(bad))
```

### Restrict analysis to some convenient, well-represented variables.

```r
str(data.all[c(good,bad),])
```

Which columns are numeric and which are categorical?

```r
n = num_columns(data.all[c(good,bad),])
as.vector(which(n))
as.vector(which(!n))
```

Which columns are complete and which are incomplete (have missing data)?

```r
f = full_columns(data.all[c(good,bad),])
as.vector(which(f))
as.vector(which(!f))
```

Which columns have at least some variation and which have no variation (all the same value)?

```r
v = vary_columns(data.all[c(good,bad),])
as.vector(which(v))
as.vector(which(!v))
```

Just the numeric, varying columns without any missing values.  Also omit the ID and Member ID variables.

```r
data = data.all[c(good,bad), n & f & v]
data = data[, -which(names(data) %in% c("id","member_id"))]

data.good = head(data, length(good))
data.bad  = tail(data, length(bad))

size(data)
head(data)
```

## Explore Distributions

### Scree plot of variables.

```r
variable = names(data)
sdev = sd_columns(data)
variance = var_columns(data)
cum_variance = cumsum(variance)
relative_variance = variance / sum(variance)
cum_relative_variance = cumsum(relative_variance)

scree = data.frame(variable, sdev, variance, cum_variance, relative_variance, cum_relative_variance)
scree

ggplot(scree) + ylim(0,1) + geom_col(aes(x=factor(variable, levels=variable), y=relative_variance, fill=variable)) +
theme.x_axis_45 + theme.no_legend
```

### 1D Scatter plots of 2 variables.

```r
options(repr.plot.width=8, repr.plot.height=1)
p1 = ggplot(data) + ylim(-1,1) + geom_point(aes(x=loan_amnt), y=0) + theme.x_axis_only
p2 = ggplot(data) + ylim(-1,1) + geom_point(aes(x=int_rate), y=0) + theme.x_axis_only
grid.arrange(p1, p2, nrow=1)
options(restore)
```

### Probability density functions of 2 variables.

```r
p1 = ggplot(data) + ggtitle("PDF for loan_amnt") +
     geom_density(aes(x=loan_amnt), kernel="gaussian", bw=1000, fill=PALETTE[1]) +
     theme.no_axis_titles

p2 = ggplot(data) + ggtitle("PDF for int_rate") +
     geom_density(aes(x=int_rate), kernel="gaussian", bw=0.5, fill=PALETTE[1]) +
     theme.no_axis_titles

grid.arrange(p1, p2, nrow=1)
```

### 2D scatter plot of 2 variables.

```r
ggplot(data) + geom_point(aes(x=loan_amnt, y=int_rate))
```

## How Are Observations Distinguished?

**Probabilities of variable values in certain ranges.**

```r
d.loan_amnt = density(data$loan_amnt, kernel="gaussian", bw=1000, from=0, to=40000)
d.int_rate = density(data$int_rate, kernel="gaussian", bw=0.5, from=0, to=30)

pdf.loan_amnt = approxfun(d.loan_amnt)
pdf.int_rate = approxfun(d.int_rate)

data.frame(variable=c("loan_amnt","int_rate"),
           probability=c(integrate(pdf.loan_amnt, 0, 18000)$value, integrate(pdf.int_rate, 10, 20)$value),
           range_low=c(0, 10),
           range_high=c(18000, 20))

p1 = ggplot(data) + xlim(0, 40000) + ggtitle("PDF for loan_amnt") +
     stat_function(fun=pdf.loan_amnt, geom="area", fill=PALETTE[1]) +
     stat_function(fun=pdf.loan_amnt, geom="area", xlim=c(0,18000), fill=PALETTE[5]) +
     geom_vline(xintercept=0) + geom_vline(xintercept=18000) +
     theme.no_axis_titles

p2 = ggplot(data) + xlim(0, 30) + ggtitle("PDF for int_rate") +
     stat_function(fun=pdf.int_rate, geom="area", fill=PALETTE[1]) +
     stat_function(fun=pdf.int_rate, geom="area", xlim=c(10,20), fill=PALETTE[5]) +
     geom_vline(xintercept=10) + geom_vline(xintercept=20) +
     theme.no_axis_titles

grid.arrange(p1, p2, nrow=1)
```

### Probabilities of variable values in certain ranges, given good loan.

```r
d.loan_amnt = density(data.good$loan_amnt, kernel="gaussian", bw=1000, from=0, to=40000)
d.int_rate = density(data.good$int_rate, kernel="gaussian", bw=0.5, from=0, to=30)

pdf.loan_amnt = approxfun(d.loan_amnt)
pdf.int_rate = approxfun(d.int_rate)

data.frame(variable=c("loan_amnt","int_rate"),
           probability=c(integrate(pdf.loan_amnt, 0, 18000)$value, integrate(pdf.int_rate, 10, 20)$value),
           range_low=c(0, 10),
           range_high=c(18000, 20))

p1 = ggplot(data.good) + xlim(0, 40000) + ggtitle("PDF for loan_amnt") +
     stat_function(fun=pdf.loan_amnt, geom="area", fill=PALETTE[3]) +
     stat_function(fun=pdf.loan_amnt, geom="area", xlim=c(0,18000), fill=PALETTE[5]) +
     geom_vline(xintercept=0) + geom_vline(xintercept=18000) +
     theme.no_axis_titles

p2 = ggplot(data.good) + xlim(0, 30) + ggtitle("PDF for int_rate") +
     stat_function(fun=pdf.int_rate, geom="area", fill=PALETTE[3]) +
     stat_function(fun=pdf.int_rate, geom="area", xlim=c(10,20), fill=PALETTE[5]) +
     geom_vline(xintercept=10) + geom_vline(xintercept=20) +
     theme.no_axis_titles

grid.arrange(p1, p2, nrow=1)
```

### Probabilities of variable values in certain ranges, given bad loan.

```r
d.loan_amnt = density(data.bad$loan_amnt, kernel="gaussian", bw=1000, from=0, to=40000)
d.int_rate = density(data.bad$int_rate, kernel="gaussian", bw=0.5, from=0, to=30)

pdf.loan_amnt = approxfun(d.loan_amnt)
pdf.int_rate = approxfun(d.int_rate)

data.frame(variable=c("loan_amnt","int_rate"),
           probability=c(integrate(pdf.loan_amnt, 0, 18000)$value, integrate(pdf.int_rate, 10, 20)$value),
           range_low=c(0, 10),
           range_high=c(18000, 20))

p1 = ggplot(data.bad) + xlim(0, 40000) + ggtitle("PDF for loan_amnt") +
     stat_function(fun=pdf.loan_amnt, geom="area", fill=PALETTE[2]) +
     stat_function(fun=pdf.loan_amnt, geom="area", xlim=c(0,18000), fill=PALETTE[5]) +
     geom_vline(xintercept=0) + geom_vline(xintercept=18000) +
     theme.no_axis_titles

p2 = ggplot(data.bad) + xlim(0, 30) + ggtitle("PDF for int_rate") +
     stat_function(fun=pdf.int_rate, geom="area", fill=PALETTE[2]) +
     stat_function(fun=pdf.int_rate, geom="area", xlim=c(10,20), fill=PALETTE[5]) +
     geom_vline(xintercept=10) + geom_vline(xintercept=20) +
     theme.no_axis_titles

grid.arrange(p1, p2, nrow=1)
```

## Transform to Principal Components

### Data expressed in terms of principal components.

```r
pc = prcomp(data, scale=TRUE, retx=TRUE)
```

```r
data.pc = as.data.frame(pc$x)

data.pc.good = head(data.pc, length(good))
data.pc.bad  = tail(data.pc, length(bad))

size(data.pc)
head(data.pc)
```

### Qualitative interpretation of principal components.

Each column contains variable names sorted by weight applied to the principal component.

```r
pc_constituents(pc)
```

### Scree plot of principal components.

```r
variable = names(data.pc)
sdev = sd_columns(data.pc)
variance = var_columns(data.pc)
cum_variance = cumsum(variance)
relative_variance = variance / sum(variance)
cum_relative_variance = cumsum(relative_variance)

scree.pc = data.frame(variable, sdev, variance, cum_variance, relative_variance, cum_relative_variance)
scree.pc

ggplot(scree.pc) + ylim(0,1) +
geom_col(aes(x=factor(variable, levels=variable), y=relative_variance, fill=variable)) +
theme.no_legend
```

## How Are Transformed Observations Distinguished?

### Probabilities of principal component values in certain ranges, given good loan.

```r
d.PC1 = density(data.pc.good$PC1, kernel="gaussian", bw=0.2, from=-5, to=5)
d.PC2 = density(data.pc.good$PC2, kernel="gaussian", bw=0.2, from=-5, to=5)

pdf.PC1 = approxfun(d.PC1)
pdf.PC2 = approxfun(d.PC2)

data.frame(variable=c("PC1","PC2"),
           probability=c(integrate(pdf.PC1, -5, 1)$value, integrate(pdf.PC2, -0.8, 0)$value),
           range_low=c(-5, -0.8),
           range_high=c(1, 0))


p1 = ggplot(data.pc.good) + xlim(-5, 5) + ggtitle("PDF for PC1") +
     stat_function(fun=pdf.PC1, geom="area", fill=PALETTE[3]) +
     stat_function(fun=pdf.PC1, geom="area", xlim=c(-5,1), fill=PALETTE[5]) +
     geom_vline(xintercept=-5) + geom_vline(xintercept=1) +
     theme.no_axis_titles

p2 = ggplot(data.pc.good) + xlim(-5, 5) + ggtitle("PDF for PC2") +
     stat_function(fun=pdf.PC2, geom="area", fill=PALETTE[3]) +
     stat_function(fun=pdf.PC2, geom="area", xlim=c(-0.8,0), fill=PALETTE[5]) +
     geom_vline(xintercept=-0.8) + geom_vline(xintercept=0) +
     theme.no_axis_titles

grid.arrange(p1, p2, nrow=1)
```

### Probabilities of principal component values in certain ranges, given bad loan.

```r
d.PC1 = density(data.pc.bad$PC1, kernel="gaussian", bw=0.2, from=-5, to=5)
d.PC2 = density(data.pc.bad$PC2, kernel="gaussian", bw=0.2, from=-5, to=5)

pdf.PC1 = approxfun(d.PC1)
pdf.PC2 = approxfun(d.PC2)

data.frame(variable=c("PC1","PC2"),
           probability=c(integrate(pdf.PC1, -5, 1)$value, integrate(pdf.PC2, -0.8, 0)$value),
           range_low=c(-5, -0.8),
           range_high=c(1, 0))

p1 = ggplot(data.pc.bad) + xlim(-5, 5) + ggtitle("PDF for PC1") +
     stat_function(fun=pdf.PC1, geom="area", fill=PALETTE[2]) +
     stat_function(fun=pdf.PC1, geom="area", xlim=c(-5,1), fill=PALETTE[5]) +
     geom_vline(xintercept=-5) + geom_vline(xintercept=1) +
     theme.no_axis_titles

p2 = ggplot(data.pc.bad) + xlim(-5, 5) + ggtitle("PDF for PC2") +
     stat_function(fun=pdf.PC2, geom="area", fill=PALETTE[2]) +
     stat_function(fun=pdf.PC2, geom="area", xlim=c(-0.8,0), fill=PALETTE[5]) +
     geom_vline(xintercept=-0.8) + geom_vline(xintercept=0) +
     theme.no_axis_titles

grid.arrange(p1, p2, nrow=1)
```

```r
d.PC1 = density(data.pc.good$PC1, kernel="gaussian", bw=0.2, from=-5, to=5)
d.PC2 = density(data.pc.good$PC2, kernel="gaussian", bw=0.2, from=-5, to=5)

pdf.PC1 = approxfun(d.PC1)
pdf.PC2 = approxfun(d.PC2)

data.frame(variable=c("PC1","PC2"),
           probability=c(integrate(pdf.PC1, -5, 0)$value, integrate(pdf.PC2, -0.3, 2)$value),
           range_low=c(-5, -0.3),
           range_high=c(0, 2))

p1 = ggplot(data.pc.good) + xlim(-5, 5) + ggtitle("PDF for PC1") +
     stat_function(fun=pdf.PC1, geom="area", fill=PALETTE[3]) +
     stat_function(fun=pdf.PC1, geom="area", xlim=c(-5,0), fill=PALETTE[5]) +
     geom_vline(xintercept=-5) + geom_vline(xintercept=0) +
     theme.no_axis_titles

p2 = ggplot(data.pc.good) + xlim(-5, 5) + ggtitle("PDF for PC2") +
     stat_function(fun=pdf.PC2, geom="area", fill=PALETTE[3]) +
     stat_function(fun=pdf.PC2, geom="area", xlim=c(-0.3,2), fill=PALETTE[5]) +
     geom_vline(xintercept=-0.3) + geom_vline(xintercept=2) +
     theme.no_axis_titles

grid.arrange(p1, p2, nrow=1)
```

```r
d.PC1 = density(data.pc.bad$PC1, kernel="gaussian", bw=0.2, from=-5, to=5)
d.PC2 = density(data.pc.bad$PC2, kernel="gaussian", bw=0.2, from=-5, to=5)

pdf.PC1 = approxfun(d.PC1)
pdf.PC2 = approxfun(d.PC2)

data.frame(variable=c("PC1","PC2"),
           probability=c(integrate(pdf.PC1, -5, 0)$value, integrate(pdf.PC2, -0.3, 2)$value),
           range_low=c(-5, -0.3),
           range_high=c(0, 2))

p1 = ggplot(data.pc.bad) + xlim(-5, 5) + ggtitle("PDF for PC1") +
     stat_function(fun=pdf.PC1, geom="area", fill=PALETTE[2]) +
     stat_function(fun=pdf.PC1, geom="area", xlim=c(-5,0), fill=PALETTE[5]) +
     geom_vline(xintercept=-5) + geom_vline(xintercept=0) +
     theme.no_axis_titles

p2 = ggplot(data.pc.bad) + xlim(-5, 5) + ggtitle("PDF for PC2") +
     stat_function(fun=pdf.PC2, geom="area", fill=PALETTE[2]) +
     stat_function(fun=pdf.PC2, geom="area", xlim=c(-0.3,2), fill=PALETTE[5]) +
     geom_vline(xintercept=-0.3) + geom_vline(xintercept=2) +
     theme.no_axis_titles

grid.arrange(p1, p2, nrow=1)
```

### 2D scatter plots of 2 variables, overlayed by type of loan.

Distinguishing features are revealed.

```r
ggplot() + xlim(-5,12) + ylim(-5,20) +
geom_point(aes(x=PC1, y=PC2), data=data.pc.good, color=PALETTE[3], alpha=0.2) +
geom_point(aes(x=PC1, y=PC2), data=data.pc.bad,  color=PALETTE[2], alpha=0.2)
```
