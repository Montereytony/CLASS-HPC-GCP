## Lineplot

Visualize many variable distributions; treat 1 variable as a shared sequence control.

### Show 2 Variable Distributions

```r
# Show 2 variables

data.frame(variable=c("x1","x2"), aspect=c("horizontal position","vertical position"), control=c("yes",""))

output_size(3,3)
ggplot(data) + xlim(0,6) + ylim(0,6) + ggtitle("Lineplot") +
geom_line(aes(x=x1, y=x2))
output_size(restore)
```

### Show 4 Variable Distributions

Color indicates which variable, not the variable value.  Pattern indicates which variable, not the variable value.

```r
# Show 4 variables (3 variables distinguished by color or pattern)

data.frame(variable=c("x1","x2","x6","x7"),
           aspect=c("horizontal position","vertical position","vertical position","vertical position"),
           control=c("yes","","",""))

p1 = ggplot(data) +  xlim(0,6) + ylim(0,6) + ylab("") + ggtitle("Lineplot, variables coded by color") +
     geom_line(aes(x=x1, y=x2, color="x2")) +
     geom_line(aes(x=x1, y=x6, color="x6")) +
     geom_line(aes(x=x1, y=x7, color="x7")) +
     scale_color_manual(values=PALETTE[1:3])

p2 = ggplot(data) + xlim(0,6) + ylim(0,6) + ylab("") + ggtitle("Lineplot, variables coded by pattern") +
     geom_line(aes(x=x1, y=x2, lty="x2")) +
     geom_line(aes(x=x1, y=x6, lty="x6")) +
     geom_line(aes(x=x1, y=x7, lty="x7")) +
     scale_linetype_manual(values=c("solid","longdash","dotdash"))

grid.arrange(p1, p2, nrow=1)
```

### Show 2 Variable Distributions | 1 Categorical, 1 Numeric

Useful when categorical variable represents a sequence.

```r
# Show 2 variables; horizontal position is categorical, so use group=1 in geom_line

data.frame(variable=c("x4","x2"), aspect=c("horizontal position","vertical position"), control=c("yes",""), type=c("categorical","numeric"))

output_size(3,3)
ggplot(data) + ylim(0,6) + ggtitle("Lineplot") +
geom_line(aes(x=x4, y=x2, group=1))
output_size(restore)
```

### Show 4 Variable Distributions | 1 Categorical, Others Numeric

Useful when comparing several numeric variable distributions across a categorical variable distribution, even if categorical variable does not represent a sequence. 

Color indicates which variable, not the variable value.  Pattern indicates which variable, not the variable value.

```r
# Show 4 variables (3 variables distinguished by color or pattern); horizontal position is categorical
# Use group=1 in geom_line

data.frame(variable=c("x4","x2","x6","x7"),
           aspect=c("horizontal position","vertical position","vertical position","vertical position"),
           control=c("yes","","",""),
           type=c("categorical", "numeric", "numeric", "numeric"))

p1 = ggplot(data) + ylim(0,6) + ylab("") + ggtitle("Lineplot, variables coded by color") +
     geom_line(aes(x=x4, y=x2, color="x2", group=1)) +
     geom_line(aes(x=x4, y=x6, color="x6", group=1)) +
     geom_line(aes(x=x4, y=x7, color="x7", group=1)) +
     scale_color_manual(values=PALETTE[1:3])

p2 = ggplot(data) + ylim(0,6) + ylab("") + ggtitle("Lineplot, variables coded by path") +
     geom_line(aes(x=x4, y=x2, lty="x2", group=1)) +
     geom_line(aes(x=x4, y=x6, lty="x6", group=1)) +
     geom_line(aes(x=x4, y=x7, lty="x7", group=1)) +
     scale_linetype_manual(values=c("solid","longdash","dotdash"))

grid.arrange(p1, p2, nrow=1)
```

## Stepplot

Visualize many variable distributions; treat 1 variable as a shared sequence control.

### Show 4 Variable Distributions

Adjust for overlap with transparencies (alpha) or nudges.  Color indicates which variable, not the variable value.  Pattern indicates which variable, not the variable value.

```r
# Show 4 variables (3 variables distinguished by color or pattern)
# Adjust for overlap with transparencies (alpha)
# Adjust for overlap with nudges

data.frame(variable=c("x1","x2","x6","x7"),
           aspect=c("horizontal position","vertical position","vertical position","vertical position"),
           control=c("yes","","",""))

p1 = ggplot(data) +  xlim(0,6) + ylim(0,6) + ylab("") + ggtitle("Stepplot, variables coded by color") +
     geom_step(aes(x=x1, y=x2, color="x2"), alpha=0.5) +
     geom_step(aes(x=x1, y=x6, color="x6"), alpha=0.5) +
     geom_step(aes(x=x1, y=x7, color="x7"), alpha=0.5) +
     scale_color_manual(values=PALETTE[1:3])

p2 = ggplot(data) + xlim(0,6) + ylim(0,6) + ylab("") + ggtitle("Stepplot, variables coded by pattern") +
     geom_step(aes(x=x1, y=x2, lty="x2"), position=position_nudge(x=0.2, y=0.2)) +
     geom_step(aes(x=x1, y=x6, lty="x6"), position=position_nudge(x=0.1, y=0.1)) +
     geom_step(aes(x=x1, y=x7, lty="x7"), position="nudge") +
     scale_linetype_manual(values=c("solid","longdash","dotdash"))

grid.arrange(p1, p2, nrow=1)
```

### Visualize Long-Format Matrices as Tile Plots, Numerical vs Categorical Color Coding

```r
clock.start() # may take about 1 minute

m = mandelbrot(xmin=-2, xmax=1, ymin=-1.5, ymax=1.5)
d1 = as.data.frame(expand.grid(m$x, m$y)); names(d1) = c("x","y")
d1$k = as.vector(m$matrix)   
   
m = mandelbrot(xmin = 0, xmax = 0.5, ymin = 0.25, ymax = 0.75)
d2 = as.data.frame(expand.grid(m$x, m$y)); names(d2) = c("x","y")
d2$k = as.vector(m$matrix)

m = mandelbrot(xmin = 0.350, xmax = 0.370, ymin = 0.630, ymax = 0.650)
d3 = as.data.frame(expand.grid(m$x, m$y)); names(d3) = c("x","y")
d3$k = as.vector(m$matrix)


p11 = ggplot(d1) + geom_tile(aes(x=x, y=y, fill=k)) + theme.no_legend
p21 = ggplot(d1) + geom_tile(aes(x=x, y=y, fill=factor(k))) + scale_fill_manual(values=stretch(PALETTE, length(unique(d1$k)), each=3, last="black")) + theme.no_legend
p12 = ggplot(d2) + geom_tile(aes(x=x, y=y, fill=k)) + theme.no_legend
p22 = ggplot(d2) + geom_tile(aes(x=x, y=y, fill=factor(k))) + scale_fill_manual(values=stretch(PALETTE, length(unique(d2$k)), each=3, last="black")) + theme.no_legend
p13 = ggplot(d3) + geom_tile(aes(x=x, y=y, fill=k)) + theme.no_legend
p23 = ggplot(d3) + geom_tile(aes(x=x, y=y, fill=factor(k))) + scale_fill_manual(values=stretch(PALETTE, length(unique(d3$k)), each=3, last="black")) + theme.no_legend


output_size(6.5,4)
grid.arrange(p11, p12, p13, p21, p22, p23, nrow=2, ncol=3)
output_size(restore)

clock.stop()
```
