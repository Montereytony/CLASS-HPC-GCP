```r
cat("<< Notebook apparatus is located here. Toggle cell input display to see code. >>")

options(warn=-1)

# Load some required functions
library(ggplot2,   verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(gridExtra, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(rgl,       verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE) # All rglwidget() calls must use elementID="myplot"
library(rmarkdown, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(htmltools, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(IRdisplay, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(base64enc, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)

# Define some useful functions
youtube = function(id, width=500, height=500) { display_html(paste0('<iframe src =https://www.youtube.com/embed/', id, ' width=', width, ', height=', height, ' allowfullscreen></iframe> ' )) }

# Set some visualization constants
restore = list(repr.plot.width=8, repr.plot.height=3)
PALETTE = c("#00A9FF", "#F8766D", "#7CAE00", "#C77CFF", "#CD9600", "#00BE67", "#FF61CC", "#00BFC4")
theme.x_axis_only = theme(axis.title.y=element_blank(), axis.text.y=element_blank(), axis.ticks.y=element_blank(), panel.grid.major.y=element_blank(), panel.grid.minor.y=element_blank())
theme.no_legend = theme(legend.position="none")
theme.legend_title = theme(legend.title=element_text(size=7))

# Set some visualization formatting defaults
options(scipen=10)
options(restore)
options(rgl.useNULL=TRUE)
update_geom_defaults("text", list(size=2.5, vjust=-1.5))
update_geom_defaults("point", list(size=3, colour=PALETTE[1]))
update_geom_defaults("line", list(size=0.5, colour=PALETTE[1]))
update_geom_defaults("path", list(size=0.5, colour=PALETTE[1]))
update_geom_defaults("col", list(color=NA, lwd=5))
update_geom_defaults("bar", list(color=NA, fill=PALETTE[1], lwd=5))
update_geom_defaults("violin", list(color=NA, lwd=5))
theme_update(plot.title=element_text(size=10, hjust=0.5), plot.subtitle=element_text(size=8, face="italic", hjust=0.5), axis.title=element_text(size=7), axis.text=element_text(size=7), strip.text=element_text(size=7), strip.text.y=element_text(angle=90), legend.title=element_blank(), legend.text=element_text(size=7))
```

### Visualization as a 1-Axis Scatterplot
Visualize up to 5 variable distributions:
* position
* color
* size
* shape
* label

```r
# Show 1 variable

data.frame(variable="x1", aspect=c("position"))

options(repr.plot.width=4, repr.plot.height=1)

ggplot(data) + xlim(0,6) +
geom_point(aes(x=x1, y=0)) +
theme.x_axis_only

options(restore)
```

```r
# Show 1 variable
# Note: color specification is outside aes, so it does not depend on variable values

data.frame(variable="x1", aspect=c("position"))

options(repr.plot.width=4, repr.plot.height=1)

ggplot(data) + xlim(0,6) +
geom_point(aes(x=x1, y=0), color=PALETTE[2]) +
theme.x_axis_only

options(restore)
```

```r
# Show 5 variables
# Note: color, size, and shape specifications are inside aes, so they do depend on variable values

data.frame(variable=c("x1","x3","x4","x5","x8"), aspect=c("position","color","size","shape","label"))

options(repr.plot.width=4, repr.plot.height=1.5)

ggplot(data) + xlim(0,6) +
geom_point(aes(x=x1, y=0, color=x3, size=x4, shape=x5)) +
geom_text(aes(x=x1, y=0, label=x8), vjust=-4) +
scale_color_manual(values=PALETTE, drop=FALSE) +
scale_size_manual(values=3*(1:5), drop=FALSE) +
scale_shape_manual(values=c(19,15,17,18,4), drop=FALSE) +
theme.x_axis_only + theme.no_legend

options(restore)
```

### Visualization as a 1-Axis Scatterplot Animation
Visualize up to 6 variable distributions:
* position
* color
* size
* shape
* label
* pane

```r
# Show 6 variables

data.frame(variable=c("x1","x3","x4","x5","x8","x6"), aspect=c("position","color","size","shape","label","pane"))

options(repr.plot.width=8, repr.plot.height=1.7)

p1 = ggplot(data[data$x6==2,]) + ggtitle("x6 = 2") + xlim(0,6) +
     geom_point(aes(x=x1, y=0, color=x3, size=x4, shape=x5)) +
     geom_text(aes(x=x1, y=0, label=x8), vjust=-4) +
     scale_color_manual(values=PALETTE, drop=FALSE) +
     scale_size_manual(values=3*(1:5), drop=FALSE) +
     scale_shape_manual(values=c(19,15,17,18,4), drop=FALSE) +
     theme.x_axis_only + theme.no_legend

p2 = ggplot(data[data$x6==3,]) + ggtitle("x6 = 4") + xlim(0,6) + 
     geom_point(aes(x=x1, y=0, color=x3, size=x4, shape=x5)) +
     geom_text(aes(x=x1, y=0, label=x8), vjust=-4) +
     scale_color_manual(values=PALETTE, drop=FALSE) +
     scale_size_manual(values=3*(1:5), drop=FALSE) +
     scale_shape_manual(values=c(19,15,17,18,4), drop=FALSE) +
     theme.x_axis_only + theme.no_legend

p3 = ggplot(data[data$x6==4,]) + ggtitle("x6 = 4") + xlim(0,6) + 
     geom_point(aes(x=x1, y=0, color=x3, size=x4, shape=x5)) +
     geom_text(aes(x=x1, y=0, label=x8), vjust=-4) +
     scale_color_manual(values=PALETTE, drop=FALSE) +
     scale_size_manual(values=3*(1:5), drop=FALSE) +
     scale_shape_manual(values=c(19,15,17,18,4), drop=FALSE) +
     theme.x_axis_only + theme.no_legend

grid.arrange(p1, p2, p3, nrow=1)

options(restore)
```

### Visualization a 3-Axis Scatterplot Projection
Visualize up to 6 variable distributions:
* horizontal position
* vertical position
* color
* size
* shape
* depth position

```r
# Show 3 variables

data.frame(variable=c("x1","x2","x6"), aspect=c("horizontal position","vertical position","depth position"))

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data[,c("x1","x2","x6")], type="s", col=PALETTE[1], xlim=c(0,6), ylim=c(0,6), zlim=c(0,6)); grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

```r
# Show 6 variables

data.frame(variable=c("x1","x2","x3","x4","x5","x6"),
           aspect=c("horizontal position","vertical position","color","size","shape","depth position"))

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data[,c("x1","x2","x6")], type="n", xlim=c(0,6), ylim=c(0,6), zlim=c(0,6)); grid3d(c("x","y","z"))
shapelist3d(cube3d        (col=PALETTE[as.numeric(data$x3[1])]), x=data$x1[1], y=data$x2[1], z=data$x6[1], size=0.15*as.numeric(data$x4[1]))
shapelist3d(tetrahedron3d (col=PALETTE[as.numeric(data$x3[2])]), x=data$x1[2], y=data$x2[2], z=data$x6[2], size=0.15*as.numeric(data$x4[2]))
shapelist3d(tetrahedron3d (col=PALETTE[as.numeric(data$x3[3])]), x=data$x1[3], y=data$x2[3], z=data$x6[3], size=0.15*as.numeric(data$x4[3]))
shapelist3d(octahedron3d  (col=PALETTE[as.numeric(data$x3[4])]), x=data$x1[4], y=data$x2[4], z=data$x6[4], size=0.15*as.numeric(data$x4[4]))
shapelist3d(dodecahedron3d(col=PALETTE[as.numeric(data$x3[5])]), x=data$x1[5], y=data$x2[5], z=data$x6[5], size=0.15*as.numeric(data$x4[5]))
shapelist3d(cube3d        (col=PALETTE[as.numeric(data$x3[6])]), x=data$x1[6], y=data$x2[6], z=data$x6[6], size=0.15*as.numeric(data$x4[6]))
rglwidget(elementId="myplot", width=400, height=400)
```

### Visualization as a Lineplot
Visualize many variable distributions; treat 1 variable as a shared sequence control:
* horizontal position
* vertical position
* color (to indicate variable, not to indicate value)
* pattern (to indicate variable, not to indicate value)

```r
# Show 2 variables

data.frame(variable=c("x1","x2"), aspect=c("horizontal position","vertical position"))

options(repr.plot.width=3, repr.plot.height=3)

ggplot(data) + xlim(0,6) + ylim(0,6) +
geom_line(aes(x=x1, y=x2))

options(restore)
```

```r
# Show 4 variables (3 variables distinguished by color or pattern)

data.frame(variable=c("x1","x2","x6","x7"),
           aspect=c("horizontal position","vertical position","vertical position","vertical position"),
           control=c("yes","","",""))

options(repr.plot.width=8, repr.plot.height=3)

p1 = ggplot(data) +  xlim(0,6) + ylim(0,6) + ylab("") +
     geom_line(aes(x=x1, y=x2, color="x2")) +
     geom_line(aes(x=x1, y=x6, color="x6")) +
     geom_line(aes(x=x1, y=x7, color="x7")) +
     scale_color_manual(values=PALETTE[1:3])

p2 = ggplot(data) + xlim(0,6) + ylim(0,6) + ylab("") +
     geom_line(aes(x=x1, y=x2, lty="x2")) +
     geom_line(aes(x=x1, y=x6, lty="x6")) +
     geom_line(aes(x=x1, y=x7, lty="x7")) +
     scale_linetype_manual(values=c("solid","longdash","dotdash"))

grid.arrange(p1, p2, nrow=1)

options(restore)
```

```r
# Show 2 variables; horizontal position is categorical
# Use group=1 in geom_line

data.frame(variable=c("x4","x2"), aspect=c("horizontal position","vertical position"))

options(repr.plot.width=3, repr.plot.height=3)

ggplot(data) + ylim(0,6) +
geom_line(aes(x=x4, y=x2, group=1))

options(restore)
```

```r
# Show 4 variables (3 variables distinguished by color or pattern); horizontal position is categorical
# use group=1 in geom_line

data.frame(variable=c("x4","x2","x6","x7"),
           aspect=c("horizontal position","vertical position","vertical position","vertical position"),
           control=c("yes","","",""))

options(repr.plot.width=8, repr.plot.height=3)

p1 = ggplot(data) + ylim(0,6) + ylab("") +
     geom_line(aes(x=x4, y=x2, color="x2", group=1)) +
     geom_line(aes(x=x4, y=x6, color="x6", group=1)) +
     geom_line(aes(x=x4, y=x7, color="x7", group=1)) +
     scale_color_manual(values=PALETTE[1:3])

p2 = ggplot(data) + ylim(0,6) + ylab("") +
     geom_line(aes(x=x4, y=x2, lty="x2", group=1)) +
     geom_line(aes(x=x4, y=x6, lty="x6", group=1)) +
     geom_line(aes(x=x4, y=x7, lty="x7", group=1)) +
     scale_linetype_manual(values=c("solid","longdash","dotdash"))

grid.arrange(p1, p2, nrow=1)

options(restore)
```

### Visualization as a Stepplot
Visualize many variable distributions; treat 1 variable as a shared sequence control:
* horizontal position
* vertical position
* color (to indicate variable, not to indicate value)
* pattern (to indicate variable, not to indicate value)

```r
# Show 2 variables

data.frame(variable=c("x1","x2"), aspect=c("horizontal position","vertical position"))

options(repr.plot.width=3, repr.plot.height=3)

ggplot(data) + xlim(0,6) + ylim(0,6) +
geom_step(aes(x=x1, y=x2))

options(restore)
```

```r
# Show 4 variables (3 variables distinguished by color or pattern)
# Adjust for overlap with transparencies (alpha)
# Adjust for overlap with nudges

data.frame(variable=c("x1","x2","x6","x7"),
           aspect=c("horizontal position","vertical position","vertical position","vertical position"),
           control=c("yes","","",""))

options(repr.plot.width=8, repr.plot.height=3)

p1 = ggplot(data) +  xlim(0,6) + ylim(0,6) + ylab("") +
     geom_step(aes(x=x1, y=x2, color="x2"), alpha=0.5) +
     geom_step(aes(x=x1, y=x6, color="x6"), alpha=0.5) +
     geom_step(aes(x=x1, y=x7, color="x7"), alpha=0.5) +
     scale_color_manual(values=PALETTE[1:3])

p2 = ggplot(data) + xlim(0,6) + ylim(0,6) + ylab("") +
     geom_step(aes(x=x1, y=x2, lty="x2"), position=position_nudge(x=0.2, y=0.2)) +
     geom_step(aes(x=x1, y=x6, lty="x6"), position=position_nudge(x=0.1, y=0.1)) +
     geom_step(aes(x=x1, y=x7, lty="x7"), position="nudge") +
     scale_linetype_manual(values=c("solid","longdash","dotdash"))

grid.arrange(p1, p2, nrow=1)

options(restore)
```

### Visualization as a Barplot
Visualize up to 3 variable distributions:
* horizontal position
* vertical position
* color

```r
# Show 3 variables

data.frame(variable=c("x2","x6","x3"), aspect=c("horizontal position","vertical position","color"))

options(repr.plot.width=8, repr.plot.height=3)

p1 = ggplot(data) +
     geom_col(aes(x=x5, y=x6, fill=x3), position="dodge") +
     scale_fill_manual(values=PALETTE, drop=FALSE) +
     theme.legend_title

p2 = ggplot(data) +
     geom_col(aes(x=x5, y=x6, fill=x3), position="stack") +
     scale_fill_manual(values=PALETTE, drop=FALSE) +
     theme.legend_title

grid.arrange(p1, p2, nrow=1)

options(restore)
```

```r
# Show 3 variables

options(repr.plot.width=8, repr.plot.height=3)

p1 = ggplot(data) +
     geom_col(aes(x=x5, y=x6, fill=x3), position="dodge") + coord_flip() +
     scale_fill_manual(values=PALETTE, drop=FALSE) +
     theme.legend_title

p2 = ggplot(data) +
     geom_col(aes(x=x5, y=x6, fill=x3), position="stack") + coord_flip() + 
     scale_fill_manual(values=PALETTE, drop=FALSE) +
     theme.legend_title

grid.arrange(p1, p2, nrow=1)

options(restore)
```

### Visualization as a Histogram
Visualize 2 variable distributions:
* horizontal position
* vertical position
* color

```r
# Show 1 variables + 1 statistic

data.frame(variable=c("x6","count statistic"), aspect=c("horizontal position","vertical position"))

options(repr.plot.width=8, repr.plot.height=3)

p1 = ggplot(data) + ylim(0,6) +
     geom_histogram(aes(x=x6), bins=2, fill=PALETTE[1]) +
     theme.legend_title

p2 = ggplot(data) + ylim(0,6) +
     geom_histogram(aes(x=x6), bins=4, fill=PALETTE[2]) +
     theme.legend_title

grid.arrange(p1, p2, nrow=1)

options(restore)
```
