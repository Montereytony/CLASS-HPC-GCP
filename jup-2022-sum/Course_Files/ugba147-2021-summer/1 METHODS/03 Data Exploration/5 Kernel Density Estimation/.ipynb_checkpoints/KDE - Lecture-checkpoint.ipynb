### Guessing a Simple Underlying Process in One Dimension
___

#### Given a probability distribution of events from an underlying process.  Pretend that the probability distribution is unknown, but that new observations can be gathered.

```r
pdf.simple = function(x) { y = rep(0, length(x)); if (1 %in% x) y[which(x==1)]=0.5; if (2 %in% x) y[which(x==2)]=0.5; y }
```

<br>
#### Gather new observations, guess the probability distribution.

Initialize.

Then iterate, 10 new observations per iteration.  Can you guess the probability distribution?

Then iterate, 1000 new observations per iteration.  Can you guess the probability distribution? 

```r
# Initialize
set.seed(12345)
x1 = c()
```

```r
# Iterate, 10 new observations per iteration ...

x1 = c(x1, sample(0:20, size=10, replace=TRUE, prob=pdf.simple(0:20)))

p1 = ggplot() + geom_point(aes(x=x1, y=1), size=0.5) + xlim(-1,21) + ylim(0,2) + theme.x_axis_only
p2 = ggplot() + geom_jitter(aes(x=x1, y=1), size=0.5, height=1, na.rm=TRUE) + xlim(-1,21) + ylim(0,2) + theme.x_axis_only
p3 = ggplot() + geom_bar(aes(x=x1), na.rm=TRUE) + xlim(-1,21)

length(x1)
grid.arrange(p1, p2, p3, nrow=1)
```

```r
# Iterate, 1000 new observations per iteration ...

x1 = c(x1, sample(0:20, size=1000, replace=TRUE, prob=pdf.simple(0:20)))

p1 = ggplot() + geom_point(aes(x=x1, y=1), size=0.5) + xlim(-1,21) + ylim(0,2) + theme.x_axis_only
p2 = ggplot() + geom_jitter(aes(x=x1, y=1), size=0.5, height=1, na.rm=TRUE) + xlim(-1,21) + ylim(0,2) + theme.x_axis_only
p3 = ggplot() + geom_bar(aes(x=x1), na.rm=TRUE) + xlim(-1,21)

length(x1)
grid.arrange(p1, p2, p3, nrow=1)
```

<br>
#### The actual probability distribution from the underlying process is ...

```r
ggplot(data.frame(0)) + stat_function(fun=pdf.simple, geom="bar", n=101) + xlim(0,20)
```

### Guessing a Complex Underlying Process in One Dimension
___

#### Given a probability distribution of events from an underlying process.  Pretend that the probability distribution is unknown, but that new observations can be gathered.

```r
reference = data.frame(x=c(0,1,2,3,4,9,10,11,12,17,18,19,20), y=c(0,0,20,0,0,0,1,1,0,0,20,20,0))
model = lm(y ~ poly(x, 6, raw=TRUE), reference)
df  = function(x) {(coef(model)[7]*x^6 + coef(model)[6]*x^5 + coef(model)[5]*x^4 + coef(model)[4]*x^3 + coef(model)[3]*x^2 + coef(model)[2]*x + coef(model)[1] + 10)}
area = integrate(df, 0, 20)
pdf.complex = function(x) {(coef(model)[7]*x^6 + coef(model)[6]*x^5 + coef(model)[5]*x^4 + coef(model)[4]*x^3 + coef(model)[3]*x^2 + coef(model)[2]*x + coef(model)[1] + 10) / area$value}
```

<br>
#### Gather new observations, guess the probability distribution.

Initialize.

Then iterate, 10 new observations per iteration.  Can you guess the probability distribution?

Then iterate, 1000 new observations per iteration.  Can you guess the probability distribution? 

```r
# Initialize
set.seed(12345)
x1 = c()
```

```r
# Iterate, 10 new observations per iteration

x1 = c(x1, sample(0:20, size=10, replace=TRUE, prob=pdf.complex(0:20)))

p1 = ggplot() + geom_point(aes(x=x1, y=1), size=0.5) + xlim(-1,21) + ylim(0,2) + theme.x_axis_only
p2 = ggplot() + geom_jitter(aes(x=x1, y=1), size=0.5, height=1, na.rm=TRUE) + xlim(-1,21) + ylim(0,2) + theme.x_axis_only
p3 = ggplot() + geom_bar(aes(x=x1), na.rm=TRUE) + xlim(-1,21)

length(x1)
grid.arrange(p1, p2, p3, nrow=1)
```

```r
# Iterate, 1000 new observations per iteration ...

x1 = c(x1, sample(0:20, size=1000, replace=TRUE, prob=pdf.complex(0:20)))

p1 = ggplot() + geom_point(aes(x=x1, y=1), size=0.5) + xlim(-1,21) + ylim(0,2) + theme.x_axis_only
p2 = ggplot() + geom_jitter(aes(x=x1, y=1), size=0.5, height=1, na.rm=TRUE) + xlim(-1,21) + ylim(0,2) + theme.x_axis_only
p3 = ggplot() + geom_bar(aes(x=x1), na.rm=TRUE) + xlim(-1,21)

length(x1)
grid.arrange(p1, p2, p3, nrow=1)
```

<br>
#### The actual probability distribution from the underlying process is ...

```r
ggplot(data.frame(0)) + stat_function(fun=pdf.complex, geom="area", n=100) + xlim(0,20)
```

### Guessing an Underlying Process in One Dimension without Gathering New Observations 
___

```r
options(repr.plot.width=8, repr.plot.height=1)
data = data.frame(x1=c(1,2,3,4,8,9,16,17,18,19,20))
ggplot(data) + geom_point(aes(x=x1, y=0)) + xlim(-1,21) + theme.x_axis_only
options(restore)
```

<br>
#### Distribution of counts of observations is not a satisfying guess of the underlying process. 

```r
ggplot(data) + geom_bar(aes(x=x1, y=..prop..), na.rm=TRUE) + xlim(-1,21)
```

#### A histogram with small number of bins is not a satisfying guess of the underlying process.

```r
ggplot(data) +
geom_histogram(aes(x=x1, y=..density..), bins=6, na.rm=TRUE) +xlim(-1,21)
```

#### A histogram with large number of bins is not a satisfying guess of the underlying process, though perhaps better than a histogram with small number of bins.

```r
ggplot(data) +
geom_histogram(aes(x=x1, y=..density..), bins=12, na.rm=TRUE) + xlim(-1,21)
```

<br>
#### A histogram with very large number of bins is not a satisfying guess of the underlying process. 

```r
ggplot(data) +
geom_histogram(aes(x=x1, y=..density..), bins=1000, na.rm=TRUE) + xlim(-1,21)
```

<br>
#### A probability distribution something like this would perhaps be most satisfying. 

```r
ggplot(data) +
stat_density(aes(x=x1), kernel="gaussian", bw=1, n=500) + xlim(0,20)
```

### Kernal Density Estimation Method for One Dimension
___

#### Find a satisfying probability distribution for the underlying process based on a few observations.

<br>
#### Rectangle function as kernel.

One rectangle for each observation, with center at the observation.  Each rectangle is of same size, halfwidth is set by bandwidth, height is set such that total area of all rectangles is 1.

Kernel density estimate is sum of heights along horizontal axis.

Try different bandwidths (e.g., 1, 5, 0.5), and note the effect.

```r
bandwidth = 1 # try bandwidth 1 first, then adjust as desired

kernel = function(x, center, halfwidth) { z=rep(0, length(x)); z[(x >= center-halfwidth) & (x <= center+halfwidth)] = 1 / (11*halfwidth); z }
sum_kernels = function(x) { aaply(x, 1, function(i) sum(aaply(1, 1, function(j) { kernel(i, data$x1[j], bandwidth) } ))) }
                                  
p = ggplot(data) + xlim(-1,20) + ylim(0,0.2)

p1 = p + stat_function(fun=kernel, args=c(center=data$x1[1], halfwidth=bandwidth), n=500, geom="area", alpha=0.3) +
         geom_point(aes(x=x1, y=0))
p1

p2 = p1 + stat_function(fun=sum_kernels, n=500, geom="area", color=PALETTE[1], alpha=0.1)
p2
```

#### Triangle function as kernel.

One isocelese triangle for each observation, with center at the observation.  Each triangle is of same size, halfwidth is set by bandwidth, height is set such that total area of all triangles is 1.

Kernel density estimate is sum of heights along horizontal axis.

Try different bandwidths (e.g., 3, 6, 1), and note the effect.

```r
bandwidth = 3 # start with bandwidth 3 first, then adjust as desired

kernel = function(x, center, hw) { z=rep(0,length(x));
                                   i = which((x >= center) & (x <= center+hw));
                                   z[i] = aaply(i, 1, function(ii) {(((center+hw)-x[ii])/hw)*(1/hw)} );
                                   j = which((x >= center-hw) & (x < center));
                                   z[j] = aaply(j, 1, function(jj) {((x[jj]-(center-hw))/hw)*(1/hw)} );
                                   z = z/11 
                                   return(z) }
sum_kernels = function(x) { aaply(x, 1, function(i) sum(aaply(1:11, 1, function(j) { kernel(i, data$x1[j], bandwidth) } ))) }

p = ggplot(data) + geom_point(aes(x=x1, y=0)) + xlim(-1,21) + ylim(0,0.1)

p1 = p + stat_function(fun=kernel, args=c(center=data$x1[1], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[2], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[3], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[4], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[5], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[6], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[7], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[8], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[9], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[10], hw=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(center=data$x1[11], hw=bandwidth), n=500, geom="area", alpha=0.3)

p2 = p1 + stat_function(fun=sum_kernels, n=500, geom="area", color=PALETTE[1], alpha=0.1)

p1
p2
```

#### Gaussian function as kernel.

One gaussian for each observation, with mean at the observation.  Each gaussian is of same size, standard deviation is set by bandwidth, height is set such that total area of all gaussians is 1.

Kernel density estimate is sum of heights along horizontal axis.

Try different bandwidths (e.g., 1, 2, 0.5), and note the effect.

```r
bandwidth = 1 # start with bandwidth 1 first, then adjust as desired

kernel = function(x, mean, sd) { (1/(sd*sqrt(2*pi))) * exp(-0.5 * ((x-mean)/sd)^2) / 11 }
sum_kernels = function(x) { aaply(x, 1, function(i) sum(aaply(1:11, 1, function(j) { kernel(i, data$x1[j], bandwidth) } ))) }


p = ggplot(data) + geom_point(aes(x=x1, y=0)) + xlim(-1,21) + ylim(0,0.1)

p1 = p + stat_function(fun=kernel, args=c(mean=data$x1[1], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[2], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[3], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[4], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[5], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[6], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[7], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[8], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[9], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[10], sd=bandwidth), n=500, geom="area", alpha=0.3) +
         stat_function(fun=kernel, args=c(mean=data$x1[11], sd=bandwidth), n=500, geom="area", alpha=0.3)


p2 = p1 + stat_function(fun=sum_kernels, color=PALETTE[1], n=500, geom="area", alpha=0.1)

p1
p2
```

### Kernel Density Estimation Method for Two Dimensions
___
#### Retrieve data.

```r
data = data.frame(x1=c(1,2,3,4,8,9,16,17,18,19,20), x2=c(1,2,3,4,  15,13,14,15,  18,19,20))

options(repr.plot.width=4, repr.plot.height=1)
ggplot(data) + geom_point(aes(x=x1, y=0)) + xlim(0,21) + theme.x_axis_only
ggplot(data) + geom_point(aes(x=x2, y=0)) + xlim(0,21) + theme.x_axis_only

options(repr.plot.width=4, repr.plot.height=4)
ggplot(data) + geom_point(aes(x=x1, y=x2)) + xlim(0,20) + ylim(0,20)

options(restore)
```

<br>
#### Plot 1-dimensional kernel density estimations, one for each dimension.

```r
ggplot(data) + stat_density(aes(x=x1), bw=1, kernel="gaussian", n=512)
ggplot(data) + stat_density(aes(x=x2), bw=1, kernel="gaussian", n=512)
```

<br>
#### Plot 2-dimensional kernel density estimation.

The kernel used here is a 3-dimensional gaussian, rather than a 2-dimensional gaussian.

```r
options(repr.plot.width=5, repr.plot.height=4)

ggplot(data) +
stat_density_2d(aes(x=x1, y=x2, fill=..level.., alpha=..level..), h=c(10,10), bins=100, geom="polygon") +
stat_density_2d(aes(x=x1, y=x2), h=c(10,10), bins=20, size=0.1, color="black") +
scale_fill_continuous(low="yellow", high="red") +
geom_point(aes(x=x1, y=x2))

options(restore)
```

```r
data$z = 0

contour = kde2d(data$x1, data$x2, h=c(10,10), n=100, lims=c(0,20,0,20))
color = rev(heat.colors(100,1))
zcol = cut(contour$z, 100)

p = open3d(); view3d(0,0, zoom=0.65)

plot3d(data[,c("x1","x2","z")], type="s", col=PALETTE[1], xlim=c(0,20), ylim=c(0,20), zlim=c(0, 30))
persp3d(contour$x, contour$y, contour$z, col=color[zcol], alpha=0.7)

grid3d(c("x","y","z"))

rglwidget(elementId="myplot", width=400, height=400)
rgl.close()
```
