```r
# Load some required functions
library(lubridate, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(ggplot2,   verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(gridExtra, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)

# Set some visualization constants
PALETTE = c("dodgerblue2", "brown3", "chartreuse4", "darkgoldenrod", "mediumorchid3")

# Set some visualization formatting defaults
options(repr.matrix.max.cols=100)
options(repr.plot.width=8, repr.plot.height=2.75)
update_geom_defaults("point", list(size=0.25, colour=PALETTE[1]))
update_geom_defaults("line", list(size=0.15))
update_geom_defaults("bar", list(colour=PALETTE[1], fill=PALETTE[1]))
update_geom_defaults("density", list(colour=PALETTE[1], fill=PALETTE[1]))
theme_update(plot.title=element_text(size=10, hjust=0.5), plot.subtitle=element_text(size=8, face="italic", hjust=0.5), axis.title=element_text(size=7), axis.text=element_text(size=7), strip.text=element_text(size=7), strip.text.y=element_text(angle=90), legend.title=element_blank(), legend.text=element_text(size=7))
```

### Kernel Density Estimation for 2016 Sales
___

Retrieve the dataset for 2016.

```r
data.2016 = read.csv("../DATASETS/DATASET Iowa Liquor Sales 2016.csv")
dim(data.2016)
head(data.2016)
```

<br>
Kernel density estimation, consider all sales.

```r
ggplot(data.2016) +
stat_density(aes(sale.dollars), kernel="gaussian", bw=10, n=500, fill=PALETTE[1])
```

```r
min(data.2016$sale.dollars) # minimum value transaction
max(data.2016$sale.dollars) # maximum value transaction
length(which(data.2016$sale.dollars >= 500)) # number of transactions over $500
length(which(data.2016$sale.dollars >= 500)) / nrow(data.2016) # fraction of transactions over $500
```

<br>
Kernel density estimation, consider only sales under $500.

```r
ggplot(data.2016[data.2016$sale.dollars<500,]) +
stat_density(aes(sale.dollars), kernel="gaussian", bw=10, n=500, fill=PALETTE[1])
```

<br>
Compare kernel denisty estimations based on various kernels.

```r
p = ggplot(data.2016[data.2016$sale.dollars<500,])

grid.arrange(p + stat_density(aes(sale.dollars), kernel="gaussian",    bw=10, n=500, fill=PALETTE[1]),
             p + stat_density(aes(sale.dollars), kernel="rectangular", bw=10, n=500, fill=PALETTE[1]),
             p + stat_density(aes(sale.dollars), kernel="triangular",  bw=10, n=500, fill=PALETTE[1]),
             nrow=1)
```

<br>
Compare kernel density estimations based on gaussian kernel and various standard deviations.

```r
p = ggplot(data.2016[data.2016$sale.dollars<500,])

grid.arrange(p + stat_density(aes(sale.dollars), bw=1,  kernel="gaussian", n=500, fill=PALETTE[1]),
             p + stat_density(aes(sale.dollars), bw=10, kernel="gaussian", n=500, fill=PALETTE[1]),
             p + stat_density(aes(sale.dollars), bw=50, kernel="gaussian", n=500, fill=PALETTE[1]),
             nrow=1)
```

<br>
Compare kernal density estimations based on gaussian kernel, bandwidth 10, and various resolutions.

```r
p = ggplot(data.2016[data.2016$sale.dollars<500,])

grid.arrange(p + stat_density(aes(sale.dollars), bw=10, kernel="gaussian", n=10,  fill=PALETTE[1]),
             p + stat_density(aes(sale.dollars), bw=10, kernel="gaussian", n=20,  fill=PALETTE[1]),
             p + stat_density(aes(sale.dollars), bw=10, kernel="gaussian", n=500, fill=PALETTE[1]),
             nrow=1)
```

### Compare Kernel Density Estimates for 2016, 2015, 2014, 2013
___

Retrieve the datasets for 2015, 2014, 2013.

```r
data.2015 = read.csv("../DATASETS/DATASET Iowa Liquor Sales 2015.csv")

dim(data.2015)
head(data.2015)
```

```r
data.2014 = read.csv("../DATASETS/DATASET Iowa Liquor Sales 2014.csv")

dim(data.2014)
head(data.2014)
```

```r
data.2013 = read.csv("../DATASETS/DATASET Iowa Liquor Sales 2013.csv")

dim(data.2013)
head(data.2013)
```

<br>
Compare total sales by year.

```r
sum.2013 = sum(data.2013$sale.dollars[data.2013$sale.dollars<500])
sum.2014 = sum(data.2014$sale.dollars[data.2014$sale.dollars<500])
sum.2015 = sum(data.2015$sale.dollars[data.2015$sale.dollars<500])
sum.2016 = sum(data.2016$sale.dollars[data.2016$sale.dollars<500])

data.compare_sales = data.frame(year=as.factor(c("2013", "2014", "2015", "2016")),
                                total_sales=c(sum.2013, sum.2014, sum.2015, sum.2016))
data.compare_sales

ggplot(data.compare_sales) +
geom_col(aes(x=year, y=total_sales, fill=year)) +
scale_fill_manual(values=rev(PALETTE[1:4]), guide="none")
```

<br>
Compare kernel density estimates of sales by year. Present kernel denisty estimates side-by-side.

```r
bandwidth = 10

ggplot(data.2013[data.2013$sale.dollars<500,]) +
stat_density(aes(sale.dollars), kernel="gaussian", bw=bandwidth, n=500, fill=PALETTE[4])

ggplot(data.2014[data.2014$sale.dollars<500,]) +
stat_density(aes(sale.dollars), kernel="gaussian", bw=bandwidth, n=500, fill=PALETTE[3])

ggplot(data.2015[data.2015$sale.dollars<500,]) +
stat_density(aes(sale.dollars), kernel="gaussian", bw=bandwidth, n=500, fill=PALETTE[2])

ggplot(data.2016[data.2016$sale.dollars<500,]) +
stat_density(aes(sale.dollars), kernel="gaussian", bw=bandwidth, n=500, fill=PALETTE[1])
```

<br>
Compare kernel density estimates of sales by year. Present kernel denisty estimates side-by-side.

Note peculiarity of 2016 as compared to the other years.

```r
bandwidth = 10

ggplot() +
stat_density(aes(sale.dollars), data=data.2013[data.2013$sale.dollars<500,], kernel="gaussian", bw=bandwidth, n=500, fill=PALETTE[4], alpha=0.4) +
stat_density(aes(sale.dollars), data=data.2014[data.2014$sale.dollars<500,], kernel="gaussian", bw=bandwidth, n=500, fill=PALETTE[3], alpha=0.4) +
stat_density(aes(sale.dollars), data=data.2015[data.2015$sale.dollars<500,], kernel="gaussian", bw=bandwidth, n=500, fill=PALETTE[2], alpha=0.4) +
stat_density(aes(sale.dollars), data=data.2016[data.2016$sale.dollars<500,], kernel="gaussian", bw=bandwidth, n=500, fill=PALETTE[1], alpha=0.4)
```
