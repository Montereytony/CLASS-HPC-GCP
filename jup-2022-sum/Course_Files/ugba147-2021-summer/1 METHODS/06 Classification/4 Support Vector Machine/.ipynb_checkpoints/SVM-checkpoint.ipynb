### SVM applied to linearly separable data

Linearly separable means ...
* 1-dimensional data (i.e., 1 variable) can be separated into non-overlapping classes by a point.
* 2-dimensional data (i.e., 2 variables) can be separated into non-overlapping classes by a line.
* 3-dimensional data (i.e., 3 variables) can be separated into non-overlapping classes by a plane.
* 4+-dimensional data (i.e., 4+ variables) can be separated into non-overlapping classes by a hyper-plane.

#### Data

Data is already classified, i.e., observations are organized into classes A and B. 

```r
data = data.frame(x1=c(1,2,3,6,14,15,16,18), class=c("A", "A", "A", "A", "B", "B", "B", "B"))
new = data.frame(x1=8)

data.frame(new)

options(repr.plot.width=8, repr.plot.height=1)
ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=0, color=class)) + geom_point(aes(x=new$x1, y=0)) +
scale_color_manual(values=PALETTE) + theme.x_axis_only
options(restore)
```

#### Find the support vectors and boundary
* **Support vectors** are those observations at the edges of the space separating the classes.
* The **boundary** is the middle of the space separating the classes.

```r
support_vectors = c(6, 14)
boundary = mean(support_vectors)

options(repr.plot.width=8, repr.plot.height=1)

ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=0, color=class)) +
geom_point(aes(x=x1, y=0), data=data[data$x1 %in% support_vectors,], shape=1, size=6, color="black") +
geom_point(x=boundary, y=0, shape=8, color="black") +
scale_color_manual(values=PALETTE) + theme.x_axis_only
options(restore)
```

#### Score observations based on their distances from the boundary

```r
score = (boundary - data$x1)

options(repr.plot.width=8, repr.plot.height=1.2)
ggplot(data) + xlim(0,20) + ylim(-0.5,0.5) +
geom_point(aes(x=x1, y=0, color=class)) +
geom_text(aes(x=x1, y=0, label=score), nudge_y=0.3) +
geom_point(aes(x=x1, y=0), data=data[data$x1 %in% support_vectors,], shape=1, size=6, color="black") +
geom_point(x=boundary, y=0, shape=8, color="black") +
scale_color_manual(values=PALETTE) + theme.x_axis_only
options(restore)
```

#### Predict new observation class based on its score sign

```r
new_score = (boundary - new$x1)

new_color = PALETTE[as.numeric(factor(sign(new_score), levels=c(1,-1), labels=c(1,2)))]    

options(repr.plot.width=8, repr.plot.height=1.4)
ggplot(data) + ggtitle("Predicted based on score sign") + xlim(0,20) + ylim(-0.5,0.5) +
geom_point(aes(x=x1, y=0, color=class)) +
geom_text(aes(x=x1, y=0, label=score), nudge_y=0.3) +
geom_point(x=new$x1, y=0, size=6, color="gold") + geom_point(x=new$x1, y=0, color=new_color) +
geom_text(aes(x=new$x1, y=0, label=new_score), nudge_y=0.3) +
geom_point(x=boundary, y=0, shape=8, color="black") +
scale_color_manual(values=PALETTE) + theme.x_axis_only 
options(restore)
```

#### Alternatively, assign probabilities to observations based on their scores, and predict new observation class based on its probability

Various implementations of the support vector machine method employ various schemes to determine probabilities (and may even incorporate some randomization due to cross-validation).

Here, we use the sigmoid function to determine probabilities.  The sigmoid function rescales any incoming value into range 0 to 1.  The sigmoid function is also known as a logistic function.

sigmoid:  $\large \frac{1}{(1+{e}^{−x})}$

sigmoid range: $0 < sigmoid(x) < 1$, for any $x$

```r
options(repr.plot.width=3, repr.plot.height=2)
ggplot(data.frame(0))+ ggtitle("sigmoid function") + xlim(-10,10) + xlab("x") + ylab("sigmoid(x)") +
stat_function(fun=sigmoid) 
options(restore)
```

Scale the distances so that distance from the edge to the boundary is 1. 

```r
score = (boundary - data$x1) / (boundary - data$x1[4])

options(repr.plot.width=8, repr.plot.height=1.2)
ggplot(data) + xlim(0,20) + ylim(-0.5,0.5) +
geom_point(aes(x=x1, y=0, color=class)) +
geom_text(aes(x=x1, y=0, label=score), nudge_y=0.3) +
geom_point(aes(x=x1, y=0), data=data[data$x1 %in% support_vectors,], shape=1, size=6, color="black") +
geom_point(x=boundary, y=0, shape=8, color="black") +
scale_color_manual(values=PALETTE) + theme.x_axis_only
options(restore)
```

Then scale these revised distances by applying the sigmoid function.  Use the probability with a  cutoff to make a prediction. 

```r
new_score = (boundary - new$x1) / (boundary - data$x1[4])

prob = data.frame(score, A=sigmoid(score), B=1-sigmoid(score))
prob

new_prob = data.frame(new_score, A=sigmoid(new_score), B=1-sigmoid(new_score))
new_prob

cutoff = 0.5
new_color = PALETTE[as.numeric(factor(sign(new_score), levels=c(1,-1), labels=c(1,2)))]

options(repr.plot.width=8, repr.plot.height=1.3)
ggplot(data) + ggtitle("Predicted based on cutoff @ prob(A) > 0.5") + xlim(0,20) + ylim(-0.5, 0.5) + 
geom_point(aes(x=x1, y=0, color=class)) +
geom_text(aes(x=x1, y=0, label=sprintf("%0.3f", prob$A)), nudge_y=0.3) + 
geom_point(x=new$x1, y=0, size=6, color="gold") + geom_point(x=new$x1, y=0, color=new_color) +
geom_text(aes(x=new$x1, y=0, label=sprintf("%0.3f", new_prob$A)), nudge_y=0.3) +
geom_point(x=boundary, y=0, shape=8, color="black") +
scale_color_manual(values=PALETTE) + theme.x_axis_only 
options(restore)
```

```r
cutoff = 0.7
new_color = PALETTE[as.numeric(factor(new_prob$A>=cutoff, levels=c(TRUE,FALSE), labels=c(1,2)))] 

options(repr.plot.width=8, repr.plot.height=1.3)
ggplot(data) + ggtitle("Predicted based on cutoff @ prob(A) > 0.7") + xlim(0,20) + ylim(-0.5, 0.5) + 
geom_point(aes(x=x1, y=0, color=class)) +
geom_text(aes(x=x1, y=0, label=sprintf("%0.3f", prob$A)), nudge_y=0.3) + 
geom_point(x=new$x1, y=0, size=6, color="gold") + geom_point(x=new$x1, y=0, color=new_color) +
geom_text(aes(x=new$x1, y=0, label=sprintf("%0.3f", new_prob$A)), nudge_y=0.3) +
geom_point(x=boundary, y=0, shape=8, color="black") +
scale_color_manual(values=PALETTE) + theme.x_axis_only 
options(restore)
```

### SVM applied to not linearly separable data: penalty

#### Data
Note, these observations are not linearly separable. 

```r
data = data.frame(x1=c(2,3,4,5, 20,21,22,23), class=c("A","A","B","A", "B","B","B","B"))
new = data.frame(x1=12)

options(repr.plot.width=8, repr.plot.height=1)
ggplot(data) + geom_point(aes(x=x1, y=0, color=class)) + geom_point(x=new$x1, y=0) +
scale_color_manual(values=PALETTE) + theme.x_axis_only
options(restore)
```

#### Find the support vectors and boundary, assign penalties based on costs
* **Support vectors** are those observations at the edges of the space separating the classes and any observations on the wrong side of the boundary.  The support vectors are determined so that space is maximized, but subject to a penalty to account for the offending observations.  The penalty depends on the cost and the offending observations' distances from the correct edges.  You choose the cost. 
* The **boundary** is the middle of the space separating the classes.

Consider observation x1=5 and observation x1=20 as candidates for the support vectors.  The space would be 20-5=15, but observation x1=4 would be distance 16 on the wrong side of the x1=20 edge.  If cost was chosen as 0.1, then the penalty would be 0.1x16=1.6, and so the penalized space would be 15-1.6=13.4.  Similarly, if cost was chosen as 1 or 10, then penalized space would be 15-(1x16)=-1 or 15-(10x16)=-145, respectively.    

```r
support_vectors = c(5,20)
boundary = mean(data$x1[4:5])

cost = c(0.1,1,10)
space = data$x1[5] - data$x1[4]
offense = abs(data$x1[5] - data$x1[3])
penalty = cost*offense
data.frame(cost, space, offense, penalty, penalized_space=space-penalty)

options(repr.plot.width=8, repr.plot.height=1)
ggplot(data) + geom_point(aes(x=x1, y=0, color=class)) +
geom_point(aes(x=x1, y=0), data=data[data$x1 %in% support_vectors,], shape=1, size=6, color="black") +
geom_point(aes(x=boundary, y=0), shape=8, color="black") +
geom_segment(aes(x=data$x1[4], y=0, xend=data$x1[5], yend=0)) +
geom_segment(aes(x=data$x1[3], y=0.3, xend=data$x1[5], yend=0.3), color=PALETTE[2], arrow=arrow(angle=10)) +
scale_color_manual(values=PALETTE) + ylim(-0.5, 0.5) + theme.x_axis_only
options(restore)
```

Consider observation x1=3 and observation x1=20 as candiates for the support vectors.  The space would be 20-3=17, but observation x1=4 would be distance 16 on wrong side of the x1=20 edge, and observation x1=5 would be distance 2 on the wrong side of the x1=3 edge.  If cost was chosen as 0.1, then the penalty would be 0.1x(16+2)=1.8, and so the penalized space would be 17-1.8=15.2.  Similarly, if cost was chosen as 1 or 10, then penalized space would be 17-(1x(16+2))=-1 or 17-(10x(16+2))=-163, respectively. 

```r
support_vectors = c(3,20)
boundary = mean(data$x1[c(2,5)])

cost = c(0.1,1,10)
space = data$x1[5] - data$x1[2]
offense_3 = abs(data$x1[5] - data$x1[3])
offense_2 = abs(data$x1[4] - data$x1[2])
penalty = cost*offense_2 + cost*offense_3
data.frame(cost, space, offense_2, offense_3, penalty, penalized_space=space-penalty)

options(repr.plot.width=8, repr.plot.height=1)
ggplot(data) + geom_point(aes(x=x1, y=0, color=class)) +
geom_point(aes(x=x1, y=0), data=data[data$x1 %in% support_vectors,], shape=1, size=6, color="black") +
geom_point(aes(x=boundary, y=0), shape=8, color="black") +
geom_segment(aes(x=data$x1[2], y=0, xend=data$x1[5], yend=0)) +
geom_segment(aes(x=data$x1[3], y=0.3, xend=data$x1[5], yend=0.3), color=PALETTE[2], arrow=arrow(angle=10)) +
geom_segment(aes(x=data$x1[4], y=-0.3, xend=data$x1[2], yend=-0.3), color=PALETTE[1], arrow=arrow(angle=10)) +
scale_color_manual(values=PALETTE) + ylim(-0.5, 0.5) + theme.x_axis_only
options(restore)
```

Note, the choice of cost influences the determination of the support vectors, which determine the boundary.  In our example, at cost=0.1, observations x1=3 and x1=20 would be preferred over observations x1=5 and x1=20 as support vectors because 15.2 is greater than 13.4.  A new observation x1=12 would be classified by sign as B.  At cost=10, the reverse would be preferred because -145 is greater than -163, and a new observation x1=12 would be classified by sign as A.

The support vector machine method relies on an optimization algorithm to find the support vectors by maximizing the penalized space between edges. 

```r
boundary = mean(data$x1[c(2,5)])

options(repr.plot.width=8, repr.plot.height=1.3)
ggplot(data) + geom_point(aes(x=x1, y=0, color=class)) +
geom_point(aes(x=new$x1, y=0), size=6, color="gold") + geom_point(aes(x=new$x1, y=0), color=PALETTE[2]) +
geom_point(aes(x=boundary, y=0), shape=8, color="black") +
geom_segment(aes(x=data$x1[2], y=0, xend=data$x1[5], yend=0)) +
geom_segment(aes(x=data$x1[3], y=0.3, xend=data$x1[5], yend=0.3), color=PALETTE[2], arrow=arrow(angle=10)) +
geom_segment(aes(x=data$x1[4], y=-0.3, xend=data$x1[2], yend=-0.3), color=PALETTE[1], arrow=arrow(angle=10)) +
scale_color_manual(values=PALETTE) + ylim(-0.5, 0.5) + theme.x_axis_only +
ggtitle("find support vectors at cost=0.1")
options(restore)
```

```r
boundary = mean(data$x1[4:5])

options(repr.plot.width=8, repr.plot.height=1.3)
ggplot(data) + geom_point(aes(x=x1, y=0, color=class)) +
geom_point(aes(x=new$x1, y=0), size=6, color="gold") + geom_point(aes(x=new$x1, y=0), color=PALETTE[1]) +
geom_point(aes(x=boundary, y=0), shape=8, color="black") +
geom_segment(aes(x=data$x1[4], y=0, xend=data$x1[5], yend=0)) +
geom_segment(aes(x=data$x1[3], y=0.3, xend=data$x1[5], yend=0.3), color=PALETTE[2], arrow=arrow(angle=10)) +
scale_color_manual(values=PALETTE) + ylim(-0.5, 0.5) + theme.x_axis_only +
ggtitle("find support vectors at cost=10")
options(restore)
```

### SVM applied to non-linearly separable data: kernel trick

#### Data

```r
data = data.frame(x1=c(4,6,7,10,11,12,14, 18, 19, 20), class=c("A", "A", "A", "B", "B", "B", "B", "A", "A", "A"))
new = data.frame(x1=8)

options(repr.plot.width=8, repr.plot.height=1)
ggplot(data) + 
geom_point(aes(x=x1, y=0, color=class)) + 
geom_point(aes(x=x1, y=0), data=new) +
scale_color_manual(values=PALETTE) + xlim(0,20) + theme.x_axis_only
options(restore)
```

#### Introduce a new variable by applying a kernel function to original data

This effectively increases the dimensionality of the data.  Here, we apply the kernel function $y = (x-12)^2$

```r
data$y = (data$x1-12)^2

ggplot(data) + geom_point(aes(x=x1, y=y, color=class)) + scale_color_manual(values=PALETTE) + xlim(0,20)
```

#### Find the support vectors and boundary

* **Support vectors** are those observations at the edge of the space separating the classes, where the space is maximized. 
* The **boundary** is the middle of the space separating the classes (a line in 2-dimensional space).

```r
ggplot(data) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_hline(aes(yintercept=data$y[3]), color="grey50") +
geom_hline(aes(yintercept=data$y[4]), color="grey50") +
geom_hline(aes(yintercept=mean(data$y[3:4]))) +
scale_color_manual(values=PALETTE) + ggtitle("space separating classes is not maximized") +
xlim(0,20)
```

```r
support_vectors = c(7,10,18)

m = (data$y[3] - data$y[8]) / (data$x1[3] - data$x1[8])
i.upper = -m*data$x1[3]+data$y[3]
i.lower = -m*data$x1[4]+data$y[4]
i = mean(c(i.upper, i.lower))

ggplot(data) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_point(aes(x=x1, y=y), data=data[data$x1 %in% support_vectors,], shape=1, size=6, color="black") +
geom_abline(aes(intercept=i.upper, slope=m), color="grey50") +
geom_abline(aes(intercept=i.lower, slope=m), color="grey50") +
geom_abline(aes(intercept=i, slope=m)) +
scale_color_manual(values=PALETTE) + ggtitle("space separating classes is maximized") +
xlim(0,20)
```

#### Score observations based on their distances from the boundary and predict a new observation class based on its score sign

```r
score = -(m*data$x1 -1*data$y + i) / sqrt(m^2 + (-1)^2)

new$y = (new$x1-12)^2
new_score = -(m*new$x1 -1*new$y + i) / sqrt(m^2 + (-1)^2)
new_color = if (sign(new_score) > 0) PALETTE[1] else PALETTE[2]

ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.1f", score)), nudge_y=5) +
geom_point(aes(x=x1, y=y), data=new, size=6, color="gold") +
geom_point(aes(x=x1, y=y), data=new, color=new_color) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.1f", new_score)), data=new, nudge_y=5) +
geom_abline(aes(intercept=i, slope=m)) +
scale_color_manual(values=PALETTE)
```

#### Alternatively, assign probabilities to observations based on their scores, and predict new observation class based on its probability

```r
score.adj = score / score[3]

new_score.adj = new_score / score[3]

ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.1f", score.adj)), nudge_y=5) +
geom_point(aes(x=x1, y=y), data=new) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.1f", new_score.adj)), data=new, nudge_y=5) +
geom_abline(aes(intercept=i, slope=m)) +
scale_color_manual(values=PALETTE)
```

```r
prob = data.frame(score=score.adj, A=sigmoid(score.adj), B=1-sigmoid(score.adj))
new_prob = data.frame(score=new_score.adj, A=sigmoid(new_score.adj), B=1-sigmoid(new_score.adj))

cutoff = 0.5
new_color = if(new_prob[,"A"] >= cutoff) PALETTE[1] else PALETTE[2]

ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.3f", prob$A)), nudge_y=5) +
geom_point(aes(x=new$x1, y=new$y)) +
geom_text(aes(x=new$x1, y=new$y, label=sprintf("%0.3f", new_prob$A)), nudge_y=5) +
geom_abline(aes(intercept=i, slope=m)) +
scale_color_manual(values=PALETTE)
```

```r
options(repr.plot.width=8, repr.plot.height=1.3)
ggplot(data) + ggtitle("Predicted based on cutoff @ prob(A) > 0.5") + xlim(0,20) + ylim(-0.5,0.5) + 
geom_point(aes(x=x1, y=0, color=class)) + 
geom_text(aes(x=x1, y=0, label=sprintf("%0.3f", prob$A)), nudge_y=0.3) +
geom_point(aes(x=new$x1, y=0), size=6, color="gold") +
geom_point(aes(x=new$x1, y=0), color=new_color) +
geom_text(aes(x=new$x1, y=0, label=sprintf("%0.3f", new_prob$A)), nudge_y=0.3) +
scale_color_manual(values=PALETTE) + theme.x_axis_only
options(restore)
```

### SVM applied to _very_ non-linearly separable data: kernel trick

#### Data

```r
data = data.frame(x1=c(1,2, 4,6,7,10,11,12,14, 18, 19, 20), class=c("B", "B", "A", "A", "A", "B", "B", "B", "B", "A", "A", "A"))
new = data.frame(x1=8)

options(repr.plot.width=8, repr.plot.height=1)
ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=0, color=class)) + 
geom_point(aes(x=x1, y=0), data=new) +
scale_color_manual(values=PALETTE) + theme.x_axis_only
options(restore)
```

#### Introduce a new variable by applying a kernel function to original data

This effectively increases the dimensionality of the data.  Here, we apply the kernel function $y = -1.8 + 3.0x -0.4x^2 +0.1x^3$

```r
data$y = -1.80475 + 3.03815*data$x1 + -0.38481*data$x1^2 + 0.01305*data$x1^3

ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=y, color=class)) +
scale_color_manual(values=PALETTE)
```

#### Find the support vectors and boundary

```r
support_vectors = c(4,10,18)

m = (data$y[3] - data$y[10]) / (data$x1[3] - data$x1[10])
i.upper = -m*data$x1[3]+data$y[3]
i.lower = -m*data$x1[6]+data$y[6]
i = mean(c(i.upper, i.lower))

ggplot(data) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_point(aes(x=x1, y=y), data=data[data$x1 %in% support_vectors,], shape=1, size=6, color="black") +
geom_abline(aes(intercept=i.upper, slope=m), color="grey50") +
geom_abline(aes(intercept=i.lower, slope=m), color="grey50") +
geom_abline(aes(intercept=i, slope=m)) +
scale_color_manual(values=PALETTE) + ggtitle("space separating classes is maximized") +
xlim(0,20)
```

#### Score observations based on their distances from the boundary and predict new observation class based on its score sign

```r
score = -(m*data$x1 -1*data$y + i) / sqrt(m^2 + (-1)^2)

new$y = -1.80475 + 3.03815*new$x1 + -0.38481*new$x1^2 + 0.01305*new$x1^3
new_score = -(m*new$x1 -1*new$y + i) / sqrt(m^2 + (-1)^2)
new_color = if(sign(new_score) > 0) PALETTE[1] else PALETTE[2]

ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.1f", score)), nudge_y=0.7) +
geom_point(aes(x=x1, y=y), data=new, size=6, color="gold") +
geom_point(aes(x=x1, y=y), data=new, color=new_color) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.1f", new_score)), data=new, nudge_y=0.7) +
geom_abline(aes(intercept=i, slope=m)) +
scale_color_manual(values=PALETTE)
```

#### Alternatively, assign probabilities to observations based on their scores, and predict new observation class based on its probability

```r
score.adj = score / score[3]

new_score.adj = new_score / score[3]

ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.1f", score.adj)), nudge_y=0.7) +
geom_point(aes(x=x1, y=y), data=new) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.1f", new_score.adj)), data=new, nudge_y=0.7) +
geom_abline(aes(intercept=i, slope=m)) +
scale_color_manual(values=PALETTE)
```

```r
prob = data.frame(score=score.adj, A=sigmoid(score.adj), B=1-sigmoid(score.adj))
prob

new_prob = data.frame(score=new_score.adj, A=sigmoid(new_score), B=1-sigmoid(new_score))
new_prob

cutoff = 0.5
new_color = if(new_prob$A >= cutoff) PALETTE[1] else PALETTE[2]

ggplot(data) + xlim(0,20) +
geom_point(aes(x=x1, y=y, color=class)) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.3f", prob$A)), nudge_y=0.7) +
geom_point(aes(x=x1, y=y), data=new) +
geom_text(aes(x=x1, y=y, label=sprintf("%0.3f", new_prob$A)), data=new, nudge_y=0.7) +
geom_abline(aes(intercept=i, slope=m)) +
scale_color_manual(values=PALETTE)
```

```r
options(repr.plot.width=8, repr.plot.height=1.3)
ggplot(data) + ggtitle("Predicted based on cutoff @ prob(A) > 0.5") + xlim(0,20) + ylim(-0.5,0.5) +
geom_point(aes(x=x1, y=0, color=class)) +
geom_text(aes(x=x1, y=0, label=sprintf("%0.3f", prob$A)), nudge_y=0.3) +
geom_point(aes(x=new$x1, y=0), size=6, color="gold") +
geom_point(aes(x=new$x1, y=0), color=new_color) +
geom_text(aes(x=new$x1, y=0, label=sprintf("%0.3f", new_prob$A)), nudge_y=0.3) +
scale_color_manual(values=PALETTE) + theme.x_axis_only
options(restore)
```

### SVM applied to non-linearly separable data with 2 variables: kernel trick

#### Data

```r
data = data.frame(x1=c(0,1,1,2,4,5,7,7,7,8,8,  3,3,4,4,4,5,5),
                  x2=c(3,5,6,6,7,7,0,4,5,2,4,  2,3,1,2,3,2,3),
                  class=c("A","A","A","A","A","A","A","A","A","A","A",  "B","B","B","B","B","B","B"))
new = data.frame(x1=6, x2=4.5)

options(repr.plot.width=4, repr.plot.height=3.3)
ggplot(data) + xlim(0,8) + ylim(0,8) +
geom_point(aes(x=x1, y=x2, color=class)) +
geom_point(aes(x=x1, y=x2), data=new) +
scale_color_manual(values=PALETTE)
options(restore)
```

#### Introduce a new variable by applying a kernel function to original data

This effectively increases the dimensionality of the data.  Here, we apply the kernel function $z = (x1-4)^2 + (x2-2)^2$

```r
data$z = (data$x1-4)^2 + (data$x2-2)^2
color = rep(NA,nrow(data))
color[data$class=="A"] = PALETTE[1]
color[data$class=="B"] = PALETTE[2]

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data[,c("x1","x2","z")], type="s", col=color, xlim=c(-5,12), ylim=c(-5,12), zlim=c(0, 30))
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

#### Find the support vectors and boundary

```r
# Compute boundary and margin planes based on support vectors 1, 2, 8, 14
p1 = as.numeric(data[1,c("x1","x2","z")])
p2 = as.numeric(data[2,c("x1","x2","z")])
p8 = as.numeric(data[8,c("x1","x2","z")])
p14 = as.numeric(data[14,c("x1","x2","z")])
v1 = p8 - p1
v2 = p2 - p1
cp = cross(v1, v2)
a = cp[1]
b = cp[2]
c = cp[3]
d1 = -dot(cp, p8)
d2 = -dot(cp, p14)
d = mean(c(d1,d2))


# Functions to computer some other other surfaces
           
kernel = function(x,y) { (x-4)^2 + (y-2)^2 }
                        
separator = function(x,y) { plane.z = -(a*x+b*y+d)/c
                            kernel.z = (x-4)^2 + (y-2)^2
                            separator.z = plane.z
                            separator.z[kernel.z > plane.z] = NA
                            separator.z }
                        
separator_shadow = function(x,y) { plane.z = -(a*x+b*y+d)/c
                                   kernel.z = (x-4)^2 + (y-2)^2
                                   separator.z = plane.z
                                   separator.z[kernel.z > plane.z] = NA
                                   separator.z[kernel.z <= plane.z] = 0  
                                   separator.z }                       
                                       
                        
# "Christmas Tree" plot

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data[,c("x1","x2","z")], type="s", col=color, xlim=c(-5,12), ylim=c(-5,12), zlim=c(0, 30))
plot3d(kernel, xlim=c(-5,12), ylim=c(-5,12), zlim=c(0,30), col="green", alpha=0.3, add=TRUE)
planes3d(a,b,c,d1, alpha=0.2)
planes3d(a,b,c,d2, alpha=0.2)
planes3d(a,b,c,d, color="blue", alpha=0.5)
plot3d(separator, xlim=c(-5,12), ylim=c(-5,12), zlim=c(0,30), col="grey",  alpha=0.5, add=TRUE)
plot3d(separator_shadow, xlim=c(-5,12), ylim=c(-5,12), zlim=c(0,30), col="grey",  alpha=1, add=TRUE)                          
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

```r
seq.x = seq(0,8,0.01)
seq.y = seq(0,7,0.01)
boundary = matrix(nrow=length(seq.x), ncol=length(seq.y))                       
rownames(boundary) = seq.x
colnames(boundary) = seq.y
for (i in 1:length(seq.x)) for (j in 1:length(seq.y)) boundary[i,j] = separator_shadow(seq.x[i],seq.y[j])
boundary.long = melt(boundary)

options(repr.plot.width=4, repr.plot.height=3.3)
ggplot() + xlab("x1") + ylab("x2") +
geom_tile(aes(x=Var1, y=Var2, fill=factor(value)), data=boundary.long, alpha=0.5) +
scale_fill_manual(values=c("gray",NA), guide="none") +
geom_point(aes(x=x1, y=x2, color=class), data=data) +
geom_point(aes(x=new$x1, y=new$x2), size=6, color="gold") +
geom_point(aes(x=new$x1, y=new$x2), color=PALETTE[1]) +
scale_color_manual(values=(PALETTE))
options(restore)
```

### Hyper-Parameters for Support Vector Machine

* kernel
  * linear
  * polynomial
    * degree: $x^2$, $x^3$, $x^4$, etc.
    * gamma: affects magnitude of kernel
    * coefficient: affects offset of kernel
  * radial basis (gaussian)
    * gamma = $\frac{1}{2\sigma^2}$: inverse standard deviation squared (variance) used for gaussian
  * hyperbolic tangent
    * coefficient
  * other kernels
* cost
* scale: normalized or not normalized
* tolerance: affects optimization
* epsilon: affects penalty calculation
* probability computation method
* other hyper-parameters

### Templates

```r
# Construct an SVM model

data.train = data.frame(x1=c(1,2,3,4,5,6,7,8), x2=c(9,7,6,5,4,3,2,1), x3=c(1,1,2,2,3,3,4,4),
                        class=c("A", "B", "A", "B", "A", "B", "A", "B"))

model = svm(class ~ x1+x2+x3, data.train, type="C-classification", kernel="polynomial", degree=3, coef0=0, gamma=0.25, cost=10, scale=FALSE, probability=TRUE)
summary(model)
model$SV
```

```r
# Use an SVM model to make predictions

new = data.frame(x1=c(5,6,7,8), x2=c(8,7,6,5), x3=c(1,5,2,6))
cutoff = 0.6

prob = attr(predict(model, new, probability=TRUE), "probabilities")
prediction = as.class(prob, "A", cutoff)

prob
prediction
cbind(new, prediction)
```
