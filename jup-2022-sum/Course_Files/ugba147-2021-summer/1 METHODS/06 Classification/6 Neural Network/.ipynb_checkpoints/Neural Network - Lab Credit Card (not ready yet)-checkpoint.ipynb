### Data Retrieval & Preparation
___

** Retrieve the dataset. **

```r
data = read.csv("../DATASETS/DATASET Credit Card Company.csv") # may take about 30 seconds
dim(data)
data
```

```r
c0 = length(which(data$Class == 0))
c1 = length(which(data$Class == 1))

c0
c1
```

```r
set.seed(12345)
library(dplyr)
data.c0.abbr = sample_n(data[data$Class == 0, ], c1)
dim(data.c0.abbr)
data.c0.abbr
```

```r
data.c1 = data[data$Class == 1, ]
dim(data.c1)
data.c1
```

```r
data.bal = rbind(data.c0.abbr, data.c1)
dim(data.bal)
data.bal
```

```r
library(neuralnet, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
model = neuralnet(Class ~ V1+V2, data.bal, hidden=c(4,3), algorithm="backprop", learningrate=0.1, threshold=0.01, stepmax=5000, rep=10, linear.output=FALSE, lifesign="minimal")
```

```r
model
```

```r
table(data$loan_status)

ggplot(data) +
geom_bar(aes(loan_status)) +
theme(axis.text.x=element_text(hjust=1.0, angle=10))
```

```r
datax = data[, c("loan_amnt", "int_rate")]

datax$outcome = NA
datax$outcome[data$loan_status %in% c("Fully Paid", "Does not meet the credit policy. Status:Fully Paid")] = "good"
datax$outcome[data$loan_status %in% c("Charged Off", "Does not meet the credit policy. Status:Charged Off")] = "bad"

datax = datax[datax$outcome %in% c("good","bad"),]
datax
```

```r
ggplot(datax) +
geom_point(aes(x=loan_amnt, y=int_rate, color=outcome))
```

<br>
### Baseline Model Based on 2007 Loans 
___

```r
outcome.predicted.0 = rep(names(tail(sort(table(datax$outcome)), 1)), nrow(datax))

cm.0 = table(datax$outcome, factor(outcome.predicted.0, c("bad", "good")))

confusionMatrix(cm.0)
```

<br>
### SVM Model Based on 2007 Loans 
___

```r
model.1 = svm(outcome ~ ., datax, scale=TRUE, type="C-classification", kernel="polynomial", cost=1, probability=TRUE)
summary(model.1)
```

```r
outcome.predicted.1 = predict(model.1, datax, probability=TRUE)

cm.1 = table(datax$outcome, outcome.predicted.1)

confusionMatrix(cm.1)
```

```r
ggplot(datax) +
geom_point(aes(x=loan_amnt, y=int_rate), data=datax[datax$outcome != outcome.predicted.1, ], size=3, color="gold") +
geom_point(aes(x=loan_amnt, y=int_rate, color=outcome.predicted.1))
```

<br>
### SVM Model Based on 2007 Loans (balanced data)
___

```r
datax.good = datax[datax$outcome=="good",]
nrow(datax.good)

datax.bad = datax[datax$outcome=="bad",]
nrow(datax.bad)

s = sample(nrow(datax.good), nrow(datax.bad), replace=FALSE)
s

datax.b = rbind(datax.good[s,], datax.bad)
dim(datax.b)
```

```r
ggplot(datax.b) +
geom_point(aes(x=loan_amnt, y=int_rate, color=outcome)) +
ggtitle("Training Data (balanced)")
```

```r
model.2 = svm(outcome ~ ., datax.b, scale=TRUE, type="C-classification", kernel="polynomial", cost=100, epsilon=0.1, probability=TRUE)
summary(model.2)
```

```r
ggplot(datax.b[model.2$index,]) +
geom_point(aes(x=loan_amnt, y=int_rate, color=outcome)) +
ggtitle("Support Vectors for Model 2")
```

```r
outcome.predicted.2 = predict(model.2, datax.b, probability=TRUE)

cm.2 = table(datax.b$outcome, outcome.predicted.2)

confusionMatrix(cm.2)
```

```r
ggplot(datax.b) +
geom_point(aes(x=loan_amnt, y=int_rate), data=datax.b[datax.b$outcome != outcome.predicted.2, ], size=3, color="gold") +
geom_point(aes(x=loan_amnt, y=int_rate, color=outcome.predicted.2)) +
ggtitle("Predictions on Training Data by Model 2")
```

```r
prob.predicted.2 = as.data.frame(attr(outcome.predicted.2, "probabilities"))

t = cbind(prob.predicted.2, outcome.predicted.2, datax.b$outcome, "hit"=outcome.predicted.2==datax.b$outcome)
t



```
