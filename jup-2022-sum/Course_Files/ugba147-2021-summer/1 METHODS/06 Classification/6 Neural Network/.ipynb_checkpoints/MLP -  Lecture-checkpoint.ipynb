### Perceptron & Linearly Separable Data
___

** Retrieve data. **

```r
data = data.frame(x1=c(1,2,3,6,14,15,16,18), x2=c(1,2,1,2,6,7,6,7), class=c("A", "A", "A", "A", "B", "B", "B", "B"))

ggplot(data) +
geom_point(aes(x=x1, y=x2, color=class)) +
geom_point(aes(x=10, y=4)) +
xlim(0,20)
```

<br>
** Diagram of a perceptron. **

* A calculating machine ...
* Weights ...
* Activation function (to force value to be 0 to 1)
* ...

```r
# Diagram of perceptron
ann = data.frame(in1=NA, in2=NA, out=NA)
model = neuralnet(out ~ in1+in2, ann, hidden=0)
plot(model, rep=1, fill="gray", cex=0.7, lwd=0.5, show.weights=FALSE, information=FALSE)
```

<br>
** Run the perceptron algorithm, step-by-step. **

The perceptron algorithm determines weights for a perceptron ... linearly separable classification ...

```r
# Initialize perceptron algorithm

set.seed(12345)

cutoff = 0.5
pass = 0
ann = data.frame(in1=data$x1, in2=data$x2, w0=NA, w1=NA, w2=NA, out=NA, predicted=NA, actual = as.numeric(data$class)-1, error=NA)
w0.new = runif(1); w1.new = runif(1); w2.new = runif(1)

cat("Initial weights: w0 =", w0.new, ", w1 =", w1.new, ", w2 =", w2.new, "\n")
cat("Cutoff =", cutoff, "\n")
ann
```

```r
# Following initialization, perform one pass at a time of perceptron algorithm, re-execute this cell for subsequent passes

pass = pass + 1

for (j in 1:nrow(ann))
  {
    
    ann$w0[j] = w0.new; ann$w1[j] = w1.new; ann$w2[j] = w2.new

    ann$out[j] = sigmoid(ann$w0[j] + ann$w1[j]*ann$in1[j] + ann$w2[j]*ann$in2[j])    
    ann$predicted[j] = as.numeric(ann$out[j] >= cutoff)
    ann$error[j] = abs(ann$actual[j] - ann$out[j])

    w0.new = ann$w0[j] + (ann$actual[j] - ann$out[j])
    w1.new = ann$w1[j] + (ann$actual[j] - ann$out[j]) * ann$in1[j]
    w2.new = ann$w2[j] + (ann$actual[j] - ann$out[j]) * ann$in2[j]  
    
  }

w0.final = ann$w0[nrow(ann)]; w1.final = ann$w1[nrow(ann)]; w2.final = ann$w2[nrow(ann)] 

cat("Pass", pass, "\n")
cat("Final weights: w0 =", w0.final, ", w1 =", w1.final, ", w2 =", w2.final, "\n")
cat("Mean Absolute Error = ", mean(abs(ann$error)), "\n")
ann
```

```r
# Compute the perceptron's output for the unclassified datapoint
out.10.4 = sigmoid(w0.final + w1.final*10 + w2.final*4)
predict.10.4 = as.numeric(out.10.4 >= cutoff)
cat("Input = (10,4) >>>> Output =", out.10.4, ">>>> Prediction =", predict.10.4, "\n")  
```

<br>
** Run the perceptron algorithm, 1000 passes. **

Note that weights stablize and error shrinks.

```r
# Initialize and perform 1000 passes of perceptron algorithm 

set.seed(12345)

cutoff = 0.5
ann = data.frame(in1=data$x1, in2=data$x2, w0=NA, w1=NA, w2=NA, out=NA, predicted=NA, actual = as.numeric(data$class)-1, error=NA)
w0.new = runif(1); w1.new = runif(1); w2.new = runif(1)
cat("Initial weights: w0 =", w0.new, ", w1 =", w1.new, ", w2 =", w2.new, "\n")
cat("Cutoff =", cutoff, "\n")

for (pass in 1:1000)
  {

    for (j in 1:nrow(ann))
      {
    
        ann$w0[j] = w0.new; ann$w1[j] = w1.new; ann$w2[j] = w2.new

        ann$out[j] = sigmoid(ann$w0[j] + ann$w1[j]*ann$in1[j] + ann$w2[j]*ann$in2[j])
        ann$predicted[j] = as.numeric(ann$out[j] >= cutoff)
        ann$error[j] = abs(ann$actual[j] - ann$out[j])

        w0.new = ann$w0[j] + (ann$actual[j] - ann$out[j])
        w1.new = ann$w1[j] + (ann$actual[j] - ann$out[j]) * ann$in1[j]
        w2.new = ann$w2[j] + (ann$actual[j] - ann$out[j]) * ann$in2[j]  
    
      }

  }

w0.final = ann$w0[nrow(ann)]; w1.final = ann$w1[nrow(ann)]; w2.final = ann$w2[nrow(ann)] 

cat("Pass", pass, "\n")
cat("Final weights: w0 =", w0.final, ", w1 =", w1.final, ", w2 =", w2.final, "\n")
cat("Mean Absolute Error = ", mean(abs(ann$error)), "\n")
ann
```

```r
# Compute the perceptron's output for the unclassified datapoint
out.10.4 = sigmoid(w0.final + w1.final*10 + w2.final*4)
predict.10.4 = as.numeric(out.10.4 >= cutoff)
cat("Input = (10,4) >>>> Output =", out.10.4, ">>>> Prediction =", predict.10.4, "\n")
```

### Multi-Layer Perceptron (with 1 hidden layer of weights)
___

** Retrieve data. **

```r
data.B = data.frame(x1=c(1,2,3,5,6,14,15,16,18), x2=c(1,2,1,2,2,6,7,6,7), class=c("A", "A", "A", "B", "A", "B", "B", "B", "B"))

ggplot(data.B) +
geom_point(aes(x=x1, y=x2, color=class)) +
geom_point(aes(x=10, y=4)) +
xlim(0,20)
```

<br>
** Diagram of a 2-layer perceptron. **

2 inputs, 1 hidden layer with 3 nodes, 1 output.

```r
# Diagram of perceptron
ann = data.frame(in1=NA, in2=NA, out=NA)
model = neuralnet(out ~ in1+in2, ann, hidden=3)
plot(model, rep=1, fill="gray", cex=0.7, lwd=0.5, show.weights=FALSE, information=FALSE)
```

```r
# Initialize and perform perceptron algorithm using neuralnet() function

set.seed(12345)

cutoff = 0.5
ann = data.frame(in1=data.B$x1, in2=data.B$x2, out=NA, predicted=NA, actual=as.numeric(data.B$class)-1, error=NA)

ann$out = ann$actual
model = neuralnet(out ~ in1+in2, ann, hidden=3, algorithm="backprop", learningrate=0.1, threshold=0.01, stepmax=5000, rep=10, linear.output=FALSE, lifesign="minimal")
best = which.min(model$result.matrix["error",])
cat("Best repetition =", best, "\n")
```

```r
# Present results of back-propagation algorithm

ann$out = model$net.result[[best]]
ann$predicted = as.numeric(ann$out >= cutoff)
ann$error = abs(ann$out - ann$actual)

cat("Cutoff =", cutoff, "\n")
cat("Mean Absolute Error = ", mean(abs(ann$error)), "\n")
model$weights[[best]]
ann
plot(model, rep=best, fill="gray", cex=0.7, lwd=0.5, information=FALSE)
```

```r
# Compute the perceptron's output for the unclassified datapoint
out.10.4 = compute(model, data.frame(in1=10,in2=4), rep=best)$net.result
predict.10.4 = as.numeric(out.10.4 >= cutoff)
cat("Input = (10,4) >>>> Output =", out.10.4, ">>>> Prediction =", predict.10.4, "\n")
```

### About Perceptrons & Multi-Layer Perceptrons
___

#### Perceptrons
* The perceptron algorithm will converge if the data are linearly separable, given enough steps.
    * Though it will converge, it may not converge to the optimal solution (i.e., maximum margin).
    * If margin is large, then the perceptron algorithm will converge quickly.
* The perceptron algorithm will not converge if the data are not linearly separable.
* The perceptron algorithm's performance/results are affected by the data class-ordering.

#### Multi-Layer Perceptrons
* ANY function can be approximated by a multi-layer perceptron with 1 hidden layer (proof by Cybenko, and later by Hornik).
* Often, a multi-layer perceptron with >1 hidden layer can approximate a multi-layer perceptron with 1 hidden layer, but use fewer nodes.
* A multi-layer perceptron with many nodes corresponds to a very complex class separation boundary.
* Building a multi-layer perceptron model involves finding the node-connection weights. 
* Building a multi-layer perceptron model with many layers is sometimes referred to as "deep learning".

<br>
### Meta-Parameters for Multi-Layer Perceptrons
___

* Activation function (sigmoid, hyperbolic tangent, or some other function)
* Number of hidden layers
* Number of hidden nodes
* Arrangement of hidden layers & nodes
* Initial weights (or randomized weights)
* Number of repetitions
* Algorithm (back-propagation or some other algorithm)
* Learning rate (if back-propagation algorithm is used)
* Other algorithm parameters (if some other algorithm is used)
* Stopping criteria (threshold, maximum number of steps, and/or some other criteria)
