Load function libraries, define additional useful functions, and set defaults here.

```r
options(warn=-1)

# Load some required functions
library(e1071,     verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(caret,     verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(ggplot2,   verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(rpart,    verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(ggdendro, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(dplyr,    verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)

# Define some visualization constants
restore = list(repr.plot.width=8, repr.plot.height=3)
PALETTE = c("#00A9FF", "#F8766D", "#7CAE00", "#C77CFF", "#CD9600", "#00BE67", "#FF61CC", "#00BFC4")
theme.x_axis_only = theme(axis.title.y=element_blank(), axis.text.y=element_blank(), axis.ticks.y=element_blank(), panel.grid.major.y=element_blank(), panel.grid.minor.y=element_blank())
theme.y_axis_only = theme(axis.title.x=element_blank(), axis.text.x=element_blank(), axis.ticks.x=element_blank(), panel.grid.major.x=element_blank(), panel.grid.minor.x=element_blank())
theme.no_axes = theme.x_axis_only + theme.y_axis_only
theme.no_legend = theme(legend.position="none")

# Set some visualization formatting defaults
options(restore)
update_geom_defaults("point", list(size=3))
theme_update(plot.title=element_text(size=10, hjust=0.5), plot.subtitle=element_text(size=8, face="italic", hjust=0.5), axis.title=element_text(size=7), axis.text=element_text(size=7), strip.text=element_text(size=7), strip.text.y=element_text(angle=90), legend.title=element_blank(), legend.text=element_text(size=7))

# Define some useful functions
as.class.cutoff = function(postprob, cutoff=0.5) { factor(colnames(postprob)[2-as.numeric(postprob[,1]>=cutoff)]) }
as.class.best = function(postprob) { factor(as.character(aaply(1:nrow(postprob), 1, function(i) { colnames(postprob)[as.numeric(postprob[i,]) == max(as.numeric(postprob[i,]))] } ))) }
ggtree = function(model) { ggplot() + 
                           geom_segment(aes(x=x, y=y, xend=xend, yend=yend), dendro_data(model)$segments) + 
                           geom_label(aes(x=x, y=y, label=label), dendro_data(model)$labels, size=2.5) +
                           geom_label(aes(x=x, y=y, label=label, fill=label), dendro_data(model)$leaf_labels, size=2.5) +
                           theme.no_axes + theme.no_legend }
pick = function(v, f) { if (class(v) %in% c("data.frame", "matrix")) colnames(v)[v == f(v)] else names(v)[v == f(v)] }
```

### Bootstrap Aggregating
___
Build a model comprising other constituent models, which are all built using a single predictive analytic method and set of hyper-parameter settings, but based on different bootstrap samples of the data.  The ensemble model makes predictions based on the most popular of the constituent models' predictions.

You can think of a bootstrap aggregated model as a committee of experts.  Each expert has studied a particular field of knowledge, different from (but possibly overlapping with) those that the other experts have studied. 

Bootstrap aggregating is also known as "bagging".  

<br>
#### Model based on original data.
Here we consider a decision tree model based on the original data.  We use the model to predict the class of a new observation.

```r
model = rpart(class ~ x1+x2, data, method="class")
prediction = predict(model, new, type="class")
data.frame(new, prediction)

options(repr.plot.width=2.5, repr.plot.height=2.5)
ggtree(model) + scale_fill_manual(values=PALETTE)
options(restore)
```

<br>
#### Bootstrap samples.
A bootstrap sample of data is a set of randomly selected observations from that data, with repetition.  The size of the bootstrap sample is the same as the size of the data, i.e., the number of observations in the bootstrap sample is the same as the number of observations in the data.  Usually, many of the observations in the bootstrap sample are duplicates of each other. 

Here we take 3 different bootstrap samples of the data.

```r
set.seed(12345)
data.1 = sample_n(data, nrow(data), replace=TRUE)
data.2 = sample_n(data, nrow(data), replace=TRUE)
data.3 = sample_n(data, nrow(data), replace=TRUE)
```

<br>
#### Models based on bootstrap samples.
We can build several models, using the same method and hyper-parameters, but based on the different bootstrap samples of data.  Each model is an "expert" on the particular subset of data represented in the bootstrap sample on which it is based.

Here we build 3 different such models and use them to make 3 different predictions of the class of a new observation.

```r
mymodel.1 = rpart(class ~ x1+x2, data.1, method="class")
prediction.1 = predict(mymodel.1, new, type="class")
data.frame(new, prediction.1)

options(repr.plot.width=2.5, repr.plot.height=2.5)
ggtree(mymodel.1) + scale_fill_manual(values=PALETTE)
options(restore)
```

```r
mymodel.2 = rpart(class ~ x1+x2, data.2, method="class")
prediction.2 = predict(mymodel.2, new, type="class")
data.frame(new, prediction.2)

options(repr.plot.width=2.5, repr.plot.height=2.5)
ggtree(mymodel.2) + scale_fill_manual(values=PALETTE)
options(restore)
```

```r
mymodel.3 = rpart(class ~ x1+x2, data.3, method="class")
prediction.3 = predict(mymodel.3, new, type="class")
data.frame(new, prediction.3)

options(repr.plot.width=2.5, repr.plot.height=2.5)
ggtree(mymodel.3) + scale_fill_manual(values=PALETTE)
options(restore)
```

<br>
#### Vote.
We can compare the predictions of models and vote for the most popular prediction.  

Here models 1 and 2 predict that the class of the new observation is A, and model 3 predicts B.  The vote is 2 to 1 in favor of A, so the bootstrap aggregated model predicts A.

```r
prediction.1 = as.character(prediction.1)
prediction.2 = as.character(prediction.2)
prediction.3 = as.character(prediction.3)

freq = table(c(prediction.1, prediction.2, prediction.3))

data.frame(prediction.1, prediction.2, prediction.3, pick(freq, max))
```

### Boosting
___
Build a model, make predictions, and note the prediction errors.  Then build another model, using the same predictive analytic method and hyper-parameter settings, but based on a resample of the data that emphasizes the observations associated with the prediction errors.  Continue in this way to build more models.  The ensemble model makes predictions based on the most popular of these models' predictions.

You can think of a boosted model as a committee of experts.  Each expert (except the first one) has studied a specialized sub-field of knowledge that focuses on the aspects least understood by another expert. 

#### Build model \#1 based on the original data.
Here we consider a decision tree model based on the original data.

```r
data.1 = data
mymodel.1 = rpart(class ~ x1+x2, data.1, method="class")

options(repr.plot.width=2.5, repr.plot.height=2.5)
ggtree(mymodel.1) + scale_fill_manual(values=PALETTE)
options(restore)
```

<br>
#### Assess model #1's ability to predict the original data's classes.

```r
prediction.data.1 = predict(mymodel.1, data.1, type="class")
hit.1 = prediction.data.1 == data.1$class

data.frame(observation_ID=row.names(data.1), data.1$class, prediction.data.1, hit.1)
```

<br>
#### Resample dataset #1, choose observations misclassified by model \#1 with high probability.  This is dataset #2.

Here we choose incorrectly predicted observations with 10 times greater chance than correctly predicted observations.

```r
emphasis = 10
prob = ((as.numeric(!hit.1) * (emphasis-1)) + 1)
prob = prob / sum(prob)

set.seed(12345)
resampled_observations = sample(1:nrow(data.1), replace=TRUE, prob=prob)
data.2 = data.1[resampled_observations, ]
cbind(data.2, hit.1=hit.1[resampled_observations])
```

<br>
#### Build model #2 based on dataset #2.

We expect this model to better predict the classes of those observations that model #1 predicted incorrectly.

```r
mymodel.2 = rpart(class ~ x1+x2, data.2, method="class")

options(repr.plot.width=2.5, repr.plot.height=2.5)
ggtree(mymodel.2) + scale_fill_manual(values=PALETTE)
options(restore)
```

<br>
#### Assess model #2's ability to predict dataset #2's classes.

```r
prediction.data.2 = predict(mymodel.2, data.2, type="class")
hit.2 = prediction.data.2 == data.2$class

data.frame(data.2$class, prediction.data.2, hit.2)
```

<br>
#### Resample dataset #2, choose observations misclassified by model #2 with high probability.  This is dataset #3.

Here we choose incorrectly predicted observations with 10 times greater chance than correctly predicted observations.

```r
emphasis = 10
prob = ((as.numeric(!hit.2) * (emphasis-1)) + 1)
prob = prob / sum(prob)

set.seed(12345)
resampled_observations = sample(1:nrow(data.2), replace=TRUE, prob=prob)
data.3 = data.2[resampled_observations, ]
cbind(data.3, hit.2=hit.2[resampled_observations])
```

<br>
#### Build model #3 based on dataset #3.

We expect this model to better predict the classes of those observations that model #2 predicted incorrectly.

```r
mymodel.3 = rpart(class ~ x1+x2, data.3, method="class")

options(repr.plot.width=2.5, repr.plot.height=2.5)
ggtree(mymodel.3) + scale_fill_manual(values=PALETTE)
options(restore)
```

<br>
#### Vote.
We can compare the predictions of models and vote for the most popular prediction.  

Here model 1 predicts that the class of the new observation is B, and models 2 and 3 predict A.  The vote is 2 to 1 in favor of A, so the bootstrap aggregated model predicts A.

```r
prediction.1 = predict(mymodel.1, new, type="class")
prediction.2 = predict(mymodel.2, new, type="class")
prediction.3 = predict(mymodel.3, new, type="class")

prediction.1 = as.character(prediction.1)
prediction.2 = as.character(prediction.2)
prediction.3 = as.character(prediction.3)

freq = table(c(prediction.1, prediction.2, prediction.3))

data.frame(prediction.1, prediction.2, prediction.3, pick(freq, max))
```

### Stacking
___
Build several models and make predictions.  Then build a new model, using some predictive analytic method, based on the predictions that the other models made about the original data.

You can think of a stacked model as an expert on a committee of other experts. 

#### Build several models, using the different methods and/or different hyper-parameter settings, but based on the same data.

Here we build 3 different such models and use them to make 3 different sets of predictions of the classes of the origial data.

```r
mymodel.1 = rpart(class ~ x1+x2, data, method="class")
mymodel.2 = naiveBayes(class ~ x1+x2, data)
mymodel.3 = svm(class ~ x1+x2, data, kernel="radial", cost=1, probability=FALSE)
```

```r
data.stack = data.frame(predict.1=predict(mymodel.1, data, type="class"),
                        predict.2=predict(mymodel.2, data),
                        predict.3=predict(mymodel.3, data),
                        class=data$class)
data.stack
```

<br>
#### Build a model based on predictions of the other models.

Here we build a logistic regression model.

```r
model.stack = glm(class ~ predict.1+predict.2+predict.3, binomial(link="logit"), data.stack)
```

<br>
#### Express a new observation in terms of how its class is predicted by the other models.

Here the new observation x1=5.2,x2=4 is expressed as the observation predict1=B,predict2=B,predict3=B, because model1 predicts that the class of x1=5.2,x2=4 is B, model2 predicts that the class of x1=5.2,x2=4 is B, and model3 predicts that the class of x1=5.2,x2=4 is B. 

```r
new.stack = data.frame(predict.1 = predict(mymodel.1, new, type="class"),
                       predict.2 = predict(mymodel.2, new),
                       predict.3 = predict(mymodel.3, new))

new.stack
```

<br>
#### Use the new model to predict the class of a new observation.

Here the new model predicts that the class of observation predict1=B,predict2=B,predict3=B is B, so it equivalently predicts that the calss of observation x1=5.2,x2=4 is B.

```r
prob.B = predict(model.stack, new.stack, type="response")
prob = data.frame(A=1-prob.B, B=prob.B)
as.class.cutoff(prob, 0.5)
```
