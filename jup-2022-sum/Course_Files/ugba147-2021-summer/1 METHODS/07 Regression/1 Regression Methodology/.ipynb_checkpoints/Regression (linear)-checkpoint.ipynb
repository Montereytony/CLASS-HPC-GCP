### Linear Regression

Linear regression is a method to construct models to predict numeric values.  The various versions of linear regression all use optimization techniques to find the best fit line (or plane or hyper-plane) through a set of observations.

Terminology:
* **response, dependent, outcome:** the variable to be predicted
* **explanatory, independent, predictor, input:** the variables used to make the prediction

<br>

### Simple Linear Regression

Construct a model to predict the numeric value of an output variable given one predictor variable.

#### Data

```r
data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), outcome=c( 1.0, 0.5, 3.0, 3.0, 4.0, 5.5, 7.0, 6.0))
new = data.frame(x1=5.0)

data
new
```

```r
options(repr.plot.width=8, repr.plot.height=1)
ggplot(data) + geom_point(aes(x=x1, y=0), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=0)) + 
geom_text(aes(x=x1, y=0, label=outcome), nudge_y=0.3) + 
ylim(-0.5, 0.5) + theme.x_axis_only
options(restore)
```

<br>

#### Prediction for new observation?

```r
ggplot(data) + geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=seq(0,7,length.out=nrow(data))))
```

<br>

#### Best fit line through the observations.

```r
model = lm(outcome~x1, data)
new$outcome = predict(model, new)

data.frame(model$coefficients)
new
ggplot(data) + geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=new$outcome))
```

<br>

#### The best fit line is determined using the ordinary least squares (OLS) optimization method.

Minimize the sum of the squares of the differences between the line and the actual values.

```r
model = lm(outcome~x1, data)
new$outcome = predict(model, new)

ggplot(data) +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=new$outcome)) +
geom_point(aes(x=x1, y=predict(model, data)), color=PALETTE[4])
```

```r
data$predict = predict(model, data)
data$error = data$outcome - data$predict
data$square_error = data$error^2
data

data.frame(sum(data$error), sum(data$square_error))
```

<br>

#### There are many lines that minimize differences.  There are many lines that minimize absolute differences.  There is **exactly one** line that minimizes square of differences.

For example, consider this simple case.

```r
data.simple_case = data.frame(x1=c(3,3,6,6), outcome=c(6,3,6,3))
model.simple_case = lm(outcome~x1, data.simple_case)
ggplot(data.simple_case) + geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_abline(aes(intercept=model.simple_case$coefficients["(Intercept)"], slope=model.simple_case$coefficients["x1"])) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model.simple_case, data.simple_case)), color=PALETTE[4]) +
xlim(0,10) + ylim(0,10)
```

### Multiple Linear Regression

Construct a model to predict the numeric value of an outcome variable given several other predictor variables.

#### Data

```r
data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), x2=c(5.0, 7.0, 2.0, 4.0, 3.0, 1.0, 2.0, 2.5), outcome=c( 1.0, 0.5, 3.0, 3.0, 4.0, 5.5, 7.0, 6.0))
new = data.frame(x1=5.0, x2=5.0)

data
new

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data, type="s", col=PALETTE[3], xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
spheres3d(data.frame(x1=new$x1, x2=new$x2, outcome=seq(0,10,1)), radius=0.35, color="grey50", add=TRUE)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

```r
model = lm(outcome~x1+x2, data)

data.frame(model$coefficients)
new

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data, type="s", col=PALETTE[3], xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
planes3d(a=model$coefficients["x1"], b=model$coefficients["x2"], c=-1, d=model$coefficients["(Intercept)"], color="grey50", alpha=0.3, add=TRUE)
spheres3d(data.frame(x1=new$x1, x2=new$x2, outcome=seq(0,10,1)), radius=0.35, color="grey50", add=TRUE)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

```r
new$outcome = predict(model, new)

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(rbind(data, new), type="s", col=c(rep(PALETTE[3],nrow(data)),"grey50"), xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
planes3d(a=model$coefficients["x1"], b=model$coefficients["x2"], c=-1, d=model$coefficients["(Intercept)"], color="grey50", alpha=0.3, add=TRUE)
segments3d(data.frame(x1=new$x1, x2=new$x2, outcome=c(0,new$outcome)), color="grey50", lw=5)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

```r
grid.arrange(ggplot(data) + geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=new$outcome)),
             ggplot(data) + geom_point(aes(x=x2, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x2, y=new$outcome)),
             nrow=1)
```

### Linear Regression with Log Outcome Transformation

Construct a model to predict the numeric value of the log of an outcome variable given predictor variable(s).  This may be useful if the variables are not well linearly correlated. 

#### Data

```r
data = data.frame(x1=c( 1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), outcome=c( 7.0, 6.5, 6.0, 2.0, 2.0, 1.0, 1.0, 1.5))
new = data.frame(x1=5.0)

data
new

ggplot(data) + geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, seq(0,7,length.out=nrow(data)))) + ylim(-1,10)
```

```r
model = lm(outcome~x1, data)
new$outcome = predict(model, new)

data.frame(model$coefficients)
new

ggplot(data) +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_point(aes(x=x1, y=predict(model, data), color="predicted")) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome, color="actual")) + geom_point(aes(x=new$x1, y=new$outcome)) +
ylim(-1,10) + ylab("outcome") +
scale_color_manual(values=c(actual=PALETTE[3], predicted=PALETTE[4]))
```

```r
data$log_outcome = log(data$outcome)
data = data[,c(1,3,2)]
new$outcome = NULL

data
new
```

```r
model = lm(log_outcome~x1, data)
new$log_outcome = predict(model, new)

data.frame(model$coefficients)
new

ggplot(data) +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_point(aes(x=x1, y=predict(model, data), color="predicted")) +
geom_segment(aes(x=x1, y=log_outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=log_outcome, color="actual")) + geom_point(aes(x=new$x1, y=new$log_outcome)) +
ylim(-1,10) + ylab("log(outcome)") +
scale_color_manual(values=c(actual=PALETTE[3], predicted=PALETTE[4]))
```

```r
new$outcome = exp(predict(model, new))

new

ggplot(data) +
geom_smooth(aes(x=x1, y=exp(predict(model, data))), method="loess", se=FALSE) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=exp(predict(model, data))), color=PALETTE[4]) +
geom_point(aes(x=x1, y=exp(predict(model, data)), color="predicted")) +
geom_point(aes(x=x1, y=outcome, color="actual")) +
geom_point(aes(x=new$x1, y=new$outcome)) +
ylim(-1,10) + ylab("outcome") +
scale_color_manual(values=c(actual=PALETTE[3], predicted=PALETTE[4]))
```

### Linear Regression with Polynomial Transformation

Construct a model to predict the numeric value of an outcome variable given predictor variables, where some of the variables are powers of other variables.  This may be useful if the variables are not well linearly correlated. 

#### Data

```r
data = data.frame(x1=c( 1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), outcome=c( 8.0, 6.5, 6.0, 2.0, 1.0, 3.0, 5.0, 9.0))
new = data.frame(x1=5.0)

data
new

ggplot(data) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, seq(0,10,length.out=nrow(data)))) +
ylim(0,10)
```

```r
model = lm(outcome~x1, data)
new$outcome = predict(model, new)

data.frame(model$coefficients)
new

ggplot(data) +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=new$outcome)) +
geom_point(aes(x=x1, y=predict(model, data)), color=PALETTE[4])
```

```r
data$x1.square = data$x1^2
data = data[,c(1,3,2)]

new$outcome = NULL
new$x1.square = new$x1^2

data
new
```

Model: $outcome = 12.70 + 0.41{x_1}^2 + -4.14x_1$

```r
model = lm(outcome~x1.square+x1, data)
new$outcome = predict(model, new)

data.frame(model$coefficients)
new

ggplot(data) +
geom_smooth(aes(x=x1, y=predict(model, data)), method="loess", se=FALSE) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=new$outcome)) +
ylab("outcome")
```

Model: $outcome = 12.70 + 0.41{x_1}^2 + -4.14x_1$

```r
model = lm(outcome~poly(x1,2,raw=TRUE), data)
new$x1.square = NULL
new$outcome = predict(model, new)

data.frame(model$coefficients)
new

ggplot(data) +
geom_smooth(aes(x=x1, y=predict(model, data)), method="loess", se=FALSE) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=new$outcome)) +
ylab("outcome")
```

### Linear Regression with Categorical Predictor Variables

The linear regression method assumes all variables are numerical, so convert categorical variables to dummy variables if necessary. 

#### Data

```r
x2_labels = c("A","B","C")

data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0),
                  x2=factor(c("A","B","C","A","A","B","C","C"), x2_labels),
                  outcome=c( 1.0, 0.5, 3.0, 3.0, 4.0, 5.5, 7.0, 6.0))
new = data.frame(x1=5.0, x2=factor("B", x2_labels))

data
new
```

```r
ggplot(data) + geom_point(aes(x=x1, y=outcome, color=x2)) +
geom_point(aes(x=new$x1, y=seq(0,7,length.out=nrow(data)))) +
scale_color_manual(values=PALETTE[c(1,2,5)])
```

<br>

#### Convert categorical variable(s) to dummy variables. 

```r
data.d = dummy.data.frame(data, drop=FALSE, omit.constants=FALSE)
data.d$x2A = NULL
data.d

new.d = dummy.data.frame(new, drop=FALSE, omit.constants=FALSE)
new.d$x2A = NULL
new.d
```

<br>

#### Build a linear regression model based on the data expressed in terms of dummy variables.

Model: $outcome = 0.05 + 0.71x_1 + -0.08x_{2B} + 0.76x_{2C}$

```r
model.d = lm(outcome~., data.d)
new$outcome = predict(model.d, new.d)

data.frame(model.d$coefficients)
new
```

<br>

#### Use the model to predict the new observation's outcome.

Also note how the model predicts the existing data's outcomes. 

```r
ggplot(data) +
geom_point(aes(x=x1, y=predict(model.d, data.d)), color=PALETTE[4]) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model.d, data.d)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome, color=x2)) +
geom_point(aes(x=new$x1, y=new$outcome, color=new$x2)) +
ylab("outcome") +
scale_color_manual(values=PALETTE[c(1,2,5)])
```

<br>

#### Alternatively, build a linear regression model based on the expressed in terms of categorical variables (to be converted to dummy variables), and use the model to predict a new observation's outcome.

The `lm` function accepts recognizes categorical variables and automatically converts them to dummy variables, but assumes all possible categorical variable values are represented in the data -  so, be careful new observations that have categorical variable values not represented in the data. 

Model: $outcome = 0.05 + 0.71x_1 + -0.08x_{2B} + 0.76x_{2C}$

```r
model = lm(outcome~., data)
data.frame(model$coefficients)
```

```r
new$outcome = predict(model, new)
new_outcome.dummies = new$outcome

ggplot(data) +
geom_point(aes(x=x1, y=predict(model, data)), color=PALETTE[4]) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome, color=x2)) + geom_point(aes(x=new$x1, y=new$outcome, color=new$x2)) +
ylab("outcome") +
scale_color_manual(values=PALETTE[c(1,2,5)])
```

```r
model.e = lm(outcome~., data.e)
model.e
```

```r
new$outcome = predict(model.e, new.e)
new_outcome.enumerated = new$outcome

ggplot(data) +
geom_point(aes(x=x1, y=predict(model.e, data.e)), color=PALETTE[4]) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model.e, data.e)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome, color=x2)) + geom_point(aes(x=new$x1, y=new$outcome, color=new$x2)) +
scale_color_manual(values=PALETTE)
```

<br>
#### Compare dummies model vs. index model. 

The two models make different predictions.  The dummies version predicts the existing data slightly better. 

```r
data.frame(new_outcome.dummies, new_outcome.enumerated)

grid.arrange(ggplot(data) +
             geom_point(aes(x=x1, y=predict(model, data)), color=PALETTE[4]) +
             geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
             geom_point(aes(x=x1, y=outcome, color=x2)) + geom_point(aes(x=new$x1, y=new_outcome.dummies, color=new$x2)) +
             scale_color_manual(values=PALETTE) + ggtitle("Dummies"),

             ggplot(data) +
             geom_point(aes(x=x1, y=predict(model.e, data.e)), color=PALETTE[4]) +
             geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model.e, data.e)), color=PALETTE[4]) +
             geom_point(aes(x=x1, y=outcome, color=x2)) + geom_point(aes(x=new$x1, y=new_outcome.enumerated, color=new$x2)) +
             scale_color_manual(values=PALETTE) + ggtitle("Enumerated"),

             nrow=1)
```

### Example of Complex Linear Regression

You can combine and expand log and polynomial linear regression to form more complex models.

#### Data

```r
x2_labels = c("A","B","C")
data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), 
                  x2=factor(c("A","B","C","A","B","C","A","A"), x2_labels),
                  outcome=c( 1.0, 3.5, 6.0, 2.0, 1.0, 3.0, 5.0, 9.0))
new = data.frame(x1=5.0, x2=factor("A", x2_labels))

dim(data)
data
data.frame(new)
```

```r
ggplot(data) + geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, seq(0,7,length.out=nrow(data)))) + ylim(0,10)
```

Model: $outcome = 15.31 + -0.083{x_1}^3 + 1.94{x_1}^2 + -17.18x_1 + log(x_1) + -0.97x_{2B} + 0.07x_{2C}$

```r
model = lm(log(outcome)~poly(x1,3,raw=TRUE)+log(x1)+x2, data)
new$outcome = predict(model, new)

model

ggplot(data) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=exp(predict(model, data))), color=PALETTE[4]) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) + geom_point(aes(x=new$x1, y=exp(new$outcome))) +
geom_point(aes(x=x1, y=exp(predict(model, data))), color=PALETTE[4])
```
