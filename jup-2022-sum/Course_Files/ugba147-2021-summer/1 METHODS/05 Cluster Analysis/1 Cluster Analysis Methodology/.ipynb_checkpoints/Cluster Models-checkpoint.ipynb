## Discourse
___

* **Descriptive Data Analytics** _(also known as Descriptive Data Mining, also known as Unsupervised Machine Learning)_**:**  <br>
Methods to construct a "descriptive model" that exposes interesting relationships in data.  No source of truth exists to confirm quality of the model.


* **Cluster Model:** A kind of descriptive model that exposes similarity/dissimilarity among observations in the data.  Subsets of observations are organized into separate "clusters".


* **Evaluation:**  Assess the quality of a cluster model.


* **Business Value Formula:**  Assess the quality of a cluster model in terms of business value.


* **Business Applications:** Market segmentation, others.

### Evaluation of Cluster Model

**Distance: A measure of observation dissimilarity**

```r
data[1:2, ]
```

<br>

Distance between 2 observations in 1 variable.

$
distance = \, \mid b_{z_1} - a_{z_1} \mid
$

```r
distance_a_b = abs(data[1,"z1"] - data[2,"z1"])
data.frame(distance_a_b)

options(repr.plot.width=8, repr.plot.height=1)
ggplot(data[1:2,]) + xlim(-2,2) + 
geom_segment(x=data[1,"z1"], y=0, xend=data[2,"z1"], yend=0) +
geom_point(aes(x=z1, y=0), color="gray50") +
geom_text(aes(x=z1, y=0, label=name)) +
scale_color_manual(values=PALETTE) + theme.x_axis_only + theme.no_legend
options(restore)
```

<br>

Distance between 2 observations in 2 variables. (Pythagorean Theorem)

$
distance = \, \sqrt{ (b_{z_1} - a_{z_1})^2 + (b_{z_2} - a_{z_2})^2 }
$

```r
distance_a_b = sqrt((data[1,"z1"]-data[2,"z1"])^2 + (data[1,"z2"]-data[2,"z2"])^2)
data.frame(distance_a_b)

ggplot(data[1:2,]) + xlim(-2,2) + ylim(-2,2) +
geom_segment(x=data[1,"z1"], y=data[1,"z2"], xend=data[2,"z1"], yend=data[2,"z2"]) +
geom_point(aes(x=z1, y=z2), color="gray50") +
geom_text(aes(x=z1, y=z2, label=name)) +

scale_color_manual(values=PALETTE) + theme.no_legend
```

<br>

Distance between 2 observations in 3 variables.

$
distance = \, \sqrt{ (b_{z_1} - a_{z_1})^2 + (b_{z_2} - a_{z_2})^2 + (b_{z_3} - a_{z_3})^2 }
$

```r
distance_a_b = sqrt((data[1,"z1"]-data[2,"z1"])^2 + (data[1,"z2"]-data[2,"z2"])^2 + (data[1,"z3"]-data[2,"z3"])^2)
data.frame(distance_a_b)
```

```r
distance_a_b = sqrt((data[1,"z1"]-data[2,"z1"])^2 + (data[1,"z2"]-data[2,"z2"])^2 + (data[1,"z3"]-data[2,"z3"])^2)
data.frame(distance_a_b)

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data[1:2,c("z1","z2","z3")], type="s", size=10, col="gray50", xlim=c(-2,2), ylim=c(-2,2), zlim=c(-2,2))
segments3d(data[1:2,"z1"], data[1:2,"z2"], data[1:2,"z3"], lwd=4, col=PALETTE[1])
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

<br>

Distance between 2 observations in many variables.

$
distance = \, \sqrt{ (b_{z_1} - a_{z_1})^2 + (b_{z_2} - a_{z_2})^2 \, + \, ... \, + \, (b_{z_n} - a_{z_n})^2 }
$

**Intracluster Diffusion**

For a cluster, the mean of the distances between each pair of observations.

```r
data
```

```r
data.A = data[data$cluster=="A",]
data.A
```

```r
dist(data.A[,c("z1","z2","z3")])

distance_matrix.A = as.matrix(dist(data.A[,c("z1","z2","z3")]))
distance_matrix.A
```

```r
data.frame(observation_pair=c("a_b","b_c","a_c"), distance=c(distance_matrix.A[1,2], distance_matrix.A[2,3], distance_matrix.A[1,3]))
data.frame(cluster="A", intracluster_diffusion=mean(dist(data.A[,c("z1","z2","z3")])))

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data.A[,c("z1","z2","z3")], type="s", size=10, col=PALETTE[1], xlim=c(-2,2), ylim=c(-2,2), zlim=c(-2,2))
segments3d(data.A[c(1,2),"z1"], data.A[c(1,2),"z2"], data.A[c(1,2),"z3"], lwd=4, col=PALETTE[1])
segments3d(data.A[c(2,3),"z1"], data.A[c(2,3),"z2"], data.A[c(2,3),"z3"], lwd=4, col=PALETTE[1])
segments3d(data.A[c(3,1),"z1"], data.A[c(3,1),"z2"], data.A[c(3,1),"z3"], lwd=4, col=PALETTE[1])
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

**Mean Intracluster Diffusion ...** (that's INTRA)

For a cluster model (comprising many clusters), the mean of intracluster diffusions.

```r
data
```

```r
data.A = data[data$cluster=="A",]
data.B = data[data$cluster=="B",]
data.C = data[data$cluster=="C",]
data.D = data[data$cluster=="D",]

data.A
data.B
data.C
data.D
```

```r
summary = data.frame(cluster=c("A","B","C","D"),
                     intracluster_diffusion=c(mean(dist(data.A[,c("z1","z2","z3")])),
                                              0,
                                              mean(dist(data.C[,c("z1","z2","z3")])),
                                              mean(dist(data.D[,c("z1","z2","z3")]))))

mean_intracluster_diffusion=mean(summary$intracluster_diffusion)

summary
data.frame(mean_intracluster_diffusion)

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data[,c("z1","z2","z3")], type="s", col=PALETTE[c(1,1,1,2,3,3,4,4)], xlim=c(-2,2), ylim=c(-2,2), zlim=c(-2,2))
segments3d(data[c(1,2),"z1"], data[c(1,2),"z2"], data[c(1,2),"z3"], lwd=4, col=PALETTE[1])
segments3d(data[c(2,3),"z1"], data[c(2,3),"z2"], data[c(2,3),"z3"], lwd=4, col=PALETTE[1])
segments3d(data[c(3,1),"z1"], data[c(3,1),"z2"], data[c(3,1),"z3"], lwd=4, col=PALETTE[1])
segments3d(data[c(5,6),"z1"], data[c(5,6),"z2"], data[c(5,6),"z3"], lwd=4, col=PALETTE[3])
segments3d(data[c(7,8),"z1"], data[c(7,8),"z2"], data[c(7,8),"z3"], lwd=4, col=PALETTE[4])
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

**Mean Intercluster Diffusion** ... (that's INTER)

For a cluster model (comprising many clusters), the mean of the distances between each pair of cluster centroids.

```r
data.A
```

```r
data.A[,c("z1","z2","z3")]
colMeans(data.A[,c("z1","z2","z3")])
t(colMeans(data.A[,c("z1","z2","z3")]))
as.data.frame(t(colMeans(data.A[,c("z1","z2","z3")])))
centroid.A = as.data.frame(t(colMeans(data.A[,c("z1","z2","z3")])))
```

**Diffusion Ratio**

Diffusion ratio accounts for observation similarity within a cluster and observation dissimilarity between clusters, both of which are desirable characteristics of a cluster model.  Diffusion ratio is a popular measure of cluster model quality. 

```r
diffusion_ratio = mean_intercluster_diffusion / mean_intracluster_diffusion

data.frame(mean_intracluster_diffusion, mean_intercluster_diffusion, diffusion_ratio)
```

```r
roster = read.csv("Student Roster.csv")
roster
```

```r
roster.norm = scale(roster[,-1])
roster.norm
```

<br>

Construct a cluster model.  Each of 5 clusters comprises similar students.

```r
cluster = cutree(hclust(dist(roster.norm), method="complete"), k=5)
as.data.frame(table(cluster))

cluster
```

<br>

Make team assignments by distributing similar students across teams, so that each team comprises dissimilar students.  

```r
cluster.1 = which(cluster==1)
cluster.2 = which(cluster==2)
cluster.3 = which(cluster==3)
cluster.4 = which(cluster==4)
cluster.5 = which(cluster==5)

member.1 = cluster.4[1:11]
member.2 = c(cluster.4[12:21], cluster.3[1])
member.3 = cluster.3[2:12]
member.4 = c(cluster.2, cluster.5[1:3])
member.5 = c(cluster.1, cluster.3[13:16], cluster.5[4])

assign = data.frame(team=1:11, member.1=roster$Name[member.1], member.2=roster$Name[member.2], member.3=roster$Name[member.3], member.4=roster$Name[member.4], member.5=roster$Name[member.5])

assign.gender = data.frame(team=1:11, roster$Gender[member.1], roster$Gender[member.2], roster$Gender[member.3], roster$Gender[member.4], roster$Gender[member.5])
assign$gender = rowMeans(assign.gender[,-1])
assign.terms = data.frame(team=1:11, roster$Terms[member.1], roster$Terms[member.2], roster$Terms[member.3], roster$Terms[member.4], roster$Terms[member.5])
assign$terms = rowMeans(assign.terms[,-1])
assign.quant = data.frame(team=1:11, roster$Quant[member.1], roster$Quant[member.2], roster$Quant[member.3], roster$Quant[member.4], roster$Quant[member.5])
assign$quant = rowMeans(assign.quant[,-1])
assign.business = data.frame(team=1:11, roster$Business[member.1], roster$Business[member.2], roster$Business[member.3], roster$Business[member.4], roster$Business[member.5])
assign$business = rowMeans(assign.business[,-1])
assign.compsci = data.frame(team=1:11, roster$CompSci[member.1], roster$CompSci[member.2], roster$CompSci[member.3], roster$CompSci[member.4], roster$CompSci[member.5])
assign$compsci = rowMeans(assign.compsci[,-1])/5

assign
```

```r
# Construct cluster model using Gaussian mixture model by expectation-maximization

datax = data.frame(x1=c(2,2.5,4,6.5,8,9,10,10.5), x2=c(3.5,9,8.5,11,2.5,1,4,7.5), x3=c(10,9,10.5,2,1.5,1,9.5,11))

number_of_clusters = 5

set.seed(12345)
datax$cluster = Mclust(datax, G=number_of_clusters)$classification
datax
```

```r
# Construct cluster model using hierarchical agglomeration

datax = data.frame(x1=c(2,2.5,4,6.5,8,9,10,10.5), x2=c(3.5,9,8.5,11,2.5,1,4,7.5), x3=c(10,9,10.5,2,1.5,1,9.5,11))
datax.norm = scale(datax)

number_of_clusters = 5

datax$cluster = cutree(hclust(dist(datax.norm), method="complete"), k=number_of_clusters)
datax
```

```r
# Visualize the dendrogram associated with hierarchical agglomeration

datax = data.frame(x1=c(2,2.5,4,6.5,8,9,10,10.5), x2=c(3.5,9,8.5,11,2.5,1,4,7.5), x3=c(10,9,10.5,2,1.5,1,9.5,11))
datax.norm = scale(datax)

tree = hclust(dist(datax.norm), method="complete")
ggdendrogram(tree, theme_dendro=FALSE) + xlab("observation #") + ylab("distance")
```

```r
# Construct cluster model using k-means

datax = data.frame(x1=c(2,2.5,4,6.5,8,9,10,10.5), x2=c(3.5,9,8.5,11,2.5,1,4,7.5), x3=c(10,9,10.5,2,1.5,1,9.5,11))
datax.norm = scale(datax)

number_of_clusters = 5

set.seed(12345)
datax$cluster = kmeans(datax.norm, number_of_clusters, nstart=1000)$cluster
datax
```
