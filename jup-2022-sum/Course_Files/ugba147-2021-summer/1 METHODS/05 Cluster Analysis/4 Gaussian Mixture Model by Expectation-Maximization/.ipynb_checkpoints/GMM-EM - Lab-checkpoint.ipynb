```r
data = read.csv("../DATASETS/DATASET Redfin.csv")
dim(data)
head(data)
```

```r
data.num = data[, c("PRICE","BEDS","BATHS","SQUARE.FEET","DAYS.ON.MARKET","HOA.MONTH","LATITUDE","LONGITUDE")]
data.ok = na.omit(data.num)
dim(data.ok)
head(data.ok)
```

```r
library(mclust)

m = Mclust(data.ok, G=2)
#summary(m, parameters=TRUE)
m$parameters


#plot(m, what="classification")
#str(m)
data.ok$class = m$classification
head(data.ok)
```

```r
data = data.frame(x1=c(2.0, 2.5, 4.0, 6.5, 8.0, 9.0, 10.0, 10.5))
data

ggplot(data) +
geom_point(aes(x=x1, y=0), size=6, color="black") + xlim(0,11) +
theme.x_axis_only
```

```r
gaussian = function(x, mean, sd) { (1/(sd*sqrt(2*pi))) * exp(-0.5 * ((x-mean)/sd)^2) }

mean.A = 9
sd.A = 1
mean.B = 10
sd.B = 1

data.frame(class=c("A","B"), mean=c(mean.A, mean.B), sd=c(sd.A, sd.B))

ggplot(data) + xlim(0,11) +
geom_vline(aes(xintercept=mean.A), color=PALETTE[1], lty="dashed") +
geom_vline(aes(xintercept=mean.B), color=PALETTE[2], lty="dashed") +
stat_function(fun=gaussian, args=c(mean=mean.A, sd=sd.A), n=500, geom="area", fill=PALETTE[1], alpha=0.3) +
stat_function(fun=gaussian, args=c(mean=mean.B, sd=sd.B), n=500, geom="area", fill=PALETTE[2], alpha=0.3) +
geom_point(aes(x=x1, y=0), color="black", size=6)
```

```r
prob.8.A = gaussian(8, mean.A, sd.A)
prob.8.B = gaussian(8, mean.B, sd.B)          
                
weight.A.8 = prob.8.A / (prob.8.A + prob.8.B)
weight.B.8 = 1 - weight.A.8

best_guess = if (weight.A.8 > weight.B.8) "A" else "B"
    
data.frame(x1=8, prob.A=prob.8.A, prob.B=prob.8.B, weight.A=weight.A.8, weight.B=weight.B.8)

ggplot(data) + xlim(0,11) +
geom_vline(aes(xintercept=mean.A), color=PALETTE[1], lty="dashed") +
geom_vline(aes(xintercept=mean.B), color=PALETTE[2], lty="dashed") +
stat_function(fun=gaussian, args=c(mean=mean.A, sd=sd.A), n=500, geom="area", fill=PALETTE[1], alpha=0.3) +
stat_function(fun=gaussian, args=c(mean=mean.B, sd=sd.B), n=500, geom="area", fill=PALETTE[2], alpha=0.3) +
geom_point(aes(x=8, y=0, color=weight.A.8), size=6) +
scale_color_gradient(high=PALETTE[1], low=PALETTE[2], guide="none") +
geom_text(aes(x=8, y=prob.8.A, label=sprintf("%0.3f",prob.8.A)), color=PALETTE[1]) +
geom_text(aes(x=8, y=prob.8.B, label=sprintf("%0.3f",prob.8.B)), color=PALETTE[2])
```

```r
# First iteration

data$prob.A = gaussian(data$x1, mean.A, sd.A)
data$prob.B = gaussian(data$x1, mean.B, sd.B)          
                     
data$weight.A = aaply(1:nrow(data), 1, function(i) { data$prob.A[i]/(data$prob.A[i]+data$prob.B[i]) })
data$weight.B = 1 - data$weight.A

data$best_guess = aaply(1:nrow(data), 1, function(i) { if (data$weight.A[i] > data$weight.B[i]) "A" else "B" })
                    
data

ggplot(data) + xlim(0,11) +
geom_vline(aes(xintercept=mean.A), color=PALETTE[1], lty="dashed") +
geom_vline(aes(xintercept=mean.B), color=PALETTE[2], lty="dashed") +
stat_function(fun=gaussian, args=c(mean=mean.A, sd=sd.A), n=500, geom="area", fill=PALETTE[1], alpha=0.3) +
stat_function(fun=gaussian, args=c(mean=mean.B, sd=sd.B), n=500, geom="area", fill=PALETTE[2], alpha=0.3) +
geom_point(aes(x=x1, y=0, color=weight.A), size=6) +
scale_color_gradient(high=PALETTE[1], low=PALETTE[2], guide="none") +
geom_text(aes(x=x1, y=prob.A, label=sprintf("%0.3f",prob.A)), color=PALETTE[1]) +
geom_text(aes(x=x1, y=prob.B, label=sprintf("%0.3f",prob.B)), color=PALETTE[2])
```

```r
# Iterate ...

new_mean.A = sum(data$weight.A * data$x1) / sum(data$weight.A)
new_sd.A = sqrt(sum(data$weight.A * (data$x1 - new_mean.A)^2) / sum(data$weight.A))

new_mean.B = sum(data$weight.B * data$x1) / sum(data$weight.B)
new_sd.B = sqrt(sum(data$weight.B * (data$x1 - new_mean.B)^2) / sum(data$weight.B))

data.frame(class=c("A","B"), mean=c(new_mean.A, new_mean.B), sd=c(new_sd.A, new_sd.B))

data$prob.A = gaussian(data$x1, new_mean.A, new_sd.A)
data$prob.B = gaussian(data$x1, new_mean.B, new_sd.B)          
                     
data$weight.A = aaply(1:nrow(data), 1, function(i) { data$prob.A[i]/(data$prob.A[i]+data$prob.B[i]) })
data$weight.B = 1 - data$weight.A
                    
data$best_guess = aaply(1:nrow(data), 1, function(i) { if (data$weight.A[i] > data$weight.B[i]) "A" else "B" })

data

ggplot(data) + xlim(0,11) +
geom_vline(aes(xintercept=new_mean.A), color=PALETTE[1], lty="dashed") +
geom_vline(aes(xintercept=new_mean.B), color=PALETTE[2], lty="dashed") +
stat_function(fun=gaussian, args=c(mean=new_mean.A, sd=new_sd.A), n=500, geom="area", fill=PALETTE[1], alpha=0.3) +
stat_function(fun=gaussian, args=c(mean=new_mean.B, sd=new_sd.B), n=500, geom="area", fill=PALETTE[2], alpha=0.3) +
geom_point(aes(x=x1, y=0, color=weight.A), size=6) +
scale_color_gradient(high=PALETTE[1], low=PALETTE[2], guide="none") +
geom_text(aes(x=x1, y=prob.A, label=sprintf("%0.3f",prob.A)), color=PALETTE[1]) +
geom_text(aes(x=x1, y=prob.B, label=sprintf("%0.3f",prob.B)), color=PALETTE[2])
```
