### Apparatus

```r
# Load some required functions
library(mclust,    verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(ggdendro,  verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(ggplot2,   verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(gridExtra, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(rgl,       verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE) # All rglwidget() calls must use elementID="myplot"
library(htmltools, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)

# Set some visualization constants
PALETTE = c("#00A9FF", "#F8766D", "#7CAE00", "#C77CFF", "#CD9600", "#00BE67", "#FF61CC", "#00BFC4")
theme.no_legend = theme(legend.position="none")

# Set some visualization formatting defaults
options(repr.plot.width=8, repr.plot.height=2.5)
options(rgl.useNULL=TRUE)
update_geom_defaults("point", list(size=3, colour=PALETTE[1]))
update_geom_defaults("text", list(size=2.5, vjust=-1.5))
theme_update(plot.title=element_text(size=10, hjust=0.5), plot.subtitle=element_text(size=8, face="italic", hjust=0.5), axis.title=element_text(size=7), axis.text=element_text(size=7), strip.text=element_text(size=7), strip.text.y=element_text(angle=90), legend.title=element_blank(), legend.text=element_text(size=7))
```

<br>
### Retrieve Data
___

```r
data = data.frame(name=c("A","B","C","D","E","F","G","H","I","J","K","L","M","N"),
                  x1=c(2.5, 2, 9, 6.5, 8, 4, 10, 10.5, 3, 5, 3, 6, 1, 1.5),
                  x2=c(9, 3.5, 1, 11, 2.5, 8.5, 4, 7.5, 3, 4, 7, 6, 1, 1.5),
                  x3=c(9, 10, 1, 2, 1.5, 10.5, 9.5, 11, 3, 5, 9, 6, 1, 1.5))

dim(data)
data
```

<br>
### Cluster by Hierarchical Agglomeration
___

** Organize the data into 3 clusters, using hierarchical agglomeration. Assume disimilarity measure is Euclidean distance, linkage is complete. **

Show the dissimilarity matrix, dendrogram, 2-D scatter plots (x1 vs x2, x1 vs x3, x2 vs x3) color-coded by cluster membership, 3-D scatter plot (x1 vs x2 vs x3) color-coded by cluster membership.

```r
dissimilarity_matrix = dist(data[,c("x1", "x2", "x3")], method="euclidean")
dissimilarity_matrix
```

```r
tree = hclust(dissimilarity_matrix, method="complete")
ggdendrogram(tree, theme_dendro=FALSE) + xlab("observation #") + ylab("distance")
```

```r
data$class = cutree(tree, k=3)
```

```r
p1 = ggplot(data) + geom_point(aes(x=x1, y=x2, color=as.factor(class))) + geom_text(aes(x=x1, y=x2, label=name)) + scale_color_manual(values=PALETTE) + theme.no_legend
p2 = ggplot(data) + geom_point(aes(x=x1, y=x3, color=as.factor(class))) + geom_text(aes(x=x1, y=x3, label=name)) + scale_color_manual(values=PALETTE) + theme.no_legend
p3 = ggplot(data) + geom_point(aes(x=x2, y=x3, color=as.factor(class))) + geom_text(aes(x=x2, y=x3, label=name)) + scale_color_manual(values=PALETTE) + theme.no_legend

grid.arrange(p1, p2, p3, nrow=1)
```

```r
p = open3d(); view3d(0,0, zoom=0.65)
plot3d(data[,c("x1","x2", "x3")], type="s", size=3, col=PALETTE[data$class], aspect=TRUE)
grid3d(c("x", "y", "z"))
rglwidget(elementId="myplot", width=300, height=300)
rgl.close()
```

### Cluster by k-Means
___

** Organize the data into 3 clusters, using k-means. Assume 1 random start (set randomization seed to 12345). **

Show the 2-D scatter plots (x1 vs x2, x1 vs x3, x2 vs x3) color-coded by cluster membership, 3-D scatter plot (x1 vs x2 vs x3) color-coded by cluster membership.

### Cluster by Gaussian Mixture Model
___

** Organize the data into 3 clusters, using a one-dimensional gaussian mixture model applied to x1. **

Show a plot of x1 values overlayed with the 3 gaussian curves, color-coded by cluster membership.

```r
m = Mclust(data[,c("x1")], G=3, modelNames="V")
data$class = m$classification
```

```r
gaussian = function(x, mean, sd) { (1/(sd*sqrt(2*pi))) * exp(-0.5 * ((x-mean)/sd)^2) }

mean.1 = as.numeric(m$parameters$mean[1])
mean.2 = as.numeric(m$parameters$mean[2])
mean.3 = as.numeric(m$parameters$mean[3])

sigma.1 = sqrt(m$parameters$variance$sigmasq[1])
sigma.2 = sqrt(m$parameters$variance$sigmasq[2])
sigma.3 = sqrt(m$parameters$variance$sigmasq[3])

ggplot(data) +
stat_function(fun=gaussian, args=c(mean=mean.1, sd=sigma.1), n=500, geom="area", fill=PALETTE[1], alpha=0.3) +
stat_function(fun=gaussian, args=c(mean=mean.2, sd=sigma.2), n=500, geom="area", fill=PALETTE[2], alpha=0.3) +
stat_function(fun=gaussian, args=c(mean=mean.3, sd=sigma.3), n=500, geom="area", fill=PALETTE[3], alpha=0.3) +
geom_point(aes(x=x1, y=0, color=as.factor(class))) +
scale_color_manual(values=PALETTE) + theme.no_legend
```
