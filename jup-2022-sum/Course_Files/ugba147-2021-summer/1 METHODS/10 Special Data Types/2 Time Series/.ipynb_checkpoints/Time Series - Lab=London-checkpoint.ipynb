### About the Business Situation & Data
___
The European Union recently asked all member governments to look at smart meters as part of measures to upgrade the energy supply and tackle climate change.  After an initial study, the British government decided to adopt smart meters as part of its plan to update its aging energy system. The goal is for all 26 million homes in England, Wales, and Scotland to have smart meters by 2020.

#### Data
This dataset comprises aggregated energy consumption readings for 5,567 London households that took part in the Low Carbon London project led by UK Power Networks from January 1, 2013 through January 31, 2014.

#### Objective
Predict energy consumption patterns in 2014 based on energy consumption patterns in 2013.

### Prepare Data
___

#### Partition into train and test data.
Train data is one year January 1, 2013 through December 31, 2013.  Test data is January 1, 2014 through January 31, 2014.

```r
data.train = data[as.character(data$day) <= "2013-12-31", ]
data.test = data[as.character(data$day) >= "2014-01-01", ]
```

```r
ggplot() +
geom_line(aes(x=as.Date(day), y=energy, group=1, color="train"), data=data.train) +
geom_line(aes(x=as.Date(day), y=energy, group=1, color="test"), data=data.test) +
geom_vline(aes(xintercept=as.Date("2013-12-31"), linetype="train/test boundary"), lwd=0.2) +
geom_vline(aes(xintercept=as.Date("2014-01-08"), linetype="vantage"), lwd=0.2) +
scale_color_manual(values=PALETTE[c(3,1)]) +
scale_linetype_manual(values=c("dashed","solid"))
```

### Model & Prediction: Classical Method
___
Prediction by the classical method uses one model to predict various numbers of time steps ahead based on a static vantage.  It treats the time variable as a numeric variable.

<br>
Here we convert the train data's time variable to a numeric variable.  The train data's earliest time coincides with numeric value 1.

```r
datax.train = data.train
datax.train$day = as.numeric(datax.train$day)

datax.train$day[1]
datax.train$day[nrow(datax.train)]
```

<br>
Here we convert the test data's time variable to a numeric variable.  The test data's earliest time coincides with numeric value 1 more than the train data's latest time.

```r
datax.test = data.test
datax.test$day = as.numeric(datax.test$day)

datax.test$day[1]
datax.test$day[nrow(datax.test)]
```

<br>
Here we construct a linear regression model (with 4th degree polynomial transformation) to predict energy consumption given a day.

```r
model = lm(energy ~ poly(day,4, raw=TRUE), datax.train)
model
```

<br>
Here we choose a vantage of day 373 (2014-01-08).

```r
vantage = as.numeric(factor("2014-01-08", data$day))
vantage
```

<br>
Use the model to predict energy consumption 1, 2, 3 days ahead of the vantage.

```r
x = 374

i = model$coefficients[1]
c1 = model$coefficients[2]
c2 = model$coefficients[3]
c3 = model$coefficients[4]
c4 = model$coefficients[5]

x
i
c1
c2
c3
c4

p = unname(i + c1*x + c2*x^2 + c3*x^3 + c4*x^4)
p
```

```r
new.1 = data.frame(day=vantage+1)
new.2 = data.frame(day=vantage+2)
new.3 = data.frame(day=vantage+3)

result.c = datax.test
result.c$energy.predicted = NA

result.c$energy.predicted[9] = predict(model, new.1)
result.c$energy.predicted[10] = predict(model, new.2)
result.c$energy.predicted[11] = predict(model, new.3)

result.c
```

```r
ggplot(result.c) +
geom_line(aes(x=day, y=energy, color="actual")) +
geom_line(aes(x=day, y=energy.predicted, color="predicted")) +
geom_vline(aes(xintercept=vantage, linetype="vantage"), lwd=0.2) +
scale_color_manual(values=PALETTE[c(3,5)]) +
scale_linetype_manual(values="solid")
```

<br>

### Prepare Data: Convert from Time Series Form to Cross-Sectional Form
___
The order of observations in **time series data** is significant.  In general, data analytic methods do not work on time series data because they do not respect the order of observations. 

The order of observations in **cross-sectional data** is not significant.  In general, data analytic methods do work on cross-sectional data.

Time series data can be converted to cross-sectional data, given a set of **lookbacks** and a set of **lookaheads**.

Here we choose these lookbacks and lookaheads:
* lookbacks: 0, 1, 2, 3, 4, 5, 6, 7 days
* lookaheads: 1, 2, 3 days (3 day horizon starting at 1 day ahead)

#### Add variables for lookbacks.
For each lookback, add a variable that aligns with the outcome variable shifted **up** by the appropriate number of days.

```r
data.train.cs = data.train
data.train.cs$lookback.1 = shift_up(data.train.cs$energy, 1)
data.train.cs$lookback.2 = shift_up(data.train.cs$energy, 2)
data.train.cs$lookback.3 = shift_up(data.train.cs$energy, 3)
data.train.cs$lookback.4 = shift_up(data.train.cs$energy, 4)
data.train.cs$lookback.5 = shift_up(data.train.cs$energy, 5)
data.train.cs$lookback.6 = shift_up(data.train.cs$energy, 6)
data.train.cs$lookback.7 = shift_up(data.train.cs$energy, 7)

head(data.train.cs, 10)
```

<br>

#### Add variables for lookaheads.
For each lookahead, add a variable that aligns with the outcome variable shifted **down** by the appropriate number of days.

```r
data.train.cs$lookahead.1 = shift_down(data.train.cs$energy, 1)
data.train.cs$lookahead.2 = shift_down(data.train.cs$energy, 2)
data.train.cs$lookahead.3 = shift_down(data.train.cs$energy, 3)

head(data.train.cs, 10)
tail(data.train.cs, 10)
```

<br>

#### Discard incomplete days.
Here we discard the last 3 days and first 7 days of train data.  The last 3 days are discarded cause they do not include 3 day lookaheads.  The first 7 days are discarded because they do not include 7 day lookbacks.  

```r
data.train.csx = data.train.cs[1:(nrow(data.train.cs)-3), ]
data.train.csx = data.train.csx[(7+1):nrow(data.train.csx), ]

head(data.train.csx)
tail(data.train.csx)
```

<br>

### Model & Prediction: Direct Method
___
Prediction by the direct method uses several models to predict various numbers of time steps ahead based on a static vantage.  

#### Build 3 models, one for each lookahead.
Here we use the linear regression method to build all 3 models.

```r
model.1 = lm(lookahead.1 ~ energy+lookback.1+lookback.2+lookback.3+lookback.4+lookback.5+lookback.6+lookback.7, data.train.csx)
model.2 = lm(lookahead.2 ~ energy+lookback.1+lookback.2+lookback.3+lookback.4+lookback.5+lookback.6+lookback.7, data.train.csx)
model.3 = lm(lookahead.3 ~ energy+lookback.1+lookback.2+lookback.3+lookback.4+lookback.5+lookback.6+lookback.7, data.train.csx)
```

<br>

#### Prepare a vantage.
Here we choose to prepare the vantage to be 2014-01-08.  The first 7 days of test data are reserved for lookbacks.

```r
vantage = data.test[8, ]

vantage$lookback.1 = data.test$energy[8-1]
vantage$lookback.2 = data.test$energy[8-2]
vantage$lookback.3 = data.test$energy[8-3]
vantage$lookback.4 = data.test$energy[8-4]
vantage$lookback.5 = data.test$energy[8-5]
vantage$lookback.6 = data.test$energy[8-6]
vantage$lookback.7 = data.test$energy[8-7]

vantage
```

<br>

#### Predict energy consumption over the horizon.
* Use model #1 given the first 7 days of test data to predict energy consumption at 1 day ahead of the vantage.
* Use model #2 given the first 7 days of test data to predict energy consumption at 2 days ahead of the vantage.
* Use model #3 given the first 7 days of test data to predict energy consumption at 3 days ahead of the vantage.

```r
result.d = data.test
result.d$energy.predicted = NA

result.d$energy.predicted[9] = predict(model.1, vantage)
result.d$energy.predicted[10] = predict(model.2, vantage)
result.d$energy.predicted[11] = predict(model.3, vantage)

result.d
```

```r
ggplot(result.d) +
geom_line(aes(x=as.Date(day), y=energy, color="actual")) +
geom_line(aes(x=as.Date(day), y=energy.predicted, color="predicted")) +
geom_vline(aes(xintercept=as.Date("2014-01-08"), linetype="vantage"), lwd=0.2) +
scale_color_manual(values=PALETTE[c(3,2)]) +
scale_linetype_manual(values="solid")
```

<br>

### Model & Prediction: Recursive Method
___
Prediction by the recursive method re-uses one model to predict one time step ahead based on a moving vantage and previous predictions.  

#### Build a model for 1 day lookahead.
Here we use the linear regression method to build the model.

```r
model.1 = lm(lookahead.1 ~ energy+lookback.1+lookback.2+lookback.3+lookback.4+lookback.5+lookback.6+lookback.7, data.train.csx)
```

<br>

#### Prepare a vantage.
Here we choose to prepare the vantage to be 2014-01-08.  The first 7 days of test data are reserved for lookbacks.

```r
vantage.0 = data.test[8, ]

vantage.0$lookback.1 = data.test$energy[8-1]
vantage.0$lookback.2 = data.test$energy[8-2]
vantage.0$lookback.3 = data.test$energy[8-3]
vantage.0$lookback.4 = data.test$energy[8-4]
vantage.0$lookback.5 = data.test$energy[8-5]
vantage.0$lookback.6 = data.test$energy[8-6]
vantage.0$lookback.7 = data.test$energy[8-7]

vantage.0
```

<br>

#### Predict energy consumption at 1 day ahead of vantage.
* Use the model given the first 7 days of test data to predict energy consumption at 1 day ahead of the vantage.

```r
result.1 = predict(model.1, vantage.0)
data.frame(result.1)
```

<br>

#### Prepare a new vantage (1 day ahead of the original vantage).
Here we prepare a new vantage to be 2014-01-09.  Use the previous prediction for energy consumption at this new vantage.

```r
vantage.1 = data.frame(day=data.test$day[9], energy=result.1)

vantage.1$lookback.1 = data.test$energy[9-1]
vantage.1$lookback.2 = data.test$energy[9-2]
vantage.1$lookback.3 = data.test$energy[9-3]
vantage.1$lookback.4 = data.test$energy[9-4]
vantage.1$lookback.5 = data.test$energy[9-5]
vantage.1$lookback.6 = data.test$energy[9-6]
vantage.1$lookback.7 = data.test$energy[9-7]

vantage.1
```

<br>

#### Predict energy consumption at 1 day ahead of new vantage.
* Use the model given the previous prediction and the earlier 6 days of test data to predict energy consumption at 1 day ahead of the **new** vantage.

```r
result.2 = predict(model.1, vantage.1)
data.frame(result.2)
```

<br>

#### Prepare a 2nd new vantage (2 days ahead of the original vantage).
Here we prepare a new vantage to be 2014-01-10.  Use the previous 2 predictions for energy consumption at the 2 new vantages.

```r
vantage.2 = data.frame(day=data.test$day[10], energy=result.2)

vantage.2$lookback.1 = result.1
vantage.2$lookback.2 = data.test$energy[10-2]
vantage.2$lookback.3 = data.test$energy[10-3]
vantage.2$lookback.4 = data.test$energy[10-4]
vantage.2$lookback.5 = data.test$energy[10-5]
vantage.2$lookback.6 = data.test$energy[10-6]
vantage.2$lookback.7 = data.test$energy[10-7]

vantage.2
```

<br>

#### Predict energy at 1 day ahead of 2nd new vantage.
* Use model given the previous 2 predictions and the earlier 5 days of test data to predict energy 1 day ahead of the 2nd new vantage.

```r
result.3 = predict(model.1, vantage.2)
data.frame(result.3)
```

<br>

#### Compile results.

```r
result.r = data.test
result.r$energy.predicted = NA

result.r$energy.predicted[9] = result.1
result.r$energy.predicted[10] = result.2
result.r$energy.predicted[11] = result.3

result.r
```

```r
ggplot(result.r) +
geom_line(aes(x=as.Date(day), y=energy, color="actual")) +
geom_line(aes(x=as.Date(day), y=energy.predicted, color="predicted")) +
geom_vline(aes(xintercept=as.Date("2014-01-08"), linetype="vantage"), lwd=0.2) +
scale_color_manual(values=PALETTE[c(3,4)]) +
scale_linetype_manual(values="solid")
```

<br>

### Compare Results
___
Predictions made using direct and recursive methods are similar, but not exactly equivalent.

```r
ggplot() +
geom_line(aes(x=as.Date(day), y=energy, color="actual"), data.test) +
geom_line(aes(x=as.Date(day), y=energy.predicted, color="predicted direct"), data=result.d) +
geom_line(aes(x=as.Date(day), y=energy.predicted, color="predicted recursive"), data=result.r) +
geom_vline(aes(xintercept=as.Date("2014-01-08"), linetype="vantage"), lwd=0.2) +
scale_color_manual(values=PALETTE[c(3,2,4)]) +
scale_linetype_manual(values="solid")
```
