### About the Business Situation & Data
___
To better follow the energy consumption, the government wants energy suppliers to install smart meters in every home in England, Wales and Scotland. There are more than 26 million homes for the energy suppliers to get to, with the goal of every home having a smart meter by 2020.

This roll out of meter is lead by the European Union who asked all member governments to look at smart meters as part of measures to upgrade our energy supply and tackle climate change. After an initial study, the British government decided to adopt smart meters as part of their plan to update our ageing energy system.
In this dataset, you will find a refactorised version of the data from the London data store, that contains the energy consumption readings for a sample of 5,567 London Households that took part in the UK Power Networks led Low Carbon London project between November 2011 and February 2014. The data from the smart meters seems associated only to the electrical consumption.

There is infomations on the ACORN classification details that you can find in this report or the website of CACI.

### Retrieve Data
___

```r
data = read.csv("../DATASETS/DATASET London.csv") # may take about 30 seconds
dim(data)
head(data)
```

```r
data = data[order(data$LCLid, data$day),]
dim(data)
head(data)
```

```r
data.long = melt(data, id.vars="LCLid", measure.vars="day") # note: this converts day (type factor) to value (type character)
dim(data.long)
head(data.long)
```

```r
data.start = dcast(data.long, LCLid ~ variable, min.character)
dim(data.start)
head(data.start)
```

```r
data.stop = dcast(data.long, LCLid ~ variable, max.character)
dim(data.stop)
head(data.stop)
```

```r
data.startstop = data.start[,1,drop=FALSE]
data.startstop$start = data.start[,2]
data.startstop$stop = data.stop[,2]
dim(data.startstop)
head(data.startstop)
```

```r
ggplot(data.startstop) +
geom_line(aes(x=LCLid, y=as.Date(start), color="start", group=1)) +
geom_line(aes(x=LCLid, y=as.Date(stop), color="stop", group=1)) + scale_color_manual(values=PALETTE[c(3,2)]) +
theme.y_axis_only
```

```r
earliest_start = min.character(data.startstop$start)
latest_stop = max.character(data.startstop$stop)

earliest_start
latest_stop
```

```r
data.ids = data.startstop[(data.startstop$start <= "2013-01-01") & (data.startstop$stop >= "2014-01-31"), ]
dim(data.ids)
data.ids
```

```r
data.ok = data[(data$LCLid %in% data.ids$LCLid) & (as.character(data$day) >= "2013-01-01") & (as.character(data$day) <= "2014-01-31"), ]
dim(data.ok)
head(data.ok)
```

```r
data.long = melt(data.ok, id.vars=c("LCLid","day"), measure.vars="energy_sum")
dim(data.long)
head(data.long,400)
```

```r
data.sum = dcast(data.long, day ~ variable, sum)
dim(data.sum)
head(data.sum)
```

```r
ggplot(data.sum) +
geom_line(aes(x=as.Date(day), y=energy_sum, group=1)) +
geom_vline(aes(xintercept=as.Date("2013-12-31")), lwd=0.2, linetype="dashed")
```

```r
shift_up = function(v, i) { c(rep(NA, i), v[1:(length(v)-i)])  }

data.cs = data.frame(day=data.sum$day, energy=data.sum$energy_sum)
data.cs$lookback.1 = shift_up(data.cs$energy, 1)
data.cs$lookback.2 = shift_up(data.cs$energy, 2)
data.cs$lookback.3 = shift_up(data.cs$energy, 3)
data.cs$lookback.4 = shift_up(data.cs$energy, 4)
data.cs$lookback.5 = shift_up(data.cs$energy, 5)
data.cs$lookback.6 = shift_up(data.cs$energy, 6)
data.cs$lookback.7 = shift_up(data.cs$energy, 7)

head(data.cs, 10)
```

```r
shift_down = function(v, i) { c(v[(i+1):length(v)], rep(NA, i)) }

data.cs$lookahead.7 = shift_down(data.cs$energy, 7)

head(data.cs, 10)
tail(data.cs, 10)
```

```r
data.cs.clip = data.cs[1:(nrow(data.cs)-7), ]
data.cs.clip = data.cs.clip[(7+1):nrow(data.cs.clip), ]

head(data.cs.clip)
tail(data.cs.clip)
```

```r
data.train = data.cs.clip[data.cs.clip$day <= "2013-12-31", ]
data.test = data.cs.clip[data.cs.clip$day > "2013-12-31", ]
```
