### Retrieve Data
___

```r
data.raw = read.csv("../DATASETS/DATASET Hotels.csv", encoding="UTF-8")
dim(data.raw)
head(data.raw)
```

```r
length(unique(data.raw$name))
data.frame(sort(unique(data.raw$name)))
```

```r
head(data.raw$reviews.text)
```

```r
i_rated = !is.na(data.raw$reviews.rating)
i_unrated = is.na(data.raw$reviews.rating)
```

```r
corpus = VCorpus(VectorSource(data.raw$reviews.text))

inspect(corpus[[1]])
inspect(corpus[[2]])
inspect(corpus[[3]])
```

```r
# May take about 1 minute

corpus = tm_map(corpus, content_transformer(iconv), from="UTF-8", to="latin1") # delete this line for some systems
corpus = tm_map(corpus, content_transformer(tolower))
corpus = tm_map(corpus, removeNumbers)
corpus = tm_map(corpus, removePunctuation, ucp=TRUE)
corpus = tm_map(corpus, removeWords, stopwords("English"))
corpus = tm_map(corpus, stripWhitespace)

inspect(corpus[[1]])
inspect(corpus[[2]])
inspect(corpus[[3]])
```

```r
data = as.data.frame(as.matrix(removeSparseTerms(DocumentTermMatrix(corpus), sparse=0.98)))

data.rated = data[i_rated,]
data.unrated = data[i_unrated,]

dim(data.rated)
head(data.rated)
dim(data.unrated)
head(data.unrated)
```

```r
data.1 = data.rated
data.1$outcome = data.raw$reviews.rating[i_rated]

data.2 = data.rated
data.2$class = factor(data.raw$reviews.rating[i_rated], levels=0:5, labels=c("terrible","bad","poor","fair","good","excellent"))

data.3 = data.rated
data.3$class = factor(rep(NA, nrow(data.rated)), c("positive","negative"))
data.3$class[data.raw$reviews.rating[i_rated] >= 3] = "positive"
data.3$class[data.raw$reviews.rating[i_rated] <= 2] = "negative"

head(data.1)
head(data.2)
head(data.3)
```

```r
model.lm = lm(outcome ~ ., data.1)
```

```r
outcome = data.1$outcome
outcome.predicted = predict(model.lm, data.1)
error = outcome.predicted - outcome
RMSE = sqrt(mean(error^2))

result = data.frame(outcome, outcome.predicted, error)
data.frame(RMSE)
head(result)
```

```r
class = data.2$class

integer_outcome.predicted = round(predict(model.lm, data.2))
integer_outcome.predicted[integer_outcome.predicted > 5] = 5
integer_outcome.predicted[integer_outcome.predicted < 0] = 0
class.predicted = factor(integer_outcome.predicted, levels=0:5, labels=c("terrible","bad","poor","fair","good","excellent"))

hit = class.predicted == class
accuracy = length(which(hit)) / length(hit)

result = data.frame(class, class.predicted, hit)
data.frame(accuracy)
head(result)
```

```r
near_hit = abs(as.numeric(class.predicted) - as.numeric(class)) <= 1
near_accuracy = length(which(near_hit)) / length(near_hit)

result = data.frame(class, class.predicted, near_hit)
data.frame(near_accuracy)
head(result)
```

```r
model.dt2 = rpart(class ~ ., data.2, method="class") # may take about 30 seconds
```

```r
class = data.2$class
class.predicted = predict(model.dt2, data.2, type="class")

hit = class.predicted == class
accuracy = length(which(hit)) / length(hit)

result = data.frame(class, class.predicted, hit)
data.frame(accuracy)
head(result)
```

```r
near_hit = abs(as.numeric(class.predicted) - as.numeric(class)) <= 1
near_accuracy = length(which(near_hit)) / length(near_hit)

result = data.frame(class, class.predicted, near_hit)
data.frame(near_accuracy)
head(result)
```

```r
class = data.3$class

class.predicted = factor(rep(NA, nrow(data.3)), c("positive","negative"))
class.predicted[predict(model.lm, data.3) >= 2.5] = "positive"
class.predicted[predict(model.lm, data.3) < 2.5] = "negative"

hit = class.predicted == class
accuracy = length(which(hit)) / length(hit)

result = data.frame(class, class.predicted, hit)
data.frame(accuracy)
head(result)
```

```r
model.dt3 = rpart(class ~ ., data.3, method="class") # may take about 30 seconds
```

```r
class = data.3$class

class.predicted = predict(model.dt3, data.3, type="class")

hit = class.predicted == class
accuracy = length(which(hit)) / length(hit)

result = data.frame(class, class.predicted, hit)
data.frame(accuracy)
head(result)
```

```r
class.predicted = factor(rep(NA, nrow(data.unrated)), c("positive","negative"))
class.predicted[predict(model.lm, data.unrated) >= 2.5] = "positive"
class.predicted[predict(model.lm, data.unrated) < 2.5] = "negative"
review = data.raw$reviews.text[i_unrated]

result.lm = data.frame(review, class.predicted)
```

```r
length(which(class.predicted=="positive"))
result.lm[class.predicted=="positive",]
```

```r
length(which(class.predicted=="negative"))
result.lm[class.predicted=="negative",]
```

```r
class.predicted = predict(model.dt3, data.unrated, type="class")
review = data.raw$reviews.text[i_unrated]

result.dt = data.frame(review, class.predicted)
```

```r
length(which(class.predicted=="positive"))
result.dt[class.predicted=="positive",]
```

```r
length(which(class.predicted=="negative"))
result.dt[class.predicted=="negative",]
```

```r
ambiguous = setdiff(union(result.lm[result.lm$class.predicted=="negative", "review"],
                          result.dt[result.dt$class.predicted=="negative", "review"]),
                    intersect(result.lm[result.lm$class.predicted=="negative", "review"],
                              result.dt[result.dt$class.predicted=="negative", "review"]))

data.frame(review=ambiguous)
```
