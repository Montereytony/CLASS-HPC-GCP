### Retrieve Data
___
text data retrieved as a one column table of text items - the data set is called a "corpus", each text item is called a "document" and can be a word, a phrase, a sentence, a paragraph, an article, a chapter, a book, or any other unit of text.

Here the data comprises 7 seven documents.

_**Implementation detail:** Use encoding="UTF-8" to ensure that any unusual characters in the text are interpretted correctly._  

```r
data = read.xlsx("../DATASETS/DATASET Jabberwocky.xlsx", sheetIndex=1, header=FALSE, encoding="UTF-8")
data$class = factor(c("A","A","B","A","B","B","A"), c("A","B"))
dim(data)
data
```

```r
new = data.frame(X1="The bird rested in the dead tree for a long time, and thought just as long about the joy of business analytics.")
new
```

___

<br>

<br>
#### Step 2: Convert the corpus from a table to a special kind of list.

_**Implementation detail:** To access an element of a list, use the [[...]] notation._

```r
corpus = VCorpus(VectorSource(data$X1))

inspect(corpus[[1]])
inspect(corpus[[2]])
inspect(corpus[[3]])
inspect(corpus[[4]])
inspect(corpus[[5]])
inspect(corpus[[6]])
inspect(corpus[[7]])
#inspect(corpus[[8]])
```

<br>
#### Step 3: Transform the corpus in various ways.

Here we transform the corpus in 6 ways:
* Transform all unusal characters to similar standard latin letters (necessary only on some systems).
* Transform all upper case letters to lower case letters.
* Remove all numbers.
* Remove all punctuation.
* Remove all inconsequential words, like "a", "the", "we", "in", etc.
* Remove all whitespace, like spaces, new lines, etc.

```r
#corpus = tm_map(corpus, content_transformer(iconv), from="UTF-8", to="latin1") # delete this line on some systems
corpus = tm_map(corpus, content_transformer(tolower))
corpus = tm_map(corpus, removeNumbers)
corpus = tm_map(corpus, removePunctuation, ucp=TRUE)
corpus = tm_map(corpus, removeWords, stopwords("en"))
corpus = tm_map(corpus, stripWhitespace)

inspect(corpus[[1]])
inspect(corpus[[2]])
inspect(corpus[[3]])
inspect(corpus[[4]])
inspect(corpus[[5]])
inspect(corpus[[6]])
inspect(corpus[[7]])
#inspect(corpus[[8]])
```

<br>
#### Step 4: Determine the document-term table (words).

Columns correspond to the set of all words in the transformed corpus.  Rows correspond to documents.  Each value is the number of times a certain words appears in a certain document.

Here the resulting document-term table comprises 79 columns

```r
dtm = DocumentTermMatrix(corpus)
data.t = as.data.frame(as.matrix(dtm))
dim(data.t)
data.t
```

<br>
#### Step 4 (alternative): Determine the document-term table (n-grams).

Columns correspond to the set of n-grams in the transformed corpus.  An n-gram is a set of words that appear adjecent to each other.  For example, "arms beamish" and "awhile thought" are 2-grams, or bigrams. For example, "arms" and "awhile" are 1-grams, or unigrams, or just words.

Rows correspond to documents.  Each value is the number of times a certain n-gram appears in a certain document.

Here we determine the document-term table for unigrams and bigrams.

```r
dtm = DocumentTermMatrix(corpus, control=list(tokenize=unibigram))
data.t = as.data.frame(as.matrix(dtm))
dim(data.t)
data.t
```

<br>
#### Step 4 (alternative): Determine the document-term table (n-grams) and remove sparse variables.

.aaaa......

```r
dtm = removeSparseTerms(DocumentTermMatrix(corpus, control=list(tokenize=unibigram)), sparse=0.90)
data.t =  as.data.frame(as.matrix(dtm))
dim(data.t)
data.t
```

<br>
#### Step 5: Restore classification.

....

```r
data.tc = data.t
data.tc$class = data$class
data.tc
```

### New .. .

#### .aaaa......

```r
corpus = VCorpus(VectorSource(new$X1))
#corpus = tm_map(corpus, content_transformer(iconv), from="UTF-8", to="latin1") # delete this line on some systems
corpus = tm_map(corpus, content_transformer(tolower))
corpus = tm_map(corpus, removeNumbers)
corpus = tm_map(corpus, removePunctuation, ucp=TRUE)
corpus = tm_map(corpus, removeWords, stopwords("en"))
corpus = tm_map(corpus, stripWhitespace)

dtm = DocumentTermMatrix(corpus, control=list(tokenize=unibigram))
new.t = as.data.frame(as.matrix(dtm))
new.t
```

```r
new.tr = regulate_columns(new.t, data.t)
new.tr
```

```r
model = naiveBayes(class ~ ., data.tc)
result = predict(model, new.tr, type="class")

data.frame(result)
```
