### Setup

** Configure the R software system for this analysis. **

```r
# Load some required functions
library(e1071,     verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(caret,     verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(ggplot2,   verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(gridExtra, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(rgl,       verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE) # All rglwidget() calls must use elementID="myplot"
library(htmltools, verbose=FALSE, warn.conflicts=FALSE, quietly=TRUE)
library(pracma)
library(reshape2)

options(warn=-1)
set.seed(12345)

# Set some visualization formatting defaults
options(rgl.useNULL=TRUE)
options(repr.matrix.max.cols=100)
options(repr.plot.width=3, repr.plot.height=3)
theme_set(theme_grey())
update_geom_defaults("point", list(size=2, colour="green4"))
update_geom_defaults("line", list(size=0.15))
update_geom_defaults("bar", list(colour="green4", fill="green4"))
update_geom_defaults("density", list(colour="dodgerblue", fill="dodgerblue"))
update_geom_defaults("text", list(size=2.5))
update_geom_defaults("abline", list(color="black", size=0.15))
update_geom_defaults("hline", list(color="black", size=0.15))
update_geom_defaults("vline", list(color="black", size=0.15))
update_geom_defaults("smooth", list(color="black", size=0.15))
update_geom_defaults("col", list(fill="green4"))
theme_update(plot.title=element_text(size=10, hjust=0.5), plot.subtitle=element_text(size=8, face="italic", hjust=0.5), axis.title=element_text(size=7), axis.text=element_text(size=7), strip.text=element_text(size=7), strip.text.y=element_text(angle=90), legend.title=element_blank(), legend.text=element_text(size=7))

# Define some useful visualization settings
standard_arrow = arrow(angle=15, type="closed", length=unit(10, "points"))
theme.x_axis_only = theme(axis.title.y=element_blank(), axis.text.y=element_blank(), axis.ticks.y=element_blank(), panel.grid.major.y=element_blank(), panel.grid.minor.y=element_blank())
```

### Principal Component Analysis Applied to 2 Variables
___

```r
data = data.frame(x=c(1, 2, 3, 6, 11, 14, 15, 16, 18), y=5*c(2.7, 2.3, 2.4, 2.8, 2.7, 3.2, 3.3, 3.25, 3.4)-5)
data
```

```r
options(repr.plot.width=3, repr.plot.height=3)

ggplot(data) + geom_point(aes(x=x, y=y)) + xlim(0,20) + ylim(0,20)
```

```r
options(repr.plot.width=6, repr.plot.height=1)

p1 = ggplot(data) + geom_point(aes(x=x, y=0)) + xlim(0,20) + theme.x_axis_only
p2 = ggplot(data) + geom_point(aes(x=y, y=0)) + xlim(0,20) + theme.x_axis_only
grid.arrange(p1, p2, nrow=1)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

sdev = c(sd(data$x), sd(data$y))
scree = data.frame(axis=c("x", "y"), standard_deviation=sdev, variance=sdev^2, cum_variance=cumsum(sdev^2), proportion=sdev^2/sum(sdev^2))
scree

ggplot(scree) + geom_col(aes(x=axis, y=proportion)) + ylim(0,1)
```

```r
pc = prcomp(data, retx=TRUE)

as.data.frame(pc$center)
as.data.frame(pc$rotation)
as.data.frame(pc$x)

PC1.point_1 = pc$rotation["x","PC1"]*(data$x[1]-pc$center["x"]) + pc$rotation["y", "PC1"]*(data$y[1]-pc$center["y"]) 
PC2.point_1 = pc$rotation["x","PC2"]*(data$x[1]-pc$center["x"]) + pc$rotation["y", "PC2"]*(data$y[1]-pc$center["y"]) 
data.frame(PC1=PC1.point_1, PC2=PC2.point_1, row.names=NULL)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

eigen.x.pc1 = pc$rotation["x","PC1"]
eigen.y.pc1 = pc$rotation["y","PC1"] 
eigen.x.pc2 = pc$rotation["x","PC2"]
eigen.y.pc2 = pc$rotation["y","PC2"] 
x1 = pc$center["x"]
y1 = pc$center["y"]
m.pc1 = eigen.y.pc1/eigen.x.pc1
m.pc2 = -1/m.pc1
i.pc1 = -m.pc1*x1 + y1
i.pc2 = -m.pc2*x1 + y1

ggplot(data) +
geom_point(aes(x=x, y=y)) +
geom_abline(aes(slope=m.pc1, intercept=i.pc1), color="blue") +
#geom_smooth(aes(x=x, y=y), data=data.frame(x=c(x1, x1+eigen.x.pc1), y=c(y1, y1+eigen.y.pc1)), method="lm", lwd=1, fullrange=TRUE) +
geom_abline(aes(slope=m.pc2, intercept=i.pc2), color="red") +
geom_segment(aes(x=x1, y=y1, xend=x1+5*eigen.x.pc1, yend=y1+5*eigen.y.pc1), arrow=standard_arrow, color="blue") +
geom_segment(aes(x=x1, y=y1, xend=x1+3*eigen.x.pc2, yend=y1+3*eigen.y.pc2), arrow=standard_arrow, color="red") +
geom_point(aes(x=x1, y=y1), size=1, color="black") +
xlim(0,20) + ylim(0,20)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

ggplot(as.data.frame(pc$x)) +
geom_point(aes(x=PC1, y=PC2)) +
geom_hline(aes(yintercept=0), color="blue") +
geom_vline(aes(xintercept=0), color="red") +
geom_segment(aes(x=0, y=0, xend=5*1, yend=0), arrow=standard_arrow, color="blue") +
geom_segment(aes(x=0, y=0, xend=0, yend=3*1), arrow=standard_arrow, color="red") +
geom_point(aes(x=0, y=0), size=1, color="black") +
xlim(-10,10) + ylim(-10,10)
```

```r
options(repr.plot.width=6, repr.plot.height=1)

p1 = ggplot(as.data.frame(pc$x)) + geom_point(aes(x=PC1, y=0)) + xlim(-10,10) + theme.x_axis_only
p2 = ggplot(as.data.frame(pc$x)) + geom_point(aes(x=PC2, y=0)) + xlim(-10,10) + theme.x_axis_only
grid.arrange(p1, p2, nrow=1)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

scree = data.frame(axis=c("PC1", "PC2"), standard_deviation=pc$sdev, variance=pc$sdev^2, cum_variance=cumsum(pc$sdev^2), proportion=pc$sdev^2/sum(pc$sdev^2), cum_proportion=cumsum(pc$sdev^2/sum(pc$sdev^2)))
scree

ggplot(scree) + geom_col(aes(x=axis, y=proportion)) + ylim(0,1)
```

### Principal Component Analysis Applied to 2 Variables (normalized)
___

```r
data.norm = as.data.frame(scale(data))
data.norm
```

```r
options(repr.plot.width=3, repr.plot.height=3)

ggplot(data.norm) + geom_point(aes(x=x, y=y)) + xlim(-2.5,2.5) + ylim(-2.5,2.5)
```

```r
options(repr.plot.width=6, repr.plot.height=1)

p1 = ggplot(data.norm) + geom_point(aes(x=x, y=0)) + xlim(-2.5,2.5) + theme.x_axis_only
p2 = ggplot(data.norm) + geom_point(aes(x=y, y=0)) + xlim(-2.5,2.5) + theme.x_axis_only
grid.arrange(p1, p2, nrow=1)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

sdev = c(sd(data.norm$x), sd(data.norm$y))
scree = data.frame(axis=c("x", "y"), standard_deviation=sdev, variance=sdev^2, cum_variance=cumsum(sdev^2), proportion=sdev^2/sum(sdev^2), cum_proportion=cumsum(sdev^2/sum(sdev^2)))
scree

ggplot(scree) + geom_col(aes(x=axis, y=proportion)) + ylim(0,1)
```

```r
pc.norm = prcomp(data.norm, retx=TRUE)

as.data.frame(pc.norm$center)
as.data.frame(pc.norm$rotation)
as.data.frame(pc.norm$x)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

eigen.x.pc1 = pc.norm$rotation["x","PC1"]
eigen.y.pc1 = pc.norm$rotation["y","PC1"] 
eigen.x.pc2 = pc.norm$rotation["x","PC2"]
eigen.y.pc2 = pc.norm$rotation["y","PC2"] 
x1 = pc.norm$center["x"]
y1 = pc.norm$center["y"]
m.pc1 = eigen.y.pc1/eigen.x.pc1
m.pc2 = -1/m.pc1
i.pc1 = -m.pc1*x1 + y1
i.pc2 = -m.pc2*x1 + y1

ggplot(data.norm) +
geom_point(aes(x=x, y=y)) +
geom_abline(aes(slope=m.pc1, intercept=i.pc1), color="blue") +
geom_abline(aes(slope=m.pc2, intercept=i.pc2), color="red") +
geom_segment(aes(x=x1, y=y1, xend=x1+eigen.x.pc1, yend=y1+eigen.y.pc1), arrow=standard_arrow, color="blue") +
geom_segment(aes(x=x1, y=y1, xend=x1+0.6*eigen.x.pc2, yend=y1+0.6*eigen.y.pc2), arrow=standard_arrow, color="red") +
geom_point(aes(x=x1, y=y1), size=1, color="black") +
xlim(-2.5,2.5) + ylim(-2.5,2.5)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

ggplot(as.data.frame(pc.norm$x)) +
geom_point(aes(x=PC1, y=PC2)) +
geom_hline(aes(yintercept=0), color="blue") +
geom_vline(aes(xintercept=0), color="red") +
geom_segment(aes(x=0, y=0, xend=1, yend=0), arrow=standard_arrow, color="blue") +
geom_segment(aes(x=0, y=0, xend=0, yend=0.6*1), arrow=standard_arrow, color="red") +
geom_point(aes(x=0, y=0), size=1, color="black") +
xlim(-2.5,2.5) + ylim(-2.5,2.5)
```

```r
options(repr.plot.width=6, repr.plot.height=1)

p1 = ggplot(as.data.frame(pc.norm$x)) + geom_point(aes(x=PC1, y=0)) + xlim(-2.5,2.5) + theme.x_axis_only
p2 = ggplot(as.data.frame(pc.norm$x)) + geom_point(aes(x=PC2, y=0)) + xlim(-2.5,2.5) + theme.x_axis_only
grid.arrange(p1, p2, nrow=1)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

scree = data.frame(axis=c("PC1", "PC2"), standard_deviation=pc.norm$sdev, variance=pc.norm$sdev^2, cum_variance=cumsum(pc.norm$sdev^2), proportion=pc.norm$sdev^2/sum(pc.norm$sdev^2), cum_proportion=cumsum(pc.norm$sdev^2/sum(pc.norm$sdev^2)))
scree

ggplot(scree) + geom_col(aes(x=axis, y=proportion)) + ylim(0,1)
```

### Principal Component Analysis Applied to 3 Variables, Normalized
___

```r
data.ext = cbind(data, data.frame(z=c(1, 2, 10, 13, 16, 15, 4, 8, 6)))
data.ext
```

```r
PALETTE1 = "green4"

p = open3d(); view3d(0,0, zoom=0.65)
plot3d(data.ext, type="s", col=PALETTE1, xlim=c(0,20), ylim=c(0,20), zlim=c(0,20))
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=300, height=300)
rgl.close()
```

```r
options(repr.plot.width=9, repr.plot.height=1)

p1 = ggplot(data.ext) + geom_point(aes(x=x, y=0)) + xlim(0,20) + theme.x_axis_only
p2 = ggplot(data.ext) + geom_point(aes(x=y, y=0)) + xlim(0,20) + theme.x_axis_only
p3 = ggplot(data.ext) + geom_point(aes(x=z, y=0)) + xlim(0,20) + theme.x_axis_only
grid.arrange(p1, p2, p3, nrow=1)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

sdev = c(sd(data.ext$x), sd(data.ext$y), sd(data.ext$z))
scree = data.frame(axis=c("x", "y", "z"), standard_deviation=sdev, variance=sdev^2, cum_variance=cumsum(sdev^2), proportion=sdev^2/sum(sdev^2), cum_proportion=cumsum(sdev^2/sum(sdev^2)))
scree

ggplot(scree) + geom_col(aes(x=axis, y=proportion)) + ylim(0,1)
```

```r
pc.ext = prcomp(data.ext, retx=TRUE)

as.data.frame(pc.ext$center)
as.data.frame(pc.ext$rotation)
as.data.frame(pc.ext$x)
```

```r
eigen.x.pc1 = pc.ext$rotation["x","PC1"]
eigen.y.pc1 = pc.ext$rotation["y","PC1"] 
eigen.z.pc1 = pc.ext$rotation["z","PC1"] 
eigen.x.pc2 = pc.ext$rotation["x","PC2"]
eigen.y.pc2 = pc.ext$rotation["y","PC2"]
eigen.z.pc2 = pc.ext$rotation["z","PC2"]
eigen.x.pc3 = pc.ext$rotation["x","PC3"]
eigen.y.pc3 = pc.ext$rotation["y","PC3"]
eigen.z.pc3 = pc.ext$rotation["z","PC3"]

x1 = pc.ext$center["x"]
y1 = pc.ext$center["y"]
z1 = pc.ext$center["z"]

center = data.frame(x=x1, y=y1, z=z1)

p = open3d(); view3d(0,0, zoom=0.65)
plot3d(data.ext, type="s", col=PALETTE1, xlim=c(0,20), ylim=c(0,20), zlim=c(0,20))
spheres3d(center, type="s", radius=0.75, add=TRUE)
abclines3d(x1, y1, z1, eigen.x.pc1, eigen.y.pc1, eigen.z.pc1, color="blue", add=TRUE)
abclines3d(x1, y1, z1, eigen.x.pc2, eigen.y.pc2, eigen.z.pc2, color="red", add=TRUE)
abclines3d(x1, y1, z1, eigen.x.pc3, eigen.y.pc3, eigen.z.pc3, color="violet", add=TRUE)

arrow3d(p0=c(x1,y1,z1), p1=c(x1+7*eigen.x.pc1,y1+7*eigen.y.pc1,z1+7*eigen.z.pc1), type="rotation", color="blue", add=TRUE)
arrow3d(p0=c(x1,y1,z1), p1=c(x1+5*eigen.x.pc2,y1+5*eigen.y.pc2,z1+5*eigen.z.pc2), type="rotation", color="red", add=TRUE)
arrow3d(p0=c(x1,y1,z1), p1=c(x1+3*eigen.x.pc3,y1+3*eigen.y.pc3,z1+3*eigen.z.pc3), type="rotation", color="violet", add=TRUE)

grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=300, height=300)
rgl.close()
```

```r
options(repr.plot.width=9, repr.plot.height=1)

p1 = ggplot(as.data.frame(pc.ext$x)) + geom_point(aes(x=PC1, y=0)) + xlim(-10,10) + theme.x_axis_only
p2 = ggplot(as.data.frame(pc.ext$x)) + geom_point(aes(x=PC2, y=0)) + xlim(-10,10) + theme.x_axis_only
p3 = ggplot(as.data.frame(pc.ext$x)) + geom_point(aes(x=PC3, y=0)) + xlim(-10,10) + theme.x_axis_only
grid.arrange(p1, p2, p3, nrow=1)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

scree = data.frame(axis=c("PC1", "PC2", "PC3"), standard_deviation=pc.ext$sdev, variance=pc.ext$sdev^2, cum_variance=cumsum(pc.ext$sdev^2), proportion=pc.ext$sdev^2/sum(pc.ext$sdev^2), cum_proportion=cumsum(pc.ext$sdev^2/sum(pc.ext$sdev^2)))
scree

ggplot(scree) + geom_col(aes(x=axis, y=proportion)) + ylim(0,1)
```
