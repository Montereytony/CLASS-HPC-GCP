```r
data = read.csv("../DATASETS/DATASET Lending Tree Loans.csv")
dim(data)
head(data)
```

```r
unique(data$loan_status)
```

```r
str(data)
```

```r
data.ok = data[,aaply(1:ncol(data), 1, function(i) !any(is.na(data[,i])))]
```

```r
library(plyr)
data.numeric = data.ok[,aaply(1:ncol(data.ok), 1, function(i) {class(data.ok[,i])}) != "factor"]

dim(data.numeric)
head(data.numeric)
```

```r
unique(data.numeric[,"policy_code"])
data.go = data.numeric[,-20]
```

```r
pc = prcomp(data.go, scale=TRUE, retx=TRUE)
as.data.frame(pc$center)
as.data.frame(pc$rotation)
ID = names(as.data.frame(pc$rotation))
ID
```

```r
scree = data.frame(axis=ID, standard_deviation=pc$sdev, variance=pc$sdev^2, cum_variance=cumsum(pc$sdev^2), proportion=pc$sdev^2/sum(pc$sdev^2), cum_proportion=cumsum(pc$sdev^2/sum(pc$sdev^2)))
scree

options(repr.plot.width=8, repr.plot.height=3)

ggplot(scree) + geom_col(aes(x=factor(axis, levels=axis), y=proportion)) + ylim(0,1)
```

```r
pcx = as.data.frame(pc$x)
head(pcx)
nrow(pcx)
```

```r
bad  =   data$loan_status %in% c("Charged Off", "Status:Charged Off")
good = !(data$loan_status %in% c("Charged Off", "Status:Charged Off"))

pcx$status = NA
pcx$status[bad] = "bad"
pcx$status[good] = "good"

ggplot(pcx[,]) +
geom_point(aes(x=PC1, y=PC2, color=status),, size=0.01)
```
