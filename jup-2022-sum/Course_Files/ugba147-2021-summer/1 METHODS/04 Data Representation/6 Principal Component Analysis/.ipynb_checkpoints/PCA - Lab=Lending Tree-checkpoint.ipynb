#### Reform data.
Add an **outcome** column for a simplified loan status (good vs. bad), and focus on the numeric columns with variation but no missing values.

```r
data.status = data.frame(status=factor(data$loan_status %in% c("Charged Off", "Status:Charged Off"), labels=c("good", "bad")))
data.misc = data[, full_columns & num_columns & vary_columns]
data.ok = cbind(data.status, data.misc)
dim(data.ok)
head(data.ok)
```

<br>
### Explore Relationship of Loan Status to Other Variables 
___

#### Explore column variance.

Find the columns with the most the variance. Perhaps these columns contain the most useful information about loan status.

```r
axis = names(data.ok[,-(1:3)])
sd = aaply(names(data.ok[,-(1:3)]), 1, function(j) { sd(data.ok[,j]) })
variance = sd^2
cum_variance = cumsum(variance)
proportion = variance / sum(variance)
cum_proportion = cumsum(proportion)

scree = data.frame(axis, sd, variance, cum_variance, proportion, cum_proportion)
scree

ggplot(scree) + geom_col(aes(x=factor(axis, levels=axis[rev(order(variance))]), y=proportion)) + ylim(0,1) +
theme(axis.text.x=element_text(hjust=1, angle=45))
```

<br>
#### Revolving Balance & Outstanding Principal
These 2 columns contribute the most variance.  Visualize distributions of revolving balance, outstanding principal, and loan status.

```r
# plot may take about 1 minute 
ggplot() +
geom_point(aes(x=revol_bal, y=out_prncp, color="good"), data=data.ok[data.ok$status=="good",]) +
geom_point(aes(x=revol_bal, y=out_prncp, color="bad"),  data=data.ok[data.ok$status=="bad",]) +
scale_color_manual(values=c(bad=PALETTE[2], good=PALETTE[3]))
```

<br>
Revolving balances for bad loans are concentrated at low amounts, but so too for good loans, so that does not distinguish them well.

```r
p1 = ggplot(data.ok[data.ok$status=="good",]) +
     geom_histogram(aes(revol_bal), bins=50, na.rm=TRUE, fill=PALETTE[3]) +
     xlim(0,200000) + ylim(0,175000)

p2 = ggplot(data.ok[data.ok$status=="bad",]) +
     geom_histogram(aes(revol_bal), bins=50, na.rm=TRUE, fill=PALETTE[2]) +
     xlim(0,200000) + ylim(0,10000)

p3 = ggplot(data.ok[data.ok$status=="bad",]) +
     geom_histogram(aes(revol_bal), bins=50, na.rm=TRUE, fill=PALETTE[2]) +
     xlim(0,200000) + ylim(0,175000)

grid.arrange(p1, p2, p3, nrow=1)
```

### Principal Components
___

#### Focus only on interesting variables.
Exclude outstanding principal and outstanding principal invested.

```r
data.okx = data.ok
data.okx$out_prncp = NULL
data.okx$out_prncp_inv = NULL
```

<br>
#### Calculate principal components.
Normalize data before calculating principal components.

```r
pc = prcomp(data.okx[,-(1:3)], scale=TRUE, retx=TRUE)
```

<br>
#### Interpret principal components with respect to normalized variables.

```r
pc$rotation
```

```r
# Qualitative interpretation of principal components
# Each column contains variable names sorted by weight applied to the principal component

pc_names = names(as.data.frame(pc$rotation))
qual = as.data.frame(t(adply(pc_names, 1, function(j) row.names(pc$rotation)[rev(order(abs(pc$rotation[,j])))])[-1]))
names(qual) = pc_names
row.names(qual) = NULL
qual
```

<br>
#### Explore dataset expressed in terms of its principal components.

```r
data.okx.pc = cbind(data.okx[,1:3], as.data.frame(pc$x))
dim(data.okx.pc)
head(data.okx.pc)
```

```r
# Confirm that PC values are weighted sum of normalized original variable values
data.okx.pc[1,"PC1"]
sum(scale(data.okx[,-(1:3)])[1,] * pc$rotation[,"PC1"])
```

<br>
#### Explore relative variation in principal components.

```r
axis = names(as.data.frame(pc$rotation))
scree = data.frame(axis=axis, standard_deviation=pc$sdev, variance=pc$sdev^2, cum_variance=cumsum(pc$sdev^2), proportion=pc$sdev^2/sum(pc$sdev^2), cum_proportion=cumsum(pc$sdev^2/sum(pc$sdev^2)))
scree

ggplot(scree) + geom_col(aes(x=factor(axis, levels=axis), y=proportion)) + ylim(0,1)
```

### Explore Relationship of Loan Status to Principal Components
___

#### Explore relationship of PC1 vs PC2.

```r
# plot may take about 1 minute 
ggplot() +
geom_point(aes(x=PC1, y=PC2, color="good"), data=data.okx.pc[data.okx.pc$status=="good",]) +
geom_point(aes(x=PC1, y=PC2, color="bad"),  data=data.okx.pc[data.okx.pc$status=="bad",]) +
scale_color_manual(values=c(bad=PALETTE[2], good=PALETTE[3]))
```

<br>
#### Explore distribution of PC1.


```r
p1 = ggplot(data.okx.pc[data.okx.pc$status=="good",]) +
     geom_histogram(aes(PC1), bins=50, na.rm=TRUE, fill=PALETTE[3]) +
     xlim(-5,12) + ylim(0,70000)

p2 = ggplot(data.okx.pc[data.okx.pc$status=="bad",]) +
     geom_histogram(aes(PC1), bins=50, na.rm=TRUE, fill=PALETTE[2]) +
     xlim(-5,12) + ylim(0,4000)

p3 = ggplot(data.okx.pc[data.okx.pc$status=="bad",]) +
     geom_histogram(aes(PC1), bins=50, na.rm=TRUE, fill=PALETTE[2]) +
     xlim(-5,12) + ylim(0,70000)

grid.arrange(p1, p2, p3, nrow=1)
```

<br>
#### Explore distribution of PC2.

Loans with PC2 < -1.3 may more than likely be good loans.  Loans with PC2 > 4.2 may more than likely be bad loans.

```r
p1 = ggplot(data.okx.pc[data.okx.pc$status=="good",]) +
     geom_histogram(aes(PC2), bins=50, na.rm=TRUE, fill=PALETTE[3]) +
     geom_vline(aes(xintercept=-1.3), lty="dashed") +
     geom_vline(aes(xintercept=4.2), lty="dashed") +
     xlim(-5,10) + ylim(0,90000)

p2 = ggplot(data.okx.pc[data.okx.pc$status=="bad",]) +
     geom_histogram(aes(PC2), bins=50, na.rm=TRUE, fill=PALETTE[2]) +
     geom_vline(aes(xintercept=-1.3), lty="dashed") +
     geom_vline(aes(xintercept=4.2), lty="dashed") +
     xlim(-5,10) + ylim(0,6000)

p3 = ggplot(data.okx.pc[data.okx.pc$status=="bad",]) +
     geom_histogram(aes(PC2), bins=50, na.rm=TRUE, fill=PALETTE[2]) +
     geom_vline(aes(xintercept=-1.3), lty="dashed") +
     geom_vline(aes(xintercept=4.2), lty="dashed") +
     xlim(-5,10) + ylim(0,90000)

grid.arrange(p1, p2, p3, nrow=1)
```

<br>
#### Explore distribution of PC3.

Observations with PC3 > 2 may more than likely be bad loans.

```r
p1 = ggplot(data.okx.pc[data.okx.pc$status=="good",]) +
     geom_histogram(aes(PC3), bins=50, na.rm=TRUE, fill=PALETTE[3]) +
     geom_vline(aes(xintercept=2), lty="dashed") +
     xlim(-5,15) + ylim(0,350000)

p2 = ggplot(data.okx.pc[data.okx.pc$status=="bad",]) +
     geom_histogram(aes(PC3), bins=50, na.rm=TRUE, fill=PALETTE[2]) +
     geom_vline(aes(xintercept=2), lty="dashed") +
     xlim(-5,15) + ylim(0,11000)

p3 = ggplot(data.okx.pc[data.okx.pc$status=="bad",]) +
     geom_histogram(aes(PC3), bins=50, na.rm=TRUE, fill=PALETTE[2]) +
     geom_vline(aes(xintercept=2), lty="dashed") +
     xlim(-5,15) + ylim(0,350000)

grid.arrange(p1, p2, p3, nrow=1)
```
